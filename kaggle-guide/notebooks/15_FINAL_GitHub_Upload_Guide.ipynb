## 3. Git 초기화 & GitHub 업로드

### 3.1 로컬 Git 초기화

```bash
cd kaggle-guide                  # 프로젝트 폴더로 이동

git init                         # Git 저장소 초기화
git branch -M main               # 기본 브랜치 이름을 main으로

git add .                        # 모든 파일 stage
git status                       # 뭐가 들어갈지 확인 ← 중요!

git commit -m "initial commit: Kaggle 30-day guide"
```

### 3.2 GitHub에서 Repo 생성

#### 방법 A: 웹에서 생성 (누구나 가능)
1. https://github.com/new 접속
2. Repository name: `kaggle-guide`
3. Public / Private 선택
4. **README, .gitignore, license 추가하지 말 것** (로컬에 이미 있음)
5. Create repository 클릭

#### 방법 B: GitHub CLI로 한 방에 (추천)

```bash
# Public
gh repo create kaggle-guide --public --source=. --remote=origin --push

# Private
gh repo create kaggle-guide --private --source=. --remote=origin --push
```

→ 이 명령 하나로 **생성 + remote 연결 + push** 끝.

### 3.3 Remote 연결 & Push (방법 A 선택 시)

GitHub에서 repo 만들면 아래 같은 안내가 나옵니다:

```bash
git remote add origin https://github.com/<YOUR_USERNAME>/kaggle-guide.git
git push -u origin main
```

이 때 **Personal Access Token** 으로 인증:
- https://github.com/settings/tokens/new
- Scopes: `repo` 체크
- 생성한 토큰을 비밀번호 대신 입력

## 4. 이후 업데이트하는 방법

새 파일을 추가했거나 기존 파일을 수정한 뒤:

### Commit 메시지 규칙 (추천)
| 접두사 | 의미 |
|--------|------|
| `feat:` | 새 기능/내용 추가 |
| `fix:` | 버그/오류 수정 |
| `docs:` | 문서(README 등) 수정 |
| `refactor:` | 코드 리팩토링 |
| `chore:` | 기타 잡무 |

## 5. 민감 정보 보호 — 절대 커밋하면 안 되는 것

### 🚨 절대 금지
- `kaggle.json` (API 토큰)
- `.env` 파일 (환경변수, 비밀번호)
- OAuth secret, API key
- 대용량 데이터 CSV/parquet (보통 100MB 넘으면 GitHub가 거부)

### 이미 실수로 올렸다면?
```bash
# 1) 파일 삭제 + 커밋
git rm --cached kaggle.json
echo "kaggle.json" >> .gitignore
git commit -m "chore: remove sensitive file"

# 2) 과거 이력에서도 제거 (주의: 강제 푸시)
git filter-branch --force --index-filter \
  "git rm --cached --ignore-unmatch kaggle.json" \
  --prune-empty --tag-name-filter cat -- --all
git push --force --all

# 3) 노출된 토큰은 즉시 Kaggle에서 재발급!
# https://www.kaggle.com/settings 에서 Expire API Token
```

> 💡 더 안전한 방법: [BFG Repo-Cleaner](https://rtyley.github.io/bfg-repo-cleaner/) 사용

### 6.2 Kaggle Notebook → GitHub 푸시 (수동)
1. Kaggle Notebook 다운로드 (.ipynb)
2. 로컬 저장소 `notebooks/` 에 복사
3. `git add / commit / push`

### 6.3 GitHub Actions로 자동 실행 (고급)

`.github/workflows/test.yml` 작성:

```yaml
name: Test Notebooks

on: [push, pull_request]

jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with:
          python-version: '3.11'
      - name: Install deps
        run: |
          pip install -r requirements.txt
          pip install nbconvert jupyter
      - name: Execute notebooks
        run: |
          for nb in notebooks/0*.ipynb; do
            jupyter nbconvert --to notebook --execute "$nb" \
              --output "/tmp/$(basename $nb)" || exit 1
          done
```

Push 때마다 모든 노트북이 자동 실행되어 **에러 감지** 가능.
