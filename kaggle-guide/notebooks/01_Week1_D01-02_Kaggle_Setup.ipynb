## 1. Kaggle이란?

Kaggle은 세계 최대의 **데이터 사이언스 커뮤니티**이자 **머신러닝 대회 플랫폼**입니다.

### Kaggle 핵심 요소
| 요소 | 설명 |
|------|------|
| **Competitions** | 실제 기업·연구 문제를 푸는 대회 (상금·메달) |
| **Datasets** | 공개 데이터셋 저장소 |
| **Notebooks** | 클라우드 Jupyter (GPU/TPU 주당 30~40h 무료) |
| **Discussions** | 대회별 토론, 솔루션 공유 |
| **Courses** | 무료 ML 강의 |

### 대회 등급
- **Featured**: 상금 대회 (최고 난이도)
- **Research**: 학술 연구 성격
- **Playground**: 연습용 (월간 Tabular Playground Series 추천)
- **Getting Started**: 입문자용 (Titanic, House Prices, Digit Recognizer)

### 메달 & 티어
Novice → Contributor → Expert → Master → Grandmaster

메달 기준 (대회):
- **Bronze**: 상위 40%
- **Silver**: 상위 20%
- **Gold**: 상위 10팀 + 0.2% × N (N=참가팀)

## 3. Kaggle API 설정 (매우 중요)

웹에서 매번 다운로드하지 않고 CLI로 바로 데이터를 받습니다.

### 3.1 API 토큰 발급
1. https://www.kaggle.com 로그인
2. 우측 상단 프로필 → **Settings**
3. **API** 섹션 → **Create New Token**
4. `kaggle.json` 파일이 자동 다운로드됨

### 3.2 토큰 파일 배치
```bash
# Linux/Mac
mkdir -p ~/.kaggle
mv ~/Downloads/kaggle.json ~/.kaggle/
chmod 600 ~/.kaggle/kaggle.json

# Windows PowerShell
mkdir $env:USERPROFILE\.kaggle
mv $env:USERPROFILE\Downloads\kaggle.json $env:USERPROFILE\.kaggle\
```

### 3.3 동작 확인

### 3.4 자주 쓰는 Kaggle CLI

```bash
# 대회 검색
kaggle competitions list -s [키워드]

# 대회 데이터 다운로드
kaggle competitions download -c [대회명] -p [경로]

# 제출
kaggle competitions submit -c [대회명] -f submission.csv -m "설명"

# 내 제출 내역
kaggle competitions submissions -c [대회명]

# 리더보드
kaggle competitions leaderboard -c [대회명] -s

# 데이터셋 다운로드
kaggle datasets download -d [user/dataset]
```

## 4. 대회 선택 전략 (초보자용)

### 🥇 입문 순서 (추천)
1. **Titanic** (이진 분류) - 생존 예측, 가장 유명한 입문 대회
2. **House Prices** (회귀) - Feature Engineering 최적
3. **Digit Recognizer** (이미지) - MNIST 기반 딥러닝 입문
4. **Playground Series** (월간) - 실전 감각 쌓기

### 대회 페이지에서 반드시 볼 것
- **Overview**: 문제 정의·배경
- **Data**: 피처 의미·설명
- **Evaluation**: **평가 지표** ← 가장 중요
- **Rules**: 제출 횟수·외부 데이터 허용 여부
- **Discussion**: 상위권 팁, 데이터 누락 정보
- **Code**: 공개 노트북 (상위권 베이스라인 학습)

### 🔑 대회 성공의 5가지 핵심
1. **평가지표 정확히 이해** - RMSE vs MAE, AUC vs Accuracy
2. **CV 전략** - 리더보드 말고 본인의 Cross-Validation 신뢰
3. **Feature Engineering** - 모델보다 피처가 더 중요한 경우가 많음
4. **앙상블** - 상위권은 대부분 여러 모델 조합
5. **Discussion 정독** - 혼자 풀지 말 것
