```python
# ğŸ¯ Pandas vs NumPy ìš©ë„ ì°¨ì´ì  ì™„ì „ ê°€ì´ë“œ

## ğŸ“Š í•µì‹¬ ì°¨ì´ì  ìš”ì•½

### NumPy (Numerical Python)
- **ëª©ì **: ìˆ˜ì¹˜ ê³„ì‚° ë° ë°°ì—´ ì—°ì‚°
- **ì£¼ìš” ë°ì´í„° êµ¬ì¡°**: ndarray (ë‹¤ì°¨ì› ë°°ì—´)
- **íŠ¹ì§•**: ë¹ ë¥¸ ìˆ˜í•™ì  ì—°ì‚°, ë©”ëª¨ë¦¬ íš¨ìœ¨ì„±
- **ìš©ë„**: ê³¼í•™ ê³„ì‚°, ì„ í˜•ëŒ€ìˆ˜, í†µê³„ ì—°ì‚°

### Pandas (Panel Data)
- **ëª©ì **: ë°ì´í„° ë¶„ì„ ë° ì¡°ì‘
- **ì£¼ìš” ë°ì´í„° êµ¬ì¡°**: Series, DataFrame
- **íŠ¹ì§•**: ë°ì´í„° ë¼ë²¨ë§, ê²°ì¸¡ì¹˜ ì²˜ë¦¬, ê·¸ë£¹í™”
- **ìš©ë„**: ë°ì´í„° ì „ì²˜ë¦¬, ë¶„ì„, ì‹œê°í™”

## ğŸ¯ ì–¸ì œ ë¬´ì—‡ì„ ì‚¬ìš©í• ê¹Œ?

| ìƒí™© | NumPy ì‚¬ìš© | Pandas ì‚¬ìš© |
|------|-----------|-------------|
| ìˆ˜í•™ ê³„ì‚° | âœ… í–‰ë ¬ ì—°ì‚°, ì‚¼ê°í•¨ìˆ˜ | âŒ ë³µì¡í•¨ |
| ë°ì´í„° ë¶„ì„ | âŒ ë¼ë²¨ ì—†ìŒ | âœ… ì¸ë±ìŠ¤, ì»¬ëŸ¼ëª… |
| íŒŒì¼ ì…ì¶œë ¥ | âŒ ì œí•œì  | âœ… CSV, Excel, JSON |
| ê²°ì¸¡ì¹˜ ì²˜ë¦¬ | âŒ ì§ì ‘ êµ¬í˜„ | âœ… fillna(), dropna() |
| ê·¸ë£¹ ë¶„ì„ | âŒ ë³µì¡í•¨ | âœ… groupby() |
| ì‹œê³„ì—´ ë¶„ì„ | âŒ ê¸°ë³¸ ê¸°ëŠ¥ ì—†ìŒ | âœ… ë‚ ì§œ ì¸ë±ìŠ¤ ì§€ì› |
```

```python
print("=== ğŸ“Š NumPy vs Pandas ê¸°ë³¸ êµ¬ì¡° ë¹„êµ ===")
import numpy as np
import pandas as pd

# ë™ì¼í•œ ë°ì´í„°ë¥¼ ë‘ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¡œ í‘œí˜„
data = [10, 20, 30, 40, 50]

print("ğŸ”¢ NumPy Array:")
numpy_array = np.array(data)
print(f"   ë°ì´í„°: {numpy_array}")
print(f"   íƒ€ì…: {type(numpy_array)}")
print(f"   í˜•íƒœ: {numpy_array.shape}")
print(f"   ë°ì´í„° íƒ€ì…: {numpy_array.dtype}")

print("\nğŸ“Š Pandas Series:")
pandas_series = pd.Series(data, index=['A', 'B', 'C', 'D', 'E'], name='values')
print(pandas_series)
print(f"   íƒ€ì…: {type(pandas_series)}")
print(f"   ì¸ë±ìŠ¤: {pandas_series.index.tolist()}")
print(f"   ì´ë¦„: {pandas_series.name}")

print("\n=== ğŸ¯ ì ‘ê·¼ ë°©ì‹ ì°¨ì´ ===")
print("NumPy - ìœ„ì¹˜ ê¸°ë°˜ ì ‘ê·¼:")
print(f"   ì²« ë²ˆì§¸ ìš”ì†Œ: {numpy_array[0]}")
print(f"   ë§ˆì§€ë§‰ ìš”ì†Œ: {numpy_array[-1]}")
print(f"   ìŠ¬ë¼ì´ì‹±: {numpy_array[1:4]}")

print("\nPandas - ë¼ë²¨ ê¸°ë°˜ ì ‘ê·¼:")
print(f"   'A' ì¸ë±ìŠ¤: {pandas_series['A']}")
print(f"   'E' ì¸ë±ìŠ¤: {pandas_series['E']}")
print(f"   'B'~'D' êµ¬ê°„: {pandas_series['B':'D'].values}")

print("\n=== ğŸ§® ì—°ì‚° ë°©ì‹ ì°¨ì´ ===")
print("NumPy - ìˆœìˆ˜ ìˆ˜ì¹˜ ì—°ì‚°:")
numpy_squared = numpy_array ** 2
print(f"   ì œê³±: {numpy_squared}")
print(f"   í‰ê· : {np.mean(numpy_array):.2f}")
print(f"   í‘œì¤€í¸ì°¨: {np.std(numpy_array):.2f}")

print("\nPandas - ì¸ë±ìŠ¤ ìœ ì§€ ì—°ì‚°:")
pandas_squared = pandas_series ** 2
print("   ì œê³±:")
print(pandas_squared)
print(f"   í‰ê· : {pandas_series.mean():.2f}")
print(f"   í‘œì¤€í¸ì°¨: {pandas_series.std():.2f}")
```

```python
print("\n=== ğŸ” ë°ì´í„° ì²˜ë¦¬ ë°©ì‹ ì°¨ì´ ===")

# ê²°ì¸¡ì¹˜ê°€ í¬í•¨ëœ ë°ì´í„°
data_with_nan = [10, np.nan, 30, 40, np.nan]

print("ğŸ”¢ NumPy - ê²°ì¸¡ì¹˜ ì²˜ë¦¬:")
numpy_nan = np.array(data_with_nan)
print(f"   ì›ë³¸: {numpy_nan}")
print(f"   í‰ê·  (NaN í¬í•¨): {np.mean(numpy_nan)}")  # NaN
print(f"   í‰ê·  (NaN ì œì™¸): {np.nanmean(numpy_nan):.2f}")
# NumPyëŠ” ê²°ì¸¡ì¹˜ ì²˜ë¦¬ë¥¼ ìœ„í•´ ë³„ë„ í•¨ìˆ˜ í•„ìš”

print("\nğŸ“Š Pandas - ê²°ì¸¡ì¹˜ ì²˜ë¦¬:")
pandas_nan = pd.Series(data_with_nan, index=['A', 'B', 'C', 'D', 'E'])
print(pandas_nan)
print(f"   í‰ê·  (ìë™ NaN ì œì™¸): {pandas_nan.mean():.2f}")
print("   ê²°ì¸¡ì¹˜ ì œê±°:")
print(pandas_nan.dropna())
print("   ê²°ì¸¡ì¹˜ ì±„ìš°ê¸° (0ìœ¼ë¡œ):")
print(pandas_nan.fillna(0))

print("\n=== ğŸ“ˆ ì‹¤ë¬´ ì‚¬ìš© ì˜ˆì‹œ ===")
print("NumPy ì í•©í•œ ê²½ìš°:")
print("   âœ… ìˆ˜í•™ì  ê³„ì‚°: í–‰ë ¬ ê³±ì…ˆ, ì„ í˜•ëŒ€ìˆ˜")
print("   âœ… ê³¼í•™ ê³„ì‚°: ë¬¼ë¦¬í•™, í†µê³„í•™ ê³µì‹")
print("   âœ… ì´ë¯¸ì§€ ì²˜ë¦¬: í”½ì…€ ë°°ì—´ ì—°ì‚°")
print("   âœ… ë¨¸ì‹ ëŸ¬ë‹: ê°€ì¤‘ì¹˜ í–‰ë ¬ ì—°ì‚°")

print("\nPandas ì í•©í•œ ê²½ìš°:")
print("   âœ… ë°ì´í„° ë¶„ì„: CSV íŒŒì¼ ì½ê¸°/ì“°ê¸°")
print("   âœ… ë¹„ì¦ˆë‹ˆìŠ¤ ë¶„ì„: ë§¤ì¶œ, ê³ ê° ë°ì´í„° ë¶„ì„")
print("   âœ… ë°ì´í„° ì „ì²˜ë¦¬: ê²°ì¸¡ì¹˜, ì¤‘ë³µê°’ ì²˜ë¦¬")
print("   âœ… ë¦¬í¬íŒ…: ê·¸ë£¹ë³„ ì§‘ê³„, í”¼ë²— í…Œì´ë¸”")

print("\n=== âš¡ ì„±ëŠ¥ ë¹„êµ ===")
import time

# ëŒ€ìš©ëŸ‰ ë°ì´í„°ë¡œ ì„±ëŠ¥ í…ŒìŠ¤íŠ¸
large_data = np.random.randint(1, 100, 100000)

# NumPy ì—°ì‚° ì„±ëŠ¥
start = time.time()
numpy_result = large_data ** 2
numpy_mean = np.mean(numpy_result)
numpy_time = time.time() - start

# Pandas ì—°ì‚° ì„±ëŠ¥
pandas_large = pd.Series(large_data)
start = time.time()
pandas_result = pandas_large ** 2
pandas_mean = pandas_result.mean()
pandas_time = time.time() - start

print(f"\nëŒ€ìš©ëŸ‰ ë°ì´í„° (100,000ê°œ) ì—°ì‚°:")
print(f"   NumPy ì‹œê°„: {numpy_time:.6f}ì´ˆ")
print(f"   Pandas ì‹œê°„: {pandas_time:.6f}ì´ˆ")
print(f"   ì†ë„ ë¹„ìœ¨: NumPyê°€ {pandas_time/numpy_time:.2f}ë°° ë¹ ë¦„")
print(f"   ê²°ê³¼ ë™ì¼: {abs(numpy_mean - pandas_mean) < 0.001}")
```

```python
# using a list of lists to create a DataFrame
data = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]
columns = ['Name', 'Age', 'City']
df = pd.DataFrame(data, columns=columns)
df
```

```python
# changing Sereies to DataFrame
data = [10, 20, 30, 40, 50, 60]
series = pd.Series(data, name='my_series')
df = pd.DataFrame(series)
print('using dataframe initialize:', df)

df2 = series.to_frame()
print('using to_frame:', df2)
```

```python
# cancat() using pandas
data1 = [1, 2, 3]
data2 = [4, 5, 6]
series1 = pd.Series(data1, name='Series1')
series2 = pd.Series(data2, name='Series2')
result_series = pd.concat([series1, series2], axis=1) # axis=1: ì—´ ë°©í–¥ìœ¼ë¡œ ì—°ê²°
print('Concatenated Series:\n', result_series)

result_series2 = pd.concat([series1, series2], axis=0) # axis=0: í–‰ ë°©í–¥ìœ¼ë¡œ ì—°ê²°
print('Concatenated Series:\n', result_series2)
```

```python
series3 = pd.Series([1,2,3], name='Series3')
df = pd.concat([result_series, series3], axis=1)
print('Concatenated DataFrame:\n', df)
```

```python
concatenated = pd.concat([df1, df2], axis=0)  # axis=0: í–‰ ë°©í–¥ìœ¼ë¡œ ì—°ê²°
print("Concatenated DataFrame (í–‰ ë°©í–¥):")
display(concatenated)
```

```python
# inner join
# ì²« ë²ˆì§¸ ë°ì´í„°í”„ë ˆì„ ìƒì„±
df1 = pd.DataFrame({'ID': [1, 2, 3],
                    'name': ['ì² ìˆ˜', 'ì˜í¬', 'ë¯¼ìˆ˜']})

# ë‘ ë²ˆì§¸ ë°ì´í„°í”„ë ˆì„ ìƒì„±
df2 = pd.DataFrame({'ID': [1, 2, 4],
                    'age': [25, 30, 28]})

display(df1)
display(df2)

# 'í•™ìƒID' ì—´ì„ ê¸°ì¤€ìœ¼ë¡œ ë‘ ë°ì´í„°í”„ë ˆì„ ì¡°ì¸
joined_df = pd.merge(df1, df2, on='ID', how='inner')
joined_df
```

```python
# left join
# 'í•™ìƒID' ì—´ì„ ê¸°ì¤€ìœ¼ë¡œ Left ì¡°ì¸
display(df1)
display(df2)
left_joined_df = pd.merge(df1, df2, on='ID', how='left')
left_joined_df
```

```python
display(df1)
display(df2)
outer_joined_df = pd.merge(df1, df2, on='ID', how='outer')
outer_joined_df
```

- **ì‚¬ìš© ì‹œê¸°**: ì™„ì „íˆ ë§¤ì¹­ë˜ëŠ” ë°ì´í„°ë§Œ í•„ìš”í•  ë•Œ

### 2ï¸âƒ£ Left Join (ì¢Œì¸¡ ì¡°ì¸)  
- **íŠ¹ì§•**: **ì™¼ìª½ DataFrameì˜ ëª¨ë“  í–‰** + ì˜¤ë¥¸ìª½ì—ì„œ ë§¤ì¹­ë˜ëŠ” í–‰
- **ê²°ê³¼**: ì™¼ìª½ ê¸°ì¤€, ë§¤ì¹­ ì•ˆ ë˜ë©´ NaN
- **ì‚¬ìš© ì‹œê¸°**: ê¸°ì¤€ ë°ì´í„°ë¥¼ ëª¨ë‘ ìœ ì§€í•˜ë©´ì„œ ì¶”ê°€ ì •ë³´ë¥¼ ê²°í•©í•  ë•Œ

### 3ï¸âƒ£ Right Join (ìš°ì¸¡ ì¡°ì¸)
- **íŠ¹ì§•**: **ì˜¤ë¥¸ìª½ DataFrameì˜ ëª¨ë“  í–‰** + ì™¼ìª½ì—ì„œ ë§¤ì¹­ë˜ëŠ” í–‰  
- **ê²°ê³¼**: ì˜¤ë¥¸ìª½ ê¸°ì¤€, ë§¤ì¹­ ì•ˆ ë˜ë©´ NaN
- **ì‚¬ìš© ì‹œê¸°**: Left Joinì˜ ë°˜ëŒ€ (ì‹¤ë¬´ì—ì„œëŠ” Left Joinì„ ë” ë§ì´ ì‚¬ìš©)

### 4ï¸âƒ£ Outer Join (ì™¸ë¶€ ì¡°ì¸)

- **ê²°ê³¼**: í•©ì§‘í•©ê³¼ ê°™ì€ ê°œë…
- **ì‚¬ìš© ì‹œê¸°**: ëª¨ë“  ë°ì´í„°ë¥¼ ë³´ì¡´í•˜ë©´ì„œ ê²°í•©í•  ë•Œ

```python
print("=== ğŸ¯ Join ì—°ì‚° ì‹¤ìŠµ ë°ì´í„° ì¤€ë¹„ ===")
import pandas as pd

# í•™ìƒ ì •ë³´ ë°ì´í„°í”„ë ˆì„ (ì™¼ìª½)
students_df = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'name': ['ê¹€ì² ìˆ˜', 'ì´ì˜í¬', 'ë°•ë¯¼ìˆ˜', 'ì •ì§€ì˜', 'ì¥ìˆ˜ë¯¼'],
    'grade': ['A', 'B', 'A', 'C', 'B']
})

# ì ìˆ˜ ì •ë³´ ë°ì´í„°í”„ë ˆì„ (ì˜¤ë¥¸ìª½)  
scores_df = pd.DataFrame({
    'student_id': [1, 2, 3, 6, 7],
    'math_score': [95, 87, 92, 89, 84],
    'english_score': [88, 94, 85, 91, 87]
})

print("ğŸ“š í•™ìƒ ì •ë³´ (students_df):")
display(students_df)
print("\nğŸ“Š ì ìˆ˜ ì •ë³´ (scores_df):")
display(scores_df)

print("\nğŸ” ë°ì´í„° ë¶„ì„:")
print(f"â€¢ í•™ìƒ ì •ë³´ ë°ì´í„°: {len(students_df)}ëª…")
print(f"â€¢ ì ìˆ˜ ì •ë³´ ë°ì´í„°: {len(scores_df)}ëª…") 
print(f"â€¢ ê³µí†µ í•™ìƒ ID: {set(students_df['student_id']) & set(scores_df['student_id'])}")
print(f"â€¢ í•™ìƒ ì •ë³´ì—ë§Œ ìˆëŠ” ID: {set(students_df['student_id']) - set(scores_df['student_id'])}")
print(f"â€¢ ì ìˆ˜ ì •ë³´ì—ë§Œ ìˆëŠ” ID: {set(scores_df['student_id']) - set(students_df['student_id'])}")
```

```python
print("\n=== 1ï¸âƒ£ INNER JOIN (ë‚´ë¶€ ì¡°ì¸) ===")
print("ğŸ¯ ëª©í‘œ: ë‘ DataFrameì—ì„œ ê³µí†µìœ¼ë¡œ ì¡´ì¬í•˜ëŠ” student_idë§Œ í¬í•¨")

# Inner Join ìˆ˜í–‰
inner_result = pd.merge(students_df, scores_df, on='student_id', how='inner')

print("ğŸ“‹ Inner Join ê²°ê³¼:")
display(inner_result)

print(f"ğŸ” ë¶„ì„:")
print(f"â€¢ ì›ë³¸ í•™ìƒ ìˆ˜: {len(students_df)}ëª…")
print(f"â€¢ ì›ë³¸ ì ìˆ˜ ìˆ˜: {len(scores_df)}ê°œ")  
print(f"â€¢ Inner Join ê²°ê³¼: {len(inner_result)}í–‰")
print(f"â€¢ í¬í•¨ëœ í•™ìƒ ID: {inner_result['student_id'].tolist()}")
print(f"â€¢ ì œì™¸ëœ í•™ìƒ: ê¹€ì² ìˆ˜ ì™¸ {len(students_df) - len(inner_result)}ëª… (ì ìˆ˜ ì—†ìŒ)")
print(f"â€¢ ì œì™¸ëœ ì ìˆ˜: {len(scores_df) - len(inner_result)}ê°œ (í•™ìƒ ì •ë³´ ì—†ìŒ)")

print("\nğŸ’¡ Inner Join íŠ¹ì§•:")
print("âœ… ì™„ì „íˆ ë§¤ì¹­ë˜ëŠ” ë°ì´í„°ë§Œ í¬í•¨")
print("âœ… ë°ì´í„° ë¬´ê²°ì„± ë³´ì¥")
print("âŒ ì¼ë¶€ ë°ì´í„° ì†ì‹¤ ê°€ëŠ¥ì„±")
print("ğŸ¯ ì‚¬ìš© ì˜ˆì‹œ: ì„±ì  ë¶„ì„ ì‹œ í•™ìƒ ì •ë³´ì™€ ì ìˆ˜ê°€ ëª¨ë‘ ìˆëŠ” ê²½ìš°ë§Œ")
```

```python
print("\n=== 2ï¸âƒ£ LEFT JOIN (ì¢Œì¸¡ ì¡°ì¸) ===")
print("ğŸ¯ ëª©í‘œ: ì™¼ìª½ DataFrame(í•™ìƒ ì •ë³´)ì˜ ëª¨ë“  í–‰ ìœ ì§€ + ë§¤ì¹­ë˜ëŠ” ì ìˆ˜ ì •ë³´ ì¶”ê°€")

# Left Join ìˆ˜í–‰
left_result = pd.merge(students_df, scores_df, on='student_id', how='left')

print("ğŸ“‹ Left Join ê²°ê³¼:")
display(left_result)

print(f"ğŸ” ë¶„ì„:")
print(f"â€¢ ì›ë³¸ í•™ìƒ ìˆ˜: {len(students_df)}ëª…")
print(f"â€¢ Left Join ê²°ê³¼: {len(left_result)}í–‰ (í•™ìƒ ìˆ˜ì™€ ë™ì¼)")
print(f"â€¢ ì ìˆ˜ê°€ ìˆëŠ” í•™ìƒ: {left_result['math_score'].notna().sum()}ëª…")
print(f"â€¢ ì ìˆ˜ê°€ ì—†ëŠ” í•™ìƒ: {left_result['math_score'].isna().sum()}ëª…")

# ì ìˆ˜ê°€ ì—†ëŠ” í•™ìƒ í™•ì¸
missing_scores = left_result[left_result['math_score'].isna()]
if not missing_scores.empty:
    print(f"â€¢ ì ìˆ˜ ì—†ëŠ” í•™ìƒ: {', '.join(missing_scores['name'].tolist())}")

print("\nğŸ’¡ Left Join íŠ¹ì§•:")
print("âœ… ê¸°ì¤€ í…Œì´ë¸”(ì™¼ìª½)ì˜ ëª¨ë“  ë°ì´í„° ë³´ì¡´")
print("âœ… ë§¤ì¹­ë˜ì§€ ì•ŠëŠ” ê²½ìš° NaNìœ¼ë¡œ í‘œì‹œ")
print("âœ… ë°ì´í„° ì†ì‹¤ ì—†ìŒ (ê¸°ì¤€ í…Œì´ë¸” ê¸°ì¤€)")
print("ğŸ¯ ì‚¬ìš© ì˜ˆì‹œ: ëª¨ë“  í•™ìƒ ëª…ë‹¨ ìœ ì§€í•˜ë©´ì„œ ì ìˆ˜ ì •ë³´ ì¶”ê°€")
print("ğŸ“Š ì‹¤ë¬´ í™œìš©: ê³ ê° ì •ë³´ + êµ¬ë§¤ ì´ë ¥, ì§ì› ì •ë³´ + í‰ê°€ ì ìˆ˜")
```

```python
print("\n=== 3ï¸âƒ£ RIGHT JOIN (ìš°ì¸¡ ì¡°ì¸) ===")
print("ğŸ¯ ëª©í‘œ: ì˜¤ë¥¸ìª½ DataFrame(ì ìˆ˜ ì •ë³´)ì˜ ëª¨ë“  í–‰ ìœ ì§€ + ë§¤ì¹­ë˜ëŠ” í•™ìƒ ì •ë³´ ì¶”ê°€")

# Right Join ìˆ˜í–‰
right_result = pd.merge(students_df, scores_df, on='student_id', how='right')

print("ğŸ“‹ Right Join ê²°ê³¼:")
display(right_result)

print(f"ğŸ” ë¶„ì„:")
print(f"â€¢ ì›ë³¸ ì ìˆ˜ ê°œìˆ˜: {len(scores_df)}ê°œ")
print(f"â€¢ Right Join ê²°ê³¼: {len(right_result)}í–‰ (ì ìˆ˜ ê°œìˆ˜ì™€ ë™ì¼)")
print(f"â€¢ í•™ìƒ ì •ë³´ê°€ ìˆëŠ” ì ìˆ˜: {right_result['name'].notna().sum()}ê°œ")
print(f"â€¢ í•™ìƒ ì •ë³´ê°€ ì—†ëŠ” ì ìˆ˜: {right_result['name'].isna().sum()}ê°œ")

# í•™ìƒ ì •ë³´ê°€ ì—†ëŠ” ì ìˆ˜ í™•ì¸
missing_students = right_result[right_result['name'].isna()]
if not missing_students.empty:
    print(f"â€¢ í•™ìƒ ì •ë³´ ì—†ëŠ” ID: {missing_students['student_id'].tolist()}")

print("\nğŸ’¡ Right Join íŠ¹ì§•:")
print("âœ… ê¸°ì¤€ í…Œì´ë¸”(ì˜¤ë¥¸ìª½)ì˜ ëª¨ë“  ë°ì´í„° ë³´ì¡´") 
print("âœ… ë§¤ì¹­ë˜ì§€ ì•ŠëŠ” ê²½ìš° NaNìœ¼ë¡œ í‘œì‹œ")
print("âœ… Left Joinì˜ ë°˜ëŒ€ ê°œë…")
print("ğŸ¯ ì‚¬ìš© ì˜ˆì‹œ: ëª¨ë“  ì ìˆ˜ ë°ì´í„° ìœ ì§€í•˜ë©´ì„œ í•™ìƒ ì •ë³´ ì¶”ê°€")
print("ğŸ“ ì°¸ê³ : ì‹¤ë¬´ì—ì„œëŠ” Left Joinì„ ë” ë§ì´ ì‚¬ìš© (ì§ê´€ì )")
```

```python
print("\n=== 4ï¸âƒ£ OUTER JOIN (ì™¸ë¶€ ì¡°ì¸) ===")
print("ğŸ¯ ëª©í‘œ: ì–‘ìª½ DataFrameì˜ ëª¨ë“  í–‰ í¬í•¨ (í•©ì§‘í•©)")

# Outer Join ìˆ˜í–‰
outer_result = pd.merge(students_df, scores_df, on='student_id', how='outer')

print("ğŸ“‹ Outer Join ê²°ê³¼:")
display(outer_result)

print(f"ğŸ” ë¶„ì„:")
print(f"â€¢ ì›ë³¸ í•™ìƒ ìˆ˜: {len(students_df)}ëª…")
print(f"â€¢ ì›ë³¸ ì ìˆ˜ ìˆ˜: {len(scores_df)}ê°œ")
print(f"â€¢ Outer Join ê²°ê³¼: {len(outer_result)}í–‰")
print(f"â€¢ ì™„ì „í•œ ë°ì´í„° (í•™ìƒì •ë³´ + ì ìˆ˜): {outer_result[['name', 'math_score']].notna().all(axis=1).sum()}í–‰")
print(f"â€¢ í•™ìƒ ì •ë³´ë§Œ ìˆëŠ” ê²½ìš°: {outer_result['name'].notna().sum() - outer_result[['name', 'math_score']].notna().all(axis=1).sum()}í–‰")
print(f"â€¢ ì ìˆ˜ ì •ë³´ë§Œ ìˆëŠ” ê²½ìš°: {outer_result['math_score'].notna().sum() - outer_result[['name', 'math_score']].notna().all(axis=1).sum()}í–‰")

print("\nğŸ’¡ Outer Join íŠ¹ì§•:")
print("âœ… ëª¨ë“  ë°ì´í„° ë³´ì¡´ (í•©ì§‘í•©)")
print("âœ… ë§¤ì¹­ë˜ì§€ ì•ŠëŠ” ê²½ìš° NaNìœ¼ë¡œ í‘œì‹œ")
print("âœ… ë°ì´í„° ì†ì‹¤ ì „í˜€ ì—†ìŒ")
print("âŒ ê²°ê³¼ í¬ê¸°ê°€ ê°€ì¥ í¼")
print("ğŸ¯ ì‚¬ìš© ì˜ˆì‹œ: ì „ì²´ ë°ì´í„° í˜„í™© íŒŒì•…, ëˆ„ë½ëœ ë°ì´í„° í™•ì¸")

print("\n=== ğŸ“Š 4ê°€ì§€ Join ê²°ê³¼ ë¹„êµ ===")
print(f"{'Join íƒ€ì…':<12} {'ê²°ê³¼ í–‰ìˆ˜':<8} {'íŠ¹ì§•'}")
print("-" * 50)
print(f"{'Inner':<12} {len(inner_result):<8} êµì§‘í•© (ì™„ì „ ë§¤ì¹­ë§Œ)")
print(f"{'Left':<12} {len(left_result):<8} ì™¼ìª½ ê¸°ì¤€ (í•™ìƒ ì •ë³´ ë³´ì¡´)")
print(f"{'Right':<12} {len(right_result):<8} ì˜¤ë¥¸ìª½ ê¸°ì¤€ (ì ìˆ˜ ì •ë³´ ë³´ì¡´)")
print(f"{'Outer':<12} {len(outer_result):<8} í•©ì§‘í•© (ëª¨ë“  ë°ì´í„° ë³´ì¡´)")
```
