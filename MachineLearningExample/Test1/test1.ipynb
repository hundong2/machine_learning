# 데이터 불균형이란?
데이터 불균형(Data Imbalance)이란 분류 문제에서 각 클래스(범주)의 데이터 샘플 수가 크게 차이 나는 현상을 의미합니다. 예를 들어, 당뇨병 예측 문제에서 '당뇨병 있음(1)'과 '당뇨병 없음(0)' 데이터의 비율이 1:9처럼 한쪽이 훨씬 많을 때 데이터 불균형이 발생합니다.

이런 상황에서는 모델이 소수 클래스(적은 데이터)보다 다수 클래스(많은 데이터)에 치우쳐 학습하게 되어, 소수 클래스의 예측 성능이 떨어질 수 있습니다. 즉, 전체 정확도는 높아 보여도 중요한 소수 클래스(예: 질병 환자)를 잘 맞추지 못할 수 있습니다.

## 답변
데이터 불균형은 분류 모델의 성능 평가와 실제 적용에서 중요한 문제입니다.  
불균형 데이터셋에서는 단순 정확도(accuracy)보다 정밀도(precision), 재현율(recall), F1-score와 같은 지표를 함께 확인해야 하며,  
필요하다면 오버샘플링(SMOTE 등), 언더샘플링, 가중치 조정 등으로 불균형을 완화할 수 있습니다.

# 박스 플롯 설명

- 박스(상자) 부분 : 데이터의 중간 50% 범위(IQR, 사분위수 범위)를 나타냄
- 박스 안의 선(-): 데이터의 중앙값 ( 50%의 지점, Median)
- 위쪽, 아래쪽의 선(수염,whisker) : 데이터의 최소/최대 범위( 이상치를 제외한 값 ) 
- 박스플롯 바깥쪽의 점 : 이상치(Outlier)

```python
def add_one(x):
    return x + 1

# apply 함수를 사용하여 add_one 함수를 각 요소에 적용할 수 있습니다.
df["age_plus_one"] = df["age"].apply(add_one) # "age" 열에 add_one 함수를 적용하여 새로운 열 "age_plus_one"을 생성합니다.
df
```

# 조건에 따라 범주화하기

```python
def categorize_age(age):
    if age < 30:
        return "young"
    elif age < 40:
        return "middle-aged"
    else:
        return "old"
```

```python
df["partition"] = df["age"].apply(categorize_age) # "age" 열에 categorize_age 함수를 적용하여 새로운 열 "partition"을 생성합니다.
```

# Using StandardScalar translation data and normalization check

- 머신러닝 모델을 학습 할 때, 각 변수의 스케일(값의 범위)이 다르면 모델이 특정 변수에 영향을 더 많이 받음. 
- 스케일이 서로 다른 데이터를 사용하면 모델이 특정 컬럼에 과도하게 영향을 받을 가능성이 커짐. 
- 모든 수치형 변수의 범위를 일정하게 맞추는 `정규화(Normalization)`가 필요.  
- `StandardScaler`를 활용하여 데이터를 `표준 정규화(Standardization)`
  - 모든 수치형 데이터를 같은 범위로 맞출 때 사용. 


```python
# StandardScaler 라이브러리 불러오기
from sklearn.preprocessing import StandardScaler #sklearn.preprocessing 모듈에서 StandardScaler 클래스를 불러옵니다.
# 각 칼럼의 평균을 0, 표준편차를 1로 변환하여 데이터의 스케일을 맞추는 역할. 

# 정규화를 적용할 수치형 변수 선택 (Outcome 제외)
numeric_columns = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", 
                   "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"]

# StandardScaler 객체 생성
scaler = StandardScaler()

# StandardScaler를 사용해 평균과 표준편차 계산
scaler.fit(train[numeric_columns])
```

```python
print("Mean:", scaler.mean_)  # 각 칼럼의 평균
print("Scale:", scaler.scale_)  # 각 칼럼의 표준편차
```

```python
# StandardScaler를 사용하여 데이터 변환
train_scaled = scaler.transform(train[numeric_columns]) # train 데이터의 수치형 변수들을 
train[numeric_columns] = train_scaled
train.head(5)
```

```python
import numpy as np

# 평균과 표준편차 계산 (변수를 활용하여 저장)
train_scaled_mean = train[numeric_columns].mean()
train_scaled_std = train[numeric_columns].std()

# 평균과 표준편차 출력 (소수점 4자리까지 반올림)
print("각 열의 정규화된 평균:", np.round(train_scaled_mean, 4))
print("각 열의 정규화된 표준편차:", np.round(train_scaled_std, 4))
```

# model validation

1. 검증 데이터가 있을 때 모델 평가
2. 모델의 성늘을 평가하는 가장 기본적인 방법은 정확도(Accuracy)
3. `accuracy_score()` function 
4. Validation Set (검증데이터)가 있는 경우 모델을 학습 한 후, 검증 데이터를 활용하여 평가
5. 검증 데이터를 따로 분리 하지 않은 경우 -> 테스트 데이터(Test Set)에 대한 예측을 제출한 후에 점수 확인가능 



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #이진 분류(Binary Classification) 모델인 로지스틱 회귀를 사용합니다.

train = pd.read_csv("train.csv")
train
```

```python
X = train.drop(columns=["Outcome", "ID"])  # 특성 데이터
y = train["Outcome"]  # 정답 데이터

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=False)

# sklearn.model_selection 모듈에서 train_test_split 함수를 사용하여 데이터를 학습용
#과 검증용으로 분리합니다.
# test_size=0.2는 전체 데이터의 20%를 검증용으로 사용하고, 나머지 80%를 학습용으로 사용합니다.
# shuffle=False는 데이터를 섞지 않고 순서대로 분리합니다.
print("X_train_type:", type(X_train))  # X_train의 타입 출력
print("X_valid_type:", type(X_valid))  # X_valid의 타입 출력
print("y_train_type:", type(y_train))  # y_train의 타입 출력
print("y_valid_type:", type(y_valid))  # y_valid의 타입 출력
model = LogisticRegression()
model.fit(X_train, y_train)  # 모델 학습
y_pred_valid = model.predict(X_valid)  # 검증 데이터에 대한 예측 결과를 계산합니다.
#predict() 메서드는 주어진 입력 데이터에 대한 예측 결과를 반환합니다.
#predict()는 각 샘플을 0 또는 1로 분류하는 이진 분류 문제에 사용됩니다.
print("Validation Predictions:", y_pred_valid[:5])  # 예측 결과의 상위
print('predict value: ', y_pred_valid, type(y_pred_valid))  # 예측값 출력
print('reaul' ,y_valid.values, type(y_valid))  # 실제 정답 값 출력
```
