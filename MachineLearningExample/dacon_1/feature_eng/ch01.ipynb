```python
import pandas as pd 

bicycle = pd.read_csv('train.csv')
display(bicycle.head())
display(bicycle.info())
display(f"{bicycle['date_time'].iloc[0]} ~ {bicycle['date_time'].iloc[-1]}")
```

```python
def seperate_datetime(dataframe):
    year = []
    month = []
    day = []

    for date in dataframe.date_time:
        year_point, month_point, day_point = date.split('-') # Split string by '-' and create a list e.g., '2016-04-01' -> ['2016', '04', '01']
        year.append(int(year_point)-2017)
        month.append(int(month_point)-3)
        day.append(int(day_point))
    return year, month, day

year, month, day = seperate_datetime(bicycle)

bicycle['year'] = year
bicycle['month'] = month
bicycle['day'] = day

bicycle.head()
```

```python
X  = bicycle.drop(['date_time', 'number_of_rentals'], axis=1)
y = bicycle.number_of_rentals
X
```

```python
display(bicycle['number_of_rentals'])
display(bicycle.loc[:, 'number_of_rentals'])
display(bicycle.number_of_rentals)
```

```python
from sklearn.linear_model import LinearRegression

import numpy as np

# NMAE (Normalized Mean Absolute Error)
# A metric to evaluate how large the error is relative to the original value
def NMAE(real, pred):
    score = np.mean(np.abs(real-pred)/ real)
    return score

model = LinearRegression()
model.fit(X, y)
y_hat = model.predict(X)
score = NMAE(y, y_hat)
print(f'model NMAE : {score:.4f}')
```

# Feature Engineering

## Purpose of Creating Features

- To help the model learn more patterns from the data
- To increase the expressiveness of the data
- To extract new information not present in the original data
- To derive potential information from the data

```python
display(bicycle['date_time'])
```

```python
from sklearn.preprocessing import LabelEncoder

X_human = X.copy()

week_day = pd.to_datetime(bicycle['date_time']).dt.day_name()
display(f'week_day : {week_day}')

# Convert string data to numerical data
le = LabelEncoder()
le.fit(week_day)
X_human['week_day'] = le.transform(week_day)
X_human.head()
```

```python
# 1. Information for "weather with a large daily temperature range"
X_human['temp_diff_info'] = X_human['high_temp'] - X_human['low_temp']

# 2. Information for "hot and humid weather"
X_human['sweat_info'] = X_human['high_temp'] * X_human['humidity']

# 3. Information for "cold and windy weather"
X_human['cold_info'] = X_human['low_temp'] * X_human['wind_speed']

X_human[['temp_diff_info', 'sweat_info', 'cold_info']].head()
```

```python
## Check Model 1 Performance

model = LinearRegression() # Define model
model.fit(X_human, y) # Train

y_hat = model.predict(X_human) # Predict y

score = NMAE(y,y_hat)

print(f'Model NMAE: {score}')
```

```python
# feature engineering by computer 

# Model 2 uses features created by the computer.
X_computer = X.copy()

col_list = X_computer.columns

# Use a nested for loop to add new features: the square of a feature and the product of two features.
for i in range(len(col_list)):
    for j in range(i, len(col_list)):
        X_computer[f'{col_list[i]}*{col_list[j]}'] = X_computer[col_list[i]] * X_computer[col_list[j]]

X_computer.head()
```

```python
## Check Model 2 Performance

model = LinearRegression() # Define model
model.fit(X_computer, y) # Train

y_hat = model.predict(X_computer) # Predict y

score = NMAE(y, y_hat)

print(f'Model NMAE: {score}')
```

## Meaning of Creating Features through Automatic Operations

- Essentially, it's one way to help machine learning models better understand non-linear relationships in the data.
- Using tools like `sklearn`'s `PolynomialFeatures` to create new features is an attempt to make the model better at capturing hidden patterns or relationships in the data.

### Addition

- Adding two features creates a linear combination of the features.
- e.g., Adding savings account balance + checking account balance to create a new feature for total account balance helps to better understand the overall asset size.

### Subtraction

- The difference between features.
- Subtracting manufacturing cost from the selling price calculates the net profit, informing the model how much profit is generated.

### Multiplication

- The product of two features often represents the interaction between them.
- You can get the total number of clicks by multiplying the number of ad impressions per hour by the click-through rate.

### Squaring

- Squaring a feature increases its importance.
- Squaring the distance to create the square of the distance makes points that are farther away more important than simple distance.

### Division

- Creates a ratio.
- You can calculate the average revenue per visitor by dividing the total sales revenue by the number of visitors. This can be important for evaluating the quality or value of visitors.
