# 딥러닝

- 인간이 문제를 해결하는 방식을 모방해서, 계속 데이터를 분석하고 그안에서 패턴을 찾아내려고 합니다. 
- 머신이 대화의 맥락을 이해하고, 사용자의 요청에 맞는 적절한 대답이나 조치를 취할 수 있도록합니다. 

## 머신러닝과 딥러닝의 차이

- 딥러닝은 머신러닝의 한분야로 더 깊고 복잡한 학습을 가능하게 합니다. 
- 머신러닝은 컴퓨터가 데이터를 분석하고 그 안에서 학습을 하여 결정을 내리는 과정, 이 과정에서 뭔가 잘못되면 사람이 개입해서 수정해야 한다. 
    - 새로운 유형의 스팸이 등장할 때마다 사람이 이를 지도해야 한다. 
- `딥러닝`의 경우 패턴을 스스로 학습하고, 더 복잡하고 미묘한 스팸의 형태까지도 자동으로 식별, 인간의 개입 없이도 스스로 학습과 판단 능력을 갖춘 인공신경망을 통해 이루어진다.  
    - 입력 -> 특징 추출 -> 분류 알고리즘 -> 출력
- `머신러닝`은 데이터로부터 배우는 컴퓨터의 능력 `딥러닝`은 이런 학습과정을 더욱 깊게 하여 스스로 판단하고 결정할 수 있는 능력을 가진 인공 신경망을 만드는 것 
    - 입력 -> 인공 신경망(특징 추출 & 분류 ) -> 출력 



## 머신러닝의 다양한 유형

1. 지도 학습

- 입력 데이터와 그에 해당하는 정답(라벨)을 함께 제공받음
- 데이터 과학자는 모델이 제시한 예측이 정확한지 평가하고, 모델을 조정하여 오류를 최소화 한다. 
- 지도 학습의 종류 
    - 분류(Classification) : 데이터를 사전에 정의된 여러 범주중 하나로 할당하는 과정 
    - 회귀(Regression): 연속적인 값을 예측하는 과정

2. 비지도 학습

- 모델에 정답 라벨이 붙어 있지 않은 데이터만 제공
- 컴퓨터가 스스로 데이터 안에서 패턴이나 구조를 발견
- 데이터에서 어떤 결과를 기대해야 할 지 명확하지 않을 때 유용 
- 비지도 학습의 종류
    - 클러스터링 ( Clustering ) : 데이터를 자연스럽게 유사한 그룹으로 나누는 과정
    - 차원 축소(Dimensionality Reduction) : 데이터의 특성을 줄이는 것을 목표, 데이터의 가장 중요한 특성을 유지하면서 노이즈를 제거하고 계산을 단순화하기 위해 사용 ( 예) PCA )

3. 강화 학습

- 자신의 행동에 대한 피드백을 받으며, 올바른 결정에는 보상을, 잘못된 결정에는 처벌을 받는다.  


# 딥러닝 알고리즘


1. 합성곱 신경망 (CNN)

- 이미지 처리와 객체 인식에 탁월한 딥러닝 알고리즘입니다.
- 사람이 이미지에서 모서리, 질감, 형태를 단계적으로 파악하듯이,
  CNN도 낮은 수준 특징(선, 경계)부터 높은 수준 특징(얼굴, 자동차)까지 자동으로 학습합니다.
- 핵심 구성:
  - 합성곱(Convolution) 층: 중요한 특징을 추출
  - 풀링(Pooling) 층: 정보는 유지하면서 크기를 줄여 계산량 감소
  - 완전연결(Fully Connected) 층: 최종 분류 수행

예시:
- 고양이/강아지 사진 분류
- 의료 영상(X-ray)에서 이상 부위 탐지
- 자율주행에서 보행자/차선 인식

간단한 동작 흐름:
- 입력 이미지 -> 특징 추출(합성곱) -> 크기 축소(풀링) -> 분류 결과 출력

2. 순환 신경망 (RNN)

- 시간에 따른 데이터의 순서와 연관성을 학습하는 데 강점을 가진 알고리즘입니다.
- 일반 신경망과 달리 이전 시점의 정보를 다음 계산에 반영하여,
  "순서"가 중요한 데이터를 다루는 데 유리합니다.
- 핵심 아이디어:
  - 현재 입력 + 이전 상태(hidden state)를 함께 사용
  - 과거 문맥을 이어받아 다음 값을 예측

예시:
- 문장 다음 단어 예측 (자연어 처리)
- 주가/전력 사용량 같은 시계열 예측
- 음성 인식, 번역, 감성 분석

간단한 동작 흐름:
- t시점 입력 + t-1 시점 기억 -> t시점 출력 생성 -> 다음 시점으로 기억 전달

참고:
- 긴 시퀀스에서는 기본 RNN이 장기 의존성 학습이 어렵기 때문에,
  실무에서는 LSTM, GRU를 자주 사용합니다.

## 딥러닝

1. 퍼셉트론

- 퍼셉트론은 인공 신경망 모델 중 가장 기본적인 형태 
- 1957년 로젠블라트(Frank Rosenblatt)에 의해 개발 되었다. 

## 가중치와 편향

1. 가중치(Weight)

- 입력 신호가 출력에 미치는 영향력을 조절 
- 중요성이나 영향을 나타냄 
- 훈련을 통해 최적화 된다.  

2. 편향(Bias)

- 퍼셉트론에서는 결정을 내리는 기준점으로 임계값(세타 $\theta$ ) 를 사용 
- 이 값을 넘어설 때 만 퍼셉트론이 1을 출력 
- 임계값($\theta$)를 $-b$ 로 치환하여 편향 b를 사용합니다. 
    - 편향은 결정 임계값을 조정하는 역할 
- 편향은 가중치들과 함께 조절되어야하는 중요한 파라미터 
- $w1x1 + w2x2 + b(편향) = y$

- 가중치는 입력 신호의 중요도를 조정하는 반면, 편향은 모델이 어느 정도 확실을 갖고 결정을 내릴 수 있도록 해준다.

### 활성화 함수(activation function)

#### 인공 신경망의 기본 구조 

- 인공 신경망의 기본 구조 
- 신경망 안의 노드가 받은 신호를 가지고, 다음 단계로 어떤 신호를 보낼지 결정하게 해준다.  
- 노드가 얼마나 `활성화` 될지를 결정하는 것. 
- 만약 입력 신호가 중요하다면, 활성화 함수는 노드를 활성화시켜 다음 레벨로 강한 신호를 보내게 할 거고, 그렇지 않다면 신호를 약하게 만들거나 전혀 보내지 않을 것. 
- 이런 활성화 함수는 신경망이 단순한 직선적인 관계만이 아닌, 곡선이나 더 복잡한 형태의 관계를 배울 수 있게 도와준다. 

$$
w1x1 + w2x2 + b = a(가중합)
$$

$$
h(a) = y
$$

# 계단 함수(step function)의 이해 

- 기본적인 활성화 함수 중 하나인 계단 함수는 이름처럼 계단 모양을 하고 있다. 
- 입력값이 특정 임계값보다 크거나 작을 때 두가지의 고유한 값(0 or 1) 중 하나를 출력하는 함수 

## 장단점 

- 구현이 매우 간단, 이해하기 쉽다. 
- 그라디언트(미분값) 0이 되기 때문에, 역전파에서 가중치를 업데이트할 때 사용할 수 없다. 이로인해 심층(깊은 신경망) 학습에는 적합하지 않음.  
- 특정 값에 따라 0 또는 1의 출력값을 가지며, 비선형 함수 

$$f(x) = \begin{cases} 0 & (x \le \theta) \\ 1 & (x > \theta) \end{cases}$$

## 활성화 함수의 유형 및 종류 

- 시그모이드(sigmoid) : 출력이 0과 1사이 이진분류 문제에서 출력층에 주로 사용. 그라디언트 소실 문제가 있을 수 있다. 
- 탄젠트(Tanh): 출력이 -1과 1 사이. 중심이 0에 가까워 학습 초기에 유리할 수 있지만, 그라디언트 소실 문제가 있다. 
- 렐루(ReLU): 음수 입력에 대해서는 0, 양수 입력에 대해서는 입력을 그대로 출력, 학습 속도가 빠르며, 일반적으로 은닉층에 사용 
- Softmax(소프트맥스): 다중 클래스 분류 문제의 출력층에 사용, 출력값을 확률 분포로 반환 

```python
import numpy as np
import matplotlib.pyplot as plt

# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 시그모이드 미분
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# x 범위
x = np.linspace(-10, 10, 500)

# y 계산
y_sigmoid = sigmoid(x)
y_derivative = sigmoid_derivative(x)

# 그래프
plt.figure(figsize=(10, 5))
plt.plot(x, y_sigmoid, label='Sigmoid: σ(x)', linewidth=2)
plt.plot(x, y_derivative, label="Derivative: σ'(x)", linewidth=2, linestyle='--')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

$$
\sigma'(x) = \sigma(x)\left(1-\sigma(x)\right)
$$


## 시그모이드 미분 공식 유도

시그모이드 함수를 다음과 같이 둡니다.

$$
\sigma(x) = \frac{1}{1+e^{-x}} = (1+e^{-x})^{-1}
$$

양변을 $x$에 대해 미분하면,

$$
\sigma'(x) = \frac{d}{dx}(1+e^{-x})^{-1}
$$

연쇄법칙을 적용하면,

$$
\sigma'(x) = -(1+e^{-x})^{-2} \cdot \frac{d}{dx}(1+e^{-x})
$$

그리고

$$
\frac{d}{dx}(1+e^{-x}) = -e^{-x}
$$

이므로,

$$
\sigma'(x) = -(1+e^{-x})^{-2}(-e^{-x})
= \frac{e^{-x}}{(1+e^{-x})^2}
$$

이 식을 시그모이드 형태로 다시 쓰면,

$$
\sigma(x)=\frac{1}{1+e^{-x}}, \quad
1-\sigma(x)=1-\frac{1}{1+e^{-x}}=\frac{e^{-x}}{1+e^{-x}}
$$

따라서,

$$
\sigma(x)(1-\sigma(x))
= \frac{1}{1+e^{-x}} \cdot \frac{e^{-x}}{1+e^{-x}}
= \frac{e^{-x}}{(1+e^{-x})^2}
= \sigma'(x)
$$

최종 결과:

$$
\boxed{\sigma'(x) = \sigma(x)(1-\sigma(x))}
$$
