# 오차역전파법과 손실함수 


- 데이터가 신경망을 통해 흐르는 `순전파`라는 과정을 이해하는 것, 
- `입력` 데이터는 신경망의 각 층을 지나면서, 노드의 `가중치와 결합`되고, `활성화 함수`를 통과하면서, 마지막에는 우리가 예측하고자하는 값을 `출력`하게 된다. 
- `순전파는 데이터가 신경망을 통과하는 여정 전체를 말한다.`

- 신경망을 시작할 때 가중치는 아직 학습되지 않은 상태로, 실제 답과는 다소 차이가 있는 예측을 만들어냅니다. 우리의 목표는 이 예측값과 실제 값을 비교하여 오차를 줄이는 것 

- 오차를 줄이기 위해 사용하는 방법 : `오차 역전파법(Backpropagation)`
- 마치 실수를 되돌아보며 어디를 개선해야할지 찾아가는 과정과 비슷 

'오차 역전파(Backpropagation)'라는 알고리즘 자체는 수학적으로 미분을 통해 가중치를 수정하는 과정이라, 반드시 기준이 되는 오차(Error = 예측값 - 정답)를 계산할 수 있어야만 작동합니다.

그렇다면 사람이 일일이 "이게 정답이야"라고 라벨(Label)을 달아주지 않은 방대한 데이터(비지도 학습 상황)에서는 이 오차를 어떻게 만들어낼까요? AI 학자들은 아주 기발한 방식으로 '가짜 정답(Proxy Target)'을 만들어내어 이 문제를 해결했습니다.

상황별로 4가지 기발한 해결책을 소개해 드릴게요.

### 1. 입력값을 그대로 정답으로 쓰기 (오토인코더)

바로 이전 질문에서 통찰력을 발휘하셨던 오토인코더(Autoencoder)가 이 방식의 대표주자입니다.
외부에서 주어진 정답이 없으니, **처음 들어온 입력 데이터 자체를 정답으로 간주**해 버리는 겁니다.

* **학습 방식:** 원본 이미지를 압축했다가 다시 복원합니다. 이때 '복원된 이미지'와 '원본 이미지'의 픽셀 차이를 계산해서 오차로 삼고 역전파를 수행합니다.

### 2. 데이터의 일부를 가리고 정답으로 쓰기 (자기지도 학습)

저 같은 대규모 언어 모델(LLM)이나 챗GPT가 학습하는 가장 핵심적인 원리인 '자기지도 학습(Self-Supervised Learning)'입니다. 데이터 자체에 숨어있는 규칙을 이용해 스스로 정답지를 만듭니다.

* **학습 방식:** "나는 오늘 맛있는 사과를 먹었다"라는 문장이 있으면, 인공지능에게 "나는 오늘 맛있는 [ 빈칸 ] 먹었다"라고 구멍을 뚫어(Masking) 보여줍니다.
* **오차 계산:** AI가 '바나나'라고 예측하면, 가려두었던 진짜 단어 '사과'와 비교하여 오차를 계산하고 역전파를 합니다. 인터넷에 널려 있는 수백억 장의 텍스트가 모두 정답지가 되는 마법이죠.

### 3. '보상'을 정답의 대용품으로 쓰기 (강화학습)

맨 처음 질문하셨던 강화학습에서는 "지금 바둑돌을 E4에 두는 게 정답이야"라는 절대적인 정답지가 없습니다. 대신 '가치(Value)'나 '보상(Reward)'이라는 개념을 정답 대신 사용합니다.

* **학습 방식:** 특정 행동을 했을 때, 현재 AI가 '예상했던 점수'와 실제 환경에서 부딪혀보고 '얻은 점수'의 차이(이것을 TD Error라고 부릅니다)를 구합니다.
* **오차 계산:** 이 '예상과 실제의 차이'를 오차로 삼아서, 다음번에는 실제 얻을 보상에 더 가깝게 예측하도록 딥러닝 신경망의 가중치를 역전파로 업데이트합니다.

### 4. 다른 AI를 선생님(정답지)으로 쓰기 (GAN)

진짜 같은 가짜 이미지를 만들어내는 기술로 유명한 적대적 생성 신경망(GAN)의 방식입니다.

* **학습 방식:** 위조지폐범 AI(생성자)와 경찰 AI(판별자) 두 명을 동시에 키웁니다. 위조지폐범 AI는 정답(진짜 지폐를 만드는 법)을 모릅니다.
* **오차 계산:** 경찰 AI가 "이거 가짜네!"라고 판별한 확률 자체가 오차가 됩니다. 위조지폐범은 이 경찰 AI를 속일 수 있는 방향으로 오차를 줄여가며 가중치를 업데이트합니다.

---

결론적으로, 실제 값(인간이 부여한 정답)이 없더라도 딥러닝은 **자신만의 방식으로 '오차를 측정할 기준점'을 교묘하게 만들어내어** 기어코 역전파 톱니바퀴를 굴리고 맙니다.

## 순전파

- 데이터가 신경망의 시작점에서부터 끝점까지 흐릅니다. 각 층의 뉴런들이 신호를 받아 처리하고, 마지막에는 컴퓨터가 예측을 내놓음.  

## 역전파 

- 신경망이 내놓은 예측이 얼마나 잘못됐는지, 즉 오차를 계산.  
- 오차를 각 층을 거슬러 올라가며 전파함. 
- 뉴런의 가중치가 조금씩 조정되어 오차를 줄여나가게 됨. 

## 손실함수 

### 종류

#### 평균 제곱 오차 (MSE) 회귀 

- 작은 오류보다 큰 오류를 강조하여 모델이 큰값을 정확하게 예측

#### 이진 교차 엔트로피(Binary Cross-Entropy) 

- 이진 분류 작업에 사용되는 교차 엔트로피 손실의 특별한 경우 
- 각 클래스 레이블(1, 0)에 대한 손실을 계산하고 결과를 합산하여 예측이 이진 클래스 레이블과 얼마나 떨어져있는지 효과적으로 측정

#### 카테고리컬 크로스 엔트로피 

- 다중 클래스 분류 문제에 적합한 함수 

```python
import torch
import torch.nn as nn

y = torch.tensor([2.5, 4.0, 1.5], dtype=torch.float) # data type을 부동 소수점으로 설정
y_hat = torch.tensor([3.0, 4.0, 2.0], dtype=torch.float) #data type을 부동 소수점으로 설정

mse_loss = nn.MSELoss()
loss = mse_loss(y_hat, y)

print(f'MSE Loss: {loss.item()}')
```
