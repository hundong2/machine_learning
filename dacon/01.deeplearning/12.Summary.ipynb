# 순방향 전파

순방향 전파(Forward Propagation)는 입력 데이터를 신경망의 앞층에서 뒤층으로 전달해 최종 예측값을 계산하는 과정입니다.

## 핵심 개념

- 입력 $x$가 각 층의 가중치 $W$, 편향 $b$와 연산됩니다.
- 각 층에서 선형변환 후 활성화 함수를 적용합니다.
- 마지막 층에서 예측값 $\hat{y}$를 만듭니다.

## 기본 수식

한 층에서:

- $z = Wx + b$
- $a = f(z)$

여러 층을 통과하면 최종 출력:

- $\hat{y} = f_L(\cdots f_2(f_1(xW_1+b_1)W_2+b_2)\cdots)$

## 왜 중요한가?

- 모델이 현재 파라미터로 어떤 예측을 하는지 계산하는 단계입니다.
- 이 결과로 손실(loss)을 계산하고, 이후 역전파(backpropagation)에서 기울기를 구해 가중치를 업데이트합니다.

## 간단한 절차

1. 입력 텐서를 모델에 넣는다.
2. 각 층에서 선형연산 + 활성화 함수를 적용한다.
3. 최종 출력(예측값)을 얻는다.
4. 예측값과 정답을 비교해 손실을 계산한다.

````python
input_data = [30, 2] # 집의 크기가 30평, 방의 개수가 2개

weights = {
    'node_0': [50, 10], # 은닉층 첫 번째 노드의 가중치
    'node_1': [-1, 0],  # 은닉층 두 번째 노드의 가중치
    'output': [0.5, 25] # 출력층 가중치
}

# 은닉층 첫 번째 노드 계산
node_0_value = input_data[0] * weights['node_0'][0] + input_data[1] * weights['node_0'][1]

# 은닉층 두 번째 노드 계산
node_1_value = input_data[0] * weights['node_1'][0] + input_data[1] * weights['node_1'][1]

# 은닉층 결과
hidden_layer_values = [node_0_value, node_1_value]
print(f"은닉층 결과 출력:",hidden_layer_values)

# 출력층에서의 최종 예측
output = hidden_layer_values[0] * weights['output'][0] + hidden_layer_values[1] * weights['output'][1]

# 결과 출력
print(f"예측된 집의 가격: {int(output)}억")
````

````python
# 입력 데이터와 가중치 설정
input_data = [10, 2]  

weights = {
    'node_0': [0.5, 1],  
    'node_1': [-1, 0],   
    'output': [1, 0.5]   
}

# 은닉층 노드 계산
node_0_value = input_data[0] * weights['node_0'][0] + input_data[1] * weights['node_0'][1]
node_1_value = input_data[0] * weights['node_1'][0] + input_data[1] * weights['node_1'][1]

hidden_layer_values = [node_0_value, node_1_value]

# 출력층 최종 예측
output = hidden_layer_values[0] * weights['output'][0] + hidden_layer_values[1] * weights['output'][1]
print(f"예측값: {output}")
````

9.다층 신경망과 ReLU 활성화 함수를 사용한 집 가격 예측 실습
이번 실습에서는 다층 신경망과 ReLU 활성화 함수를 사용하여 집 가격을 예측하는 모델을 구현해 볼 것입니다.
본 실습의 목표는 주어진 집의 크기와 방의 개수를 기반으로 집 가격을 예측하는 것입니다.

우리는 이를 위해 입력 데이터로부터 시작하여 두 개의 은닉층을 거치고, 최종적으로 출력층에서 예측 결과를 도출하는 과정을 구현할 것입니다.
이 과정을 통해, 신경망이 어떻게 다양한 층과 활성화 함수를 사용하여 입력 데이터를 처리하고, 복잡한 문제 해결에 접근하는지 이해할 수 있습니다.

[문제 2]

hidden2_node1_output를 계산할 때, 첫 번째 은닉층의 출력값인 hidden1_node0_activated과 hidden1_node1_activated을 각각 해당 가중치와 곱한 후 더해야 합니다.
hidden2_node1_activated를 계산할 때, relu 함수를 사용하여 양수는 그대로 두고 음수는 0으로 변환해야 합니다.
출력층의 예측값을 계산할 때, 두 번째 은닉층의 출력값인 hidden2_node0_activated와 hidden2_node1_activated을 각각 해당 가중치와 곱한 후 더해야 합니다.

````python
def relu(x):
    """ReLU 활성화 함수"""
    return max(0, x)

# 입력 데이터 설정
input_data = [30.0, 2.0]

# 가중치 설정
weights = {
    'hidden1_node0': [50.0, 10.0],
    'hidden1_node1': [-1.0, 0.0],
    'hidden2_node0': [0.1, 10.0],
    'hidden2_node1': [-0.1, 0.0],
    'output': [0.1, 25.0]
}

# 첫 번째 은닉층 계산 및 ReLU 활성화 함수 적용
hidden1_node0_output = input_data[0] * weights['hidden1_node0'][0] + input_data[1] * weights['hidden1_node0'][1]
hidden1_node0_activated = relu(hidden1_node0_output)

hidden1_node1_output = input_data[0] * weights['hidden1_node1'][0] + input_data[1] * weights['hidden1_node1'][1]
hidden1_node1_activated = relu(hidden1_node1_output)

# 두 번째 은닉층 계산 및 ReLU 활성화 함수 적용
hidden2_node0_output = hidden1_node0_activated * weights['hidden2_node0'][0] + hidden1_node1_activated * weights['hidden2_node0'][1]
hidden2_node0_activated = relu(hidden2_node0_output)

hidden2_node1_output = hidden1_node0_activated * weights['hidden2_node1'][0] + hidden1_node1_activated * weights['hidden2_node1'][1]
hidden2_node1_activated = relu(hidden2_node1_output)

# 출력층 계산
output = hidden2_node0_activated * weights['output'][0] + hidden2_node1_activated * weights['output'][1]

# 결과 출력
print(f"예측된 집의 가격: {int(output)}억")
````

# 가중치 조정을 통한 신경망의 예측 오차 최소화 실습



````python
input_data = [30, 2] # 입력 데이터
target_actual = 3000  # 실제 목표값
weights_0 = {'node_0': [100, 10]} # 첫 번째 가중치 설정

# 첫 번째 가중치를 사용한 예측값
output_0 = input_data[0] * weights_0['node_0'][0] + input_data[1] * weights_0['node_0'][1]
error_0 = output_0 - target_actual # 오차

# 가중치 조정 및 두 번째 예측
weights_1 = {'node_0': [90, 150]} # 두 번째 가중치 설정 (가중치 조정)
output_1 = input_data[0] * weights_1['node_0'][0] + input_data[1] * weights_1['node_0'][1]
error_1 = output_1 - target_actual

print("첫 번째 예측 오차:", error_0)
print("조정된 가중치로의 예측 오차:", error_1)
````

````python
def relu_derivative(output):
    """ReLU 함수의 미분값
    ReLU(x) = max(0, x) 이므로,
    미분은 x > 0이면 1, x <= 0이면 0
    역전파 시 gradient가 이 게이트를 통과할 수 있는지 결정한다.
    음수였던 노드(이미 0으로 클리핑된 노드)는 gradient를 뒤로 전달하지 않는다.
    """
    return 1 if output > 0 else 0

# 1. 오차 계산하기
# 모델 예측값과 실제 정답의 차이를 구한다.
# 이 오차가 역전파의 출발점이 된다.
predicted_value = 15
target_value = 18
error = predicted_value - target_value  # error = -3

# 2. 예측값에 대한 오류의 기울기 계산
# MSE 손실 함수 L = (predicted - target)^2 을 사용할 때,
# dL/d(predicted) = 2 * (predicted - target)
# 즉, 손실이 예측값에 대해 얼마나 민감하게 변하는지를 나타낸다.
error_gradient = 2 * error  # = -6

# 두 번째 은닉층 노드의 ReLU 이후 출력값
# 체인룰에서 이 값이 해당 가중치의 기울기 계산에 직접 곱해진다.
hidden_node_output = 152  # ReLU를 거친 후의 값(양수이므로 gradient가 통과됨)

# 3. 가중치에 대한 손실 함수의 기울기 계산 (체인룰 적용)
# dL/dw = dL/d(pred) * d(pred)/d(hidden_a) * d(hidden_a)/d(z) * d(z)/dw
#       = error_gradient * w_output * relu'(z) * hidden_node_output
# 여기서는 단순화하여:
#   - hidden_node_gradient: ReLU 미분 → gradient가 통과 가능한지 여부
#   - weight_gradient: 가중치를 1 증가시킬 때 손실이 얼마나 변하는지
hidden_node_gradient = relu_derivative(hidden_node_output)  # 양수이므로 1
weight_gradient = hidden_node_output * error_gradient * hidden_node_gradient
# = 152 * (-6) * 1 = -912

# 4. 다음 가중치 계산하기 (경사하강법 적용)
# w_new = w_old - lr * dL/dw
# gradient가 음수(-912)이므로 가중치는 증가하는 방향으로 업데이트된다.
# → 예측값을 키워서 정답(18)에 가까워지려는 방향
learning_rate = 0.01
next_weight = 0.1 - learning_rate * weight_gradient
# = 0.1 - 0.01 * (-912) = 0.1 + 9.12 = 9.22

print(f"업데이트된 가중치 값: {next_weight}")
````

# 순전파 & 역전파 완전 이해

## 1. 신경망 구조 (이 예제)

```
입력층          은닉층1         은닉층2         출력층
                                             
 x₁=30  ──w[50]──→ h1n0 ──w[0.1]──→ h2n0 ──w[0.1]──┐
         ──w[10]──↗  ↑ReLU ──w[10]──↗   ↑ReLU       ├──→ ŷ (예측값)
 x₂=2   ──w[-1]──→ h1n1 ──w[-0.1]→ h2n1 ──w[25]──┘
         ──w[0]──↗  ↑ReLU ──w[0]──↗   ↑ReLU

───────────────────────────────────────────────────────────
 순전파 →→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→→
 역전파 ←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←←
```

## 2. 각 노드에서 하는 일 (공통 원리)

1. **선형결합**: $z = w_1 x_1 + w_2 x_2$  
2. **활성화**: $a = \text{ReLU}(z) = \max(0, z)$  
3. **다음 층으로 전달**: $a$가 다음 노드의 입력이 됨

## 3. 역전파 - 왜 체인룰인가?

가중치 $w$가 손실 $L$에 미치는 영향은 **여러 단계를 거치기** 때문에  
각 단계 미분을 모두 곱해야 합니다 (연쇄법칙):

$$\frac{\partial L}{\partial w} = \underbrace{\frac{\partial L}{\partial \hat{y}}}_{\text{① 손실→출력}} \times \underbrace{\frac{\partial \hat{y}}{\partial a}}_{\text{② 출력→활성값}} \times \underbrace{\frac{\partial a}{\partial z}}_{\text{③ ReLU 미분}} \times \underbrace{\frac{\partial z}{\partial w}}_{\text{④ 선형결합→가중치}}$$

- ① `dL/dŷ = 2(ŷ - y)` : MSE 미분
- ② `dŷ/da = 연결 가중치` : 선형결합이므로 해당 가중치 값
- ③ `da/dz = ReLU'(z)` : z>0이면 1, z≤0이면 0  
- ④ `dz/dw = 입력값` : z=wx 이므로 x 그 자체

→ 최종: **가중치 기울기 = 오차신호(δ) × 해당 가중치로 들어오는 입력**

## 4. 은닉층 오차신호가 가중합인 이유

hidden1 노드는 hidden2의 **여러 노드에 연결**되어 있습니다.  
각 경로로 전달된 오차를 모두 더해야 하므로:

$$\delta_{h1} = \left(\sum_k w_{h1 \to h2_k} \cdot \delta_{h2_k}\right) \times \text{ReLU}'(z_{h1})$$

예를 들어 h1n0는 h2n0와 h2n1 양쪽에 연결되어 있으므로  
두 경로에서 내려오는 오차를 **가중합** 후 ReLU 미분으로 게이팅합니다.

## 5. ReLU 미분이 0이면 무슨 의미?

- z≤0 → 해당 노드는 순전파 때 이미 0으로 클리핑됨  
- 그 노드를 거치는 경로는 **역전파 시 gradient가 차단**됨 (기여 없음)  
- 이를 "Dead Neuron" 문제라고도 부름

# 10번 셀: 순전파/역전파 통합 코드

````python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False

# ──────────────────────────────────────────────────────────────
# 설정
# ──────────────────────────────────────────────────────────────
inp = [30.0, 2.0]   # 입력: 집 크기(평), 방 개수
target = 18.0        # 정답
lr = 0.01            # 학습률

W = {
    'h1n0': [50.0, 10.0],   # hidden1 node0의 가중치 [w₁, w₂]
    'h1n1': [-1.0,  0.0],   # hidden1 node1의 가중치
    'h2n0': [ 0.1, 10.0],   # hidden2 node0의 가중치 (h1n0, h1n1에서 들어옴)
    'h2n1': [-0.1,  0.0],   # hidden2 node1의 가중치
    'out':  [ 0.1, 25.0],   # 출력층 가중치 (h2n0, h2n1에서 들어옴)
}

def relu(x):   return max(0.0, x)
def drelu(x):  return 1.0 if x > 0 else 0.0  # ReLU 미분: z>0 → 1, 아니면 0

# ══════════════════════════════════════════════════════════════
# STEP 1. 순전파 (Forward Pass)
# 원리: z = w₁x₁ + w₂x₂  →  a = ReLU(z)  →  다음 층 입력
# ══════════════════════════════════════════════════════════════
# 은닉층 1
h1n0_z = inp[0]*W['h1n0'][0] + inp[1]*W['h1n0'][1]   # 선형결합
h1n0_a = relu(h1n0_z)                                  # 활성화

h1n1_z = inp[0]*W['h1n1'][0] + inp[1]*W['h1n1'][1]
h1n1_a = relu(h1n1_z)

# 은닉층 2 (hidden1 출력이 입력)
h2n0_z = h1n0_a*W['h2n0'][0] + h1n1_a*W['h2n0'][1]
h2n0_a = relu(h2n0_z)

h2n1_z = h1n0_a*W['h2n1'][0] + h1n1_a*W['h2n1'][1]
h2n1_a = relu(h2n1_z)

# 출력층 (활성화 없음 - 회귀 문제이므로 그대로 출력)
y_hat = h2n0_a*W['out'][0] + h2n1_a*W['out'][1]

print("=" * 55)
print("[ STEP 1 ] 순전파(Forward Pass)")
print("=" * 55)
print(f"  h1n0: z={h1n0_z:.1f}  ReLU→ a={h1n0_a:.1f}")
print(f"  h1n1: z={h1n1_z:.1f}  ReLU→ a={h1n1_a:.1f}  ← 음수 → 0으로 클리핑")
print(f"  h2n0: z={h2n0_z:.2f}  ReLU→ a={h2n0_a:.2f}")
print(f"  h2n1: z={h2n1_z:.2f}  ReLU→ a={h2n1_a:.2f}  ← 음수 → 0으로 클리핑")
print(f"  예측값 ŷ = {y_hat:.4f}  (정답 y = {target})")

# ══════════════════════════════════════════════════════════════
# STEP 2. 손실 계산 (MSE)
# L = (ŷ - y)²  →  dL/dŷ = 2(ŷ - y)
# ══════════════════════════════════════════════════════════════
loss = (y_hat - target) ** 2
dL_dy = 2 * (y_hat - target)   # 손실의 출력에 대한 미분 (역전파 출발점)

print(f"\n[ STEP 2 ] 손실 L = ({y_hat:.4f} - {target})² = {loss:.4f}")
print(f"           dL/dŷ = 2×({y_hat:.4f}-{target}) = {dL_dy:.4f}")

# ══════════════════════════════════════════════════════════════
# STEP 3. 역전파 (Backward Pass) - 체인룰
#
# 출력층 가중치 기울기:
#   dL/dw_out = dL/dŷ × h2_노드_출력값
#   이유: ŷ = w0*h2n0 + w1*h2n1 이므로 dŷ/dw0 = h2n0
#
# 은닉층2 오차신호(δ):
#   δ_h2 = w_out × dL/dŷ × ReLU'(z_h2)
#   이유: 체인룰 - 출력층에서 내려온 gradient × 연결 가중치 × ReLU 게이트
#
# 은닉층1 오차신호(δ):
#   δ_h1 = (Σ w_h2k × δ_h2k) × ReLU'(z_h1)
#   이유: h1 노드는 h2의 여러 노드에 연결 → 모든 경로 오차를 가중합
# ══════════════════════════════════════════════════════════════
# 출력층 가중치 gradient
dL_dw_out0 = dL_dy * h2n0_a   # ∂L/∂w_out[0]
dL_dw_out1 = dL_dy * h2n1_a   # ∂L/∂w_out[1]

# 은닉층2 오차신호
d_h2n0 = W['out'][0] * dL_dy * drelu(h2n0_z)
d_h2n1 = W['out'][1] * dL_dy * drelu(h2n1_z)   # drelu=0 → gradient 차단

# 은닉층1 오차신호 (두 경로에서 오는 gradient를 가중합)
d_h1n0 = (W['h2n0'][0]*d_h2n0 + W['h2n1'][0]*d_h2n1) * drelu(h1n0_z)
d_h1n1 = (W['h2n0'][1]*d_h2n0 + W['h2n1'][1]*d_h2n1) * drelu(h1n1_z)  # drelu=0 → 차단

print("\n" + "=" * 55)
print("[ STEP 3 ] 역전파(Backward Pass)")
print("=" * 55)
print(f"  출력층  dL/dw0 = {dL_dy:.4f}×{h2n0_a:.2f} = {dL_dw_out0:.4f}")
print(f"           dL/dw1 = {dL_dy:.4f}×{h2n1_a:.2f} = {dL_dw_out1:.4f}")
print(f"  h2n0  δ = {W['out'][0]}×{dL_dy:.4f}×ReLU'({h2n0_z:.2f})={drelu(h2n0_z)} = {d_h2n0:.4f}")
print(f"  h2n1  δ = {W['out'][1]}×{dL_dy:.4f}×ReLU'({h2n1_z:.2f})={drelu(h2n1_z)} = {d_h2n1:.4f}  ← z≤0이라 gradient=0")
print(f"  h1n0  δ = ({W['h2n0'][0]}×{d_h2n0:.4f}+{W['h2n1'][0]}×{d_h2n1:.4f})×{drelu(h1n0_z)} = {d_h1n0:.6f}")
print(f"  h1n1  δ = ({W['h2n0'][1]}×{d_h2n0:.4f}+{W['h2n1'][1]}×{d_h2n1:.4f})×{drelu(h1n1_z)} = {d_h1n1:.6f}  ← z≤0이라 gradient=0")

# ══════════════════════════════════════════════════════════════
# STEP 4. 가중치 업데이트 (경사하강법)
# w_new = w_old - lr × ∂L/∂w
# 이유: gradient 방향이 손실 증가 방향이므로 반대로 이동
# ══════════════════════════════════════════════════════════════
W_new = {
    'h1n0': [W['h1n0'][0] - lr*inp[0]*d_h1n0,  W['h1n0'][1] - lr*inp[1]*d_h1n0],
    'h1n1': [W['h1n1'][0] - lr*inp[0]*d_h1n1,  W['h1n1'][1] - lr*inp[1]*d_h1n1],
    'h2n0': [W['h2n0'][0] - lr*h1n0_a*d_h2n0,  W['h2n0'][1] - lr*h1n1_a*d_h2n0],
    'h2n1': [W['h2n1'][0] - lr*h1n0_a*d_h2n1,  W['h2n1'][1] - lr*h1n1_a*d_h2n1],
    'out':  [W['out'][0]  - lr*dL_dw_out0,      W['out'][1]  - lr*dL_dw_out1],
}

print("\n" + "=" * 55)
print("[ STEP 4 ] 가중치 업데이트 (w ← w - lr × gradient)")
print("=" * 55)
for k in W_new:
    print(f"  {k}: {[round(v,4) for v in W[k]]} → {[round(v,4) for v in W_new[k]]}")

# ══════════════════════════════════════════════════════════════
# 시각화: 순전파(좌) & 역전파(우) 신경망 다이어그램
# ══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('신경망 순전파(Forward) & 역전파(Backward) 흐름', fontsize=14, fontweight='bold')

# 노드 위치 (x, y)
pos = {
    'x0':   (0.5, 3.7), 'x1':   (0.5, 1.8),
    'h1n0': (1.5, 4.2), 'h1n1': (1.5, 1.3),
    'h2n0': (2.5, 4.2), 'h2n1': (2.5, 1.3),
    'out':  (3.5, 2.75),
}
edges_list = [
    ('x0','h1n0'),('x0','h1n1'),
    ('x1','h1n0'),('x1','h1n1'),
    ('h1n0','h2n0'),('h1n0','h2n1'),
    ('h1n1','h2n0'),('h1n1','h2n1'),
    ('h2n0','out'), ('h2n1','out'),
]
colors = {
    'x0':'#AED6F1','x1':'#AED6F1',
    'h1n0':'#A9DFBF','h1n1':'#A9DFBF',
    'h2n0':'#FAD7A0','h2n1':'#FAD7A0',
    'out':'#F1948A',
}
fwd_labels = {
    'x0': f'x₁\n=30', 'x1': f'x₂\n=2',
    'h1n0': f'h1n0\na={h1n0_a:.0f}',
    'h1n1': f'h1n1\na={h1n1_a:.0f}\n(차단)',
    'h2n0': f'h2n0\na={h2n0_a:.2f}',
    'h2n1': f'h2n1\na={h2n1_a:.0f}\n(차단)',
    'out':  f'ŷ={y_hat:.2f}',
}
bwd_labels = {
    'x0': f'x₁=30', 'x1': f'x₂=2',
    'h1n0': f'δ=\n{d_h1n0:.4f}',
    'h1n1': f'δ=\n{d_h1n1:.4f}\n(차단)',
    'h2n0': f'δ=\n{d_h2n0:.4f}',
    'h2n1': f'δ=\n{d_h2n1:.4f}\n(차단)',
    'out':  f'dL/dŷ\n={dL_dy:.3f}',
}

for ax, node_labels, arrow_clr, forward, title in [
    (axes[0], fwd_labels, '#2E86C1', True,  '순전파 (→) 값이 앞으로 흐른다'),
    (axes[1], bwd_labels, '#C0392B', False, '역전파 (←) Gradient가 뒤로 흐른다'),
]:
    ax.set_xlim(0, 4); ax.set_ylim(0, 5.3)
    ax.set_title(title, fontsize=11, fontweight='bold', pad=8)
    ax.axis('off')

    for s, d in edges_list:
        sx, sy = pos[s]; dx, dy = pos[d]
        if forward:
            ax.annotate('', xy=(dx, dy), xytext=(sx, sy),
                        arrowprops=dict(arrowstyle='->', color=arrow_clr, lw=1.1))
        else:
            ax.annotate('', xy=(sx, sy), xytext=(dx, dy),
                        arrowprops=dict(arrowstyle='->', color=arrow_clr, lw=1.1))

    for name, (x, y) in pos.items():
        c = plt.Circle((x, y), 0.38, color=colors[name], zorder=3, ec='gray', lw=0.8)
        ax.add_patch(c)
        ax.text(x, y, node_labels[name], ha='center', va='center', fontsize=8, zorder=4)

    for xp, lb in [(0.5,'입력층'),(1.5,'은닉층1'),(2.5,'은닉층2'),(3.5,'출력층')]:
        ax.text(xp, 5.05, lb, ha='center', fontsize=9,
                bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

axes[1].text(2.0, 0.25, '← 체인룰: 뒤 gradient × 연결 가중치 × ReLU\'(z)',
             ha='center', color='tomato', fontsize=8.5)

plt.tight_layout()
plt.show()
````
