### 1. 로지스틱 회귀와 그 기원: 왜 탄생했을까?

결과표를 읽기 전에 이 모델이 무엇인지 아는 것이 중요하네. 로지스틱 회귀는 이름에 '회귀(Regression)'가 들어가지만, 실제로는 **분류(Classification)**, 특히 데이터가 어떤 클래스에 속할 확률(0과 1 사이의 값)을 예측하는 알고리즘이라네.

**역사적 배경:** 19세기에 피에르 프랑수아 베르휠스트(Pierre François Verhulst)가 인구 증가 모델을 연구하며 '로지스틱 함수(Logistic function)'를 고안했네. 인구가 무한히 증가할 수 없고 특정 한계치에 수렴한다는 자연의 법칙을 수학적으로 표현한 것이지. 20세기에 들어 통계학자 데이비드 콕스(David Cox)가 이를 통계적 분류 문제에 적용하면서 오늘날 AI와 통계학에서 가장 기초적이고 중요한 알고리즘으로 자리 잡게 되었네.

수학적으로, 선형 회귀의 결과값 $z = \beta_0 + \beta_1 x_1$를 0과 1 사이의 확률값 $p$로 변환하기 위해 **시그모이드 함수(Sigmoid Function)**를 사용하네.

$$p = \frac{1}{1 + e^{-z}} = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1)}}$$

### 3. 상단 요약표: 모델의 전반적인 성능 평가

> `Method: MLE` (최대우도추정법)
> `Log-Likelihood: -147.21` (로그 가능도)
> `LL-Null: -207.94` (널 로그 가능도)
> `Pseudo R-squ.: 0.2921` (유사 결정계수)
> `LLR p-value: 3.005e-28` (우도비 검정 p-값)

이 부분이 이 보고서의 수학적 핵심이네. 하나씩 뜯어보지.

**1) Method: MLE (Maximum Likelihood Estimation, 최대우도추정법)**
*   **정의:** 주어진 데이터가 관측될 확률(Likelihood)을 가장 높게(Maximum) 만드는 모델의 파라미터(가중치)를 찾는 통계적 방법론이네. 로널드 피셔(R.A. Fisher)가 확립한 통계학의 근간이지.
*   **수학 식:** 데이터가 독립적이라고 가정할 때 우도 함수 $L(\beta)$는 다음과 같네.
    $$L(\beta) = \prod_{i=1}^{n} p_i^{y_i} (1-p_i)^{1-y_i}$$
    여기에 계산을 쉽게 하기 위해 자연로그($\ln$)를 씌운 것이 바로 **Log-Likelihood(로그 가능도)**라네. 이 식을 미분하여 기울기가 0이 되는 지점을 찾는 것이 바로 딥러닝과 머신러닝 학습의 본질이지.

**2) Log-Likelihood & LL-Null**
*   **Log-Likelihood (-147.21):** 자네가 만든 모델(변수 $x_1$이 포함된 모델)의 로그 가능도 값이네. 음수이며, 0에 가까울수록 좋은 것이지.
*   **LL-Null (-207.94):** 아무런 독립 변수 없이 오직 '상수항(const)'만으로 예측했을 때(즉, 무식하게 평균 확률로만 예측했을 때)의 로그 가능도라네.
*   *해석:* 모델을 적용하니 -207.94에서 -147.21로 값이 증가(개선)했군! 즉, 자네의 변수 $x_1$이 매우 유용하다는 뜻이네.

**3) Pseudo R-squ. (유사 R제곱, 0.2921)**
*   **설명:** 선형 회귀에는 설명력을 나타내는 $R^2$가 있지만, 로지스틱 회귀에는 완벽히 대응되는 지표가 없네. 그래서 맥파든(McFadden)이라는 학자가 고안한 유사 R제곱을 사용하지.
*   **수학 식:** $R_{pseudo}^2 = 1 - \frac{\ln(L_{model})}{\ln(L_{null})}$
*   *해석:* 모델이 데이터를 약 29.2% 정도 설명하고 있음을 의미하네. (보통 0.2~0.4 사이면 훌륭한 모델로 평가받네.)

**4) LLR p-value (우도비 검정 유의확률, 3.005e-28)**
*   **설명:** "이 모델이 변수가 아예 없는 껍데기 모델(Null model)보다 통계적으로 유의미하게 나은가?"를 검증하는 지표네.
*   *해석:* 값이 $3.005 \times 10^{-28}$이네. 0에 한없이 가까운 매우 작은 수치지. (보통 0.05 미만이면 유의미하다고 봄). 즉, 자네의 모델은 통계적으로 매우 훌륭하고 유의미하다는 결론을 내릴 수 있네.

### 4. 하단 계수표: 각 변수의 영향력 분석

> `coef` (회귀계수), `std err` (표준오차), `z` (Z-통계량), `P>|z|` (p-값)

이제 구체적인 방정식의 형태를 볼 차례네. 자네의 모델은 다음과 같은 수식으로 도출되었네.
$$\ln\left(\frac{p}{1-p}\right) = 0.0837 + 2.1228 \cdot x_1$$
(여기서 좌변은 '로그 오즈(Log-Odds)'라고 부르며, 어떤 사건이 일어날 확률을 일어나지 않을 확률로 나눈 값에 로그를 취한 것이네.)

*   **coef (계수):** $x_1$의 계수가 2.1228이군. 이는 $x_1$이 1단위 증가할 때마다 로그 오즈가 2.1228만큼 증가한다는 뜻이네. (즉, 양의 상관관계가 아주 강함)
*   **std err (표준오차):** 계수 추정치의 불확실성을 나타내네. 작을수록 좋지.
*   **z (Wald Z-통계량):** $z = \frac{\text{coef}}{\text{std err}}$ 로 계산되네. $x_1$의 경우 $2.1228 / 0.254 \approx 8.343$ 이군. 이 값이 0에서 멀어질수록 해당 변수가 중요하다는 강력한 증거가 되네.
*   **P>|z| (p-값):** $x_1$의 p-값이 0.000이군. 이는 $x_1$이 종속 변수 $y$를 예측하는 데 있어 우연이 아니라 통계적으로 **절대적으로 유의미한 변수**임을 증명하네.

```python
# 필요한 라이브러리 불러오기
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd

# 데이터셋 생성
X, y = make_classification(n_samples=100, n_features=2,
                           n_informative=2, n_redundant=0,
                           random_state=0)
data = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
data['Target'] = y

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(data[['Feature1', 'Feature2']], data['Target'], test_size=0.3, random_state=0)

# 모델 구축 및 훈련
model = LogisticRegression()
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test) 

# 성능 평가
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# 모델 해석
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
