# 피처 선택 ( Feature Selection ) 

- overfitting을 줄이고 중요한 피처들만 선정하는 과정 
- 피처 생성은 "어떻게 유용한 피처를 만들 것인가에 대한" 문제 였다면, 피처 선택은 "어떻게 유용한 피처를 고르는가"의 문제이다. 

## 피처 선택을 해야 하는 상황은 

- 많은 수의 피처가 있는 경우 
- 피처 간 높은 상관관계가 있는 경우 
  - `다중 공산성의 문제`
- 노이즈 피처가 포함 된 경우 
- 계산 비용을 절감해야 할 때 
- 모델의 해석성을 높이고 싶을 때

## 피처 선택 기법

### 필터 방법 ( Filter Methods )

- 모델 학습 전에 피처들의 통계적 특성을 평가하여 중요도를 측정
- `상관계수`, `상호정보량`, `카이제곱 검정` 등의 기법을 사용해 각 피처와 타겟 변수 간의 통계적 연관성을 계산
- 연관성이 낮은 피처들은 제거하고, 중요도가 높은 피처들만 남김 

### 래퍼 방법 ( Wrapper Methods )

- 피처를 하나씩 빼고, 넣고를 반복하면서 실제 성능을 확인하면서 선택 
- 여러 모델을 학습시켜야 하므로 계산 비용이 상당이 높아짐. 
- 특정 피처 집합에 과적합될 위험도 있음

### 임베디드 방법 ( Embedded Methods )

- 모델이 스스로 피처의 중요도를 계산하고, 그 중요도를 바탕으로 어떤 피처를 사용할지를 결정합니다. 즉, 피처 선택과 모델학습이 동시에 이루어짐. 
- 계산 효율성과 모델 성능을 모두 고려
- 모델 종류에 따라 사용 가능 여부가 달라짐 

## Example 

-  target변수인 'credit'을 제외하고 총 몇개의 피처가 있는지 봅시다

```
import numpy as np 
import pandas as pd

train_credit = pd.read_csv('train_credit.csv')

# 상위행 출력
display(train_credit.head())
# 컬럼명 출력
display(train_credit.columns)
# 컬럼 갯수
num_features = len(train_credit.columns)-1

display(num_features)
```

## 기본 성능 검증 

- [RandomForest Information](./2.3.1.RandomForest.md)  


```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = train_credit.drop('credit', axis=1); 
y = train_credit['credit'] 

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=42) 
model_rf = RandomForestClassifier(random_state = 42) #학습 과정에서의 무작위 성을 제어 

# 모델 학습
model_rf.fit(X_train,y_train)

# 예측
y_pred = model_rf.predict(X_val) 

# accuracy 정확도
accuracy = accuracy_score(y_val, y_pred)

# precision 정밀도, 모델이 얼마나 정밀하게 양성 샘플을 예측했는지 
precision = precision_score(y_val, y_pred, average='macro')

# recall 재현율, 모델이 얼마나 정밀하게 양성 샘플을 예측 했는지
recall = recall_score(y_val, y_pred, average='macro')

# f1 score, 정밀도와 재현율의 조화 평균으로, 두 지표의 균형을 고려한 성능 지표 
f1 = f1_score(y_val, y_pred, average='macro')

# average='macro'는 각 클래스에 대한 성능을 동일하게 평균 내어 전체 성능을 계산 

display(f'Accuracy: {accuracy}')
display(f'Precision: {precision}')
display(f'Recall: {recall}')
display(f'F1-Score: {f1}')
```
