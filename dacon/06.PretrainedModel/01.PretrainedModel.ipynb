- 사전 학습 된 모델은 AI가 생물, 객체, 장소 및 사람을 빠르게 인식할 수 있도록 도와준다. 
- 이를 통해 사전 학습 된 모델은 스포츠, 스마트 시티 등에서 인간과 같은 비전 기느을 제공 
- 객체 인식 모델은 CNN 또는 영역 기반 합성곱 신경망(R-CNN) 을 기초로 사용, 사진이나 비니오에서 객체를 인식하고 분류하도록 훈련 된다.

- 사전 학습 된 AI 모델은 화학의 언어를 이해하고 실제 분자의 원자 간 관계를 학습하여 빠른 신약 발견을 위한 강력한 도구를 제공

- 사이버 보안 솔루션을 구현, 인간 보안 분석가의 능력을 확장하여 위협을 더 빠르게 감지할 수 있도록 도와준다. 
- 예로는, 사람과 기걔의 디지털 지문 인식, 이상 감지, 민감 정보 및 피싱 탐지 등이 있다. 

## 사전 학습된 모델의 주요 아키텍처

사전 학습 모델을 사용할 때는 최신 모델인지보다 **현재 데이터와 배포 조건에 맞는 표현력·연산량·메모리·추론 속도**를 갖췄는지가 더 중요하다. 아래 모델들은 컴퓨터 비전 아키텍처가 `작은 CNN → 깊은 CNN → 효율적인 CNN → Transformer`로 발전한 주요 흐름을 보여준다.

### 1. 전통적인 CNN 모델

#### LeNet-5 (1998)

- **등장 배경:** 우편번호와 수표에 쓰인 손글씨 숫자를 자동으로 인식하기 위해 제안된 초기 CNN이다. 사람이 특징을 직접 설계하는 대신 픽셀에서 분류 결과까지 전체 과정을 gradient descent로 학습할 수 있음을 보여주었다.
- **주요 구조:** `Conv → Subsampling/Pooling → Conv → Pooling → Fully Connected` 순서로 공간 특징을 점차 압축한다. 원래 입력은 작은 흑백 이미지이며, 오늘날 CNN에서 사용하는 지역 연결(local connectivity)과 가중치 공유(weight sharing)의 기본 형태를 갖췄다.
- **핵심 아이디어:** 같은 convolution kernel을 이미지 전체에 공유하므로 위치마다 별도 파라미터를 사용하는 fully connected network보다 적은 파라미터로 선·모서리·획 같은 지역 특징을 감지한다. Pooling은 공간 크기를 줄이고 작은 위치 변화에 덜 민감한 표현을 만든다.
- **장점:** 구조가 작고 이해하기 쉬워 CNN의 convolution, pooling, flatten, classifier 흐름을 학습하기 좋다. MNIST와 같은 작은 단순 이미지의 기준 모델로 사용할 수 있다.
- **한계:** 깊이와 채널 수가 작아 고해상도 자연 이미지나 복잡한 객체를 표현하기에는 부족하다. 현대적인 BatchNorm, residual connection, 대규모 사전 학습 구조도 포함하지 않는다.
- **역사적 의미:** 현대 이미지 분류 CNN의 기본 설계 원리를 확립한 모델이다. [원 논문](https://leon.bottou.org/papers/lecun-98h)

#### AlexNet (2012)

- **등장 배경:** 대규모 자연 이미지 데이터인 ImageNet에서 깊은 CNN과 GPU 학습의 효과를 입증했으며 ILSVRC 2012 분류 부문에서 큰 성능 차이로 우승했다.
- **주요 구조:** 5개의 convolution layer와 3개의 fully connected layer로 구성된다. 앞부분의 큰 kernel과 stride로 이미지를 빠르게 축소하고 뒤쪽에서 더 추상적인 특징을 학습한다.
- **ReLU:** 당시 자주 쓰이던 sigmoid나 tanh 대신 `ReLU(x) = max(0, x)`를 적극적으로 사용해 깊은 CNN의 학습을 빠르게 했다. ReLU를 최초로 만든 모델이라는 뜻보다는 대규모 이미지 학습에서 효과를 널리 입증하고 대중화했다는 의미가 정확하다.
- **Dropout:** 큰 fully connected layer의 일부 뉴런을 학습 중 무작위로 비활성화해 특정 뉴런 조합에 지나치게 의존하는 과적합을 줄였다.
- **Data augmentation:** random crop, horizontal reflection, 색상 변화 등을 사용해 제한된 학습 이미지에서 더 다양한 입력을 만들었다.
- **장점:** CNN, 대규모 데이터, GPU 병렬 연산을 결합하면 수작업 특징보다 뛰어난 표현을 학습할 수 있음을 보여주며 딥러닝 기반 컴퓨터 비전의 전환점이 되었다.
- **한계:** 초기의 큰 convolution kernel과 거대한 fully connected layer 때문에 파라미터와 계산 비용이 크다. 현재는 효율성과 정확도 면에서 더 좋은 모델이 많아 교육·비교 목적 외에는 신규 서비스의 첫 선택이 되는 경우가 적다.
- **논문:** [ImageNet Classification with Deep Convolutional Neural Networks](https://papers.nips.cc/paper_files/paper/2012/hash/c399862d3b9d6b76c8436e924a68c45b-Abstract.html)

### 2. 깊은 네트워크 모델

#### VGGNet (2014)

- **핵심 질문:** CNN의 다른 조건을 비교적 단순하게 유지하면서 depth를 증가시키면 이미지 표현력이 좋아지는가를 실험했다. 대표 모델은 VGG-16과 VGG-19이다.
- **주요 구조:** 큰 kernel 하나보다 `3×3 Conv`를 반복해서 쌓고, stage 사이에서 `2×2 MaxPool`로 공간 크기를 절반씩 줄인다. 채널 수는 일반적으로 64 → 128 → 256 → 512처럼 증가한다.
- **작은 kernel을 반복하는 이유:** 3×3 convolution 두 개는 5×5에 가까운 receptive field를, 세 개는 7×7에 가까운 receptive field를 만들면서 그 사이에 activation을 더 넣을 수 있다. 동일한 입력·출력 채널을 단순 가정하면 하나의 큰 kernel보다 파라미터를 줄일 수도 있다.
- **장점:** 규칙적이고 단순한 블록 구조 덕분에 이해·구현하기 쉽다. 중간 feature가 이미지 분류뿐 아니라 객체 탐지, 스타일 전이 등의 backbone으로 널리 활용되었다.
- **한계:** 마지막 fully connected layer를 포함해 파라미터 수가 매우 크고, convolution 연산량과 activation 메모리도 많다. Residual connection이 없어 더 깊게 확장할 때 최적화가 어렵다.
- **실무 관점:** 구조 해석이나 feature visualization에는 좋지만, 제한된 GPU·모바일 환경에서는 ResNet 또는 경량 모델이 보통 더 효율적이다. [원 논문](https://arxiv.org/abs/1409.1556)

#### GoogLeNet / Inception v1 (2014)

- **핵심 질문:** 정확도를 높이기 위해 depth와 width를 늘리면서도 연산 예산을 어떻게 제한할 것인가에 초점을 맞췄다. GoogLeNet은 Inception architecture를 사용한 22-layer 모델이다.
- **Inception module:** 하나의 입력에 `1×1 Conv`, `3×3 Conv`, `5×5 Conv`, `Pooling`을 병렬 적용한 후 channel 방향으로 결과를 연결한다. 서로 다른 크기의 receptive field로 작은 패턴과 큰 패턴을 동시에 추출하는 multi-scale 구조다.
- **1×1 bottleneck:** 비싼 3×3 또는 5×5 convolution 전에 1×1 convolution으로 channel을 줄여 연산량을 낮춘다. 1×1 convolution은 공간 위치는 유지하면서 channel feature를 혼합하고 축소한다.
- **Global Average Pooling:** 큰 fully connected classifier 대신 각 feature map의 공간 평균을 사용해 파라미터와 과적합을 줄였다.
- **Auxiliary classifier:** 깊은 중간 layer에 보조 분류기를 연결해 학습 신호 전달과 regularization을 돕도록 설계했다. 추론 시에는 일반적으로 최종 classifier를 사용한다.
- **장점:** VGG처럼 단순히 layer를 크게 쌓는 방식보다 계산 자원을 세밀하게 배분한다. Multi-branch와 bottleneck 설계는 이후 여러 효율적 architecture에 영향을 주었다.
- **한계:** Branch별 kernel과 channel 수를 설계해야 해 구조가 복잡하며, 병렬 branch의 실제 속도는 hardware와 구현에 따라 이론적 FLOPs만큼 효율적이지 않을 수 있다.
- **논문:** [Going Deeper with Convolutions](https://openaccess.thecvf.com/content_cvpr_2015/html/Szegedy_Going_Deeper_With_2015_CVPR_paper.html)

#### ResNet (2015, CVPR 2016)

- **해결한 문제:** Plain CNN은 깊이를 늘렸을 때 표현력은 충분해도 optimization이 어려워 training error까지 증가하는 degradation problem이 발생할 수 있다.
- **Residual block:** 목표 변환 전체 `H(x)`를 직접 학습하는 대신 입력에 더할 변화량 `F(x)`를 학습한다. 블록 출력은 `y = x + F(x)`이다.
- **Shortcut connection:** 입력 `x`가 residual branch를 우회해 출력으로 직접 전달된다. Backpropagation에도 identity 경로가 생겨 매우 깊은 모델에서 정보와 gradient 전달이 쉬워진다. 자세한 수학과 구현은 [Residual Connections](./01.Residual_connections.md)를 참고한다.
- **Shape 변경:** 입력과 출력의 channel 또는 공간 크기가 다르면 1×1 projection convolution으로 shortcut shape을 맞춘다.
- **모델 계열:** ResNet-18/34는 주로 BasicBlock을, ResNet-50/101/152는 1×1 → 3×3 → 1×1 형태의 Bottleneck block을 사용한다.
- **장점:** 학습이 안정적이고 범용 feature extractor로 검증되어 이미지 분류·탐지·분할의 강력한 baseline이다. 사전 학습 모델과 참고 구현도 풍부하다.
- **한계:** 깊은 버전은 순차 layer 수 때문에 latency가 증가하며, 최신 경량 모델보다 정확도 대비 연산 효율이 낮을 수 있다. Residual connection이 모든 vanishing/exploding gradient 문제를 자동으로 제거하는 것도 아니다.
- **실무 관점:** 새 이미지 프로젝트에서 ResNet-18 또는 ResNet-50은 데이터 파이프라인과 성능을 검증하기 좋은 기준선이다. [원 논문](https://openaccess.thecvf.com/content_cvpr_2016/html/He_Deep_Residual_Learning_CVPR_2016_paper.html)

### 3. 효율성과 경량화를 강조한 모델

#### MobileNet v1 (2017)

- **목표:** 모바일·임베디드 장치처럼 계산량, 메모리, 전력 사용이 제한된 환경에서 사용할 수 있는 CNN을 설계한다.
- **Depthwise separable convolution:** Standard convolution을 `Depthwise Conv → 1×1 Pointwise Conv`로 분리한다. Depthwise convolution은 각 입력 channel의 공간 특징을 독립적으로 처리하고, pointwise convolution은 channel 사이 정보를 혼합한다.
- **연산량 차이:** Standard convolution의 대략적인 비용은 `H×W×Cin×Cout×K²`이고, depthwise separable convolution은 `H×W×Cin×K² + H×W×Cin×Cout`이다. 출력 채널이 클수록 계산 절감 효과가 커질 수 있다.
- **Width multiplier:** 모든 channel 수를 비율 `α`로 줄여 모델의 표현력과 연산량을 함께 조절한다.
- **Resolution multiplier:** 입력과 중간 feature map의 해상도를 `ρ` 비율로 줄여 연산량을 조절한다.
- **장점:** 크기와 계산량을 체계적으로 줄일 수 있고, 분류뿐 아니라 탐지·얼굴 속성·위치 인식 등에 backbone으로 사용할 수 있다.
- **한계:** FLOPs가 낮다고 목표 스마트폰에서 반드시 빠른 것은 아니다. Depthwise 연산의 kernel 최적화와 memory access 특성에 따라 실제 latency가 달라지므로 배포 장치에서 측정해야 한다.
- **후속 모델:** MobileNetV2는 inverted residual과 linear bottleneck, MobileNetV3는 architecture search와 squeeze-and-excitation 등을 결합해 발전했다. [원 논문](https://arxiv.org/abs/1704.04861)

#### EfficientNet (2019)

- **핵심 문제:** 기존 CNN Scaling은 depth, width, input resolution 중 한 축만 경험적으로 크게 만드는 경우가 많았다. EfficientNet은 세 축의 균형을 함께 고려했다.
- **EfficientNet-B0:** Neural Architecture Search로 찾은 기준 모델이며 MBConv, depthwise convolution, Squeeze-and-Excitation, SiLU/Swish 계열 activation과 residual connection을 사용한다.
- **Compound Scaling:** 하나의 계수 `φ`로 depth, width, resolution을 `d = α^φ`, `w = β^φ`, `r = γ^φ`처럼 함께 확장한다. 세부 수학과 계산량 유도는 [Compound Scaling](./01-02.Compound_Scaling.md)을 참고한다.
- **모델 계열:** B0에서 B7로 갈수록 block 반복 수, channel 수, 대표 입력 해상도가 함께 증가한다. 큰 번호가 항상 실무 최적 모델이라는 뜻은 아니다.
- **장점:** 같은 수준의 정확도를 목표로 할 때 기존 CNN보다 적은 파라미터와 FLOPs를 사용할 수 있는 좋은 accuracy-efficiency 균형을 제시했다.
- **한계:** Depthwise convolution과 큰 입력 해상도 때문에 이론적 FLOPs 대비 실제 학습·추론 속도가 hardware에서 기대와 다를 수 있다. 큰 B 계열은 activation memory와 학습 시간이 많이 필요하다.
- **실무 관점:** 작은 데이터에서는 B0/B1부터 비교하고, 정확도뿐 아니라 목표 장치의 latency와 peak memory를 함께 측정해야 한다. [원 논문](https://proceedings.mlr.press/v97/tan19a.html)

### 4. Transformer 기반 모델

#### Vision Transformer, ViT (2020, ICLR 2021)

- **핵심 아이디어:** 이미지를 고정 크기 patch로 나누고 각 patch를 자연어의 token처럼 취급해 Transformer encoder로 처리한다. Transformer의 기초와 attention 수식은 [Transformer Architecture](./01-03.Transformer.md)를 참고한다.
- **Patch embedding:** `H×W` 이미지를 `P×P` patch로 나누면 patch token 수는 `(H/P)×(W/P)`이다. 각 patch를 펼친 뒤 선형 projection으로 `d_model` 차원의 token embedding을 만든다. Kernel과 stride가 모두 patch 크기인 convolution으로 동일한 projection을 구현할 수도 있다.
- **Position embedding:** Self-attention만으로는 patch 순서를 알 수 없으므로 각 patch의 위치 정보를 더한다.
- **Class token:** 학습 가능한 `[CLS]` token을 patch 앞에 추가하고, encoder를 통과한 class token 표현을 최종 이미지 분류에 사용한다. Mean pooling을 사용하는 변형도 있다.
- **Self-attention:** 각 patch가 멀리 떨어진 다른 patch도 직접 참고할 수 있어 이미지 전역 관계를 모델링한다. Attention 비용은 token 수의 제곱에 비례하므로 작은 patch나 큰 해상도에서는 비용이 급격히 증가한다.
- **장점:** 대규모 사전 학습과 적절한 regularization을 사용할 때 뛰어난 전이 성능을 낼 수 있고, architecture를 텍스트·이미지·멀티모달 영역에 공통으로 적용하기 쉽다.
- **한계:** 기본 ViT는 CNN의 locality와 translation equivariance 같은 inductive bias가 약해 데이터가 작을 때 불리할 수 있다. 또한 단일 해상도의 patch sequence를 유지해 탐지·분할에 필요한 다중 해상도 feature를 직접 제공하지 않는다.
- **실무 관점:** 사전 학습 weight가 기대하는 image size, crop, interpolation, normalization을 맞추고 position embedding의 입력 해상도 호환성을 확인해야 한다. [원 논문](https://arxiv.org/abs/2010.11929)

#### Swin Transformer (2021)

- **목표:** ViT의 global attention 비용과 단일 해상도 표현을 보완해 분류뿐 아니라 객체 탐지·semantic segmentation에 사용할 범용 vision backbone을 만든다.
- **Window Multi-Head Self-Attention:** 전체 patch 사이에서 attention을 계산하지 않고 고정된 작은 window 안에서만 계산한다. Window 크기를 고정하면 이미지 크기에 대해 계산량이 대략 선형으로 증가한다.
- **Shifted Window:** 다음 block에서 window 경계를 이동한다. 첫 block의 서로 다른 window에 속해 직접 만나지 못한 patch들이 다음 block에서 같은 window에 포함되어 영역 사이 정보가 전달된다.
- **Hierarchical representation:** Stage 사이의 patch merging으로 공간 해상도를 줄이고 channel을 늘린다. CNN의 feature pyramid처럼 저수준 고해상도 특징에서 고수준 저해상도 특징으로 진행한다.
- **장점:** Local attention으로 비용을 줄이면서 shifted window를 통해 영역 사이 연결을 만들고, 다중 scale feature 덕분에 분류·탐지·분할에 폭넓게 사용할 수 있다.
- **한계:** Window partition, cyclic shift, attention mask, patch merging 때문에 기본 ViT보다 구현이 복잡하다. Local window만으로는 한 layer에서 이미지 전체 관계를 직접 계산하지 못하고 여러 block을 거쳐 정보가 전파된다.
- **실무 관점:** 고해상도 입력이나 dense prediction에서는 기본 ViT보다 효율적일 수 있지만, 실제 latency는 window 연산과 framework kernel 최적화에 따라 달라지므로 목표 hardware에서 측정해야 한다. [원 논문](https://openaccess.thecvf.com/content/ICCV2021/html/Liu_Swin_Transformer_Hierarchical_Vision_Transformer_Using_Shifted_Windows_ICCV_2021_paper.html)

### 5. 아키텍처 비교 요약

| 모델 | 핵심 구조 | 강점 | 주요 한계 | 적합한 사용 예 |
|---|---|---|---|---|
| LeNet-5 | 작은 Conv + Pooling | 단순하고 교육에 적합 | 복잡한 자연 이미지 표현 부족 | MNIST, CNN 기초 실습 |
| AlexNet | Deep CNN + ReLU + Dropout | 대규모 CNN 학습의 역사적 기준 | 큰 FC layer와 높은 비용 | 역사 학습, 고전 baseline |
| VGGNet | 반복 3×3 Conv | 규칙적 구조, 범용 feature | 많은 파라미터와 연산량 | Feature 분석, 고전 backbone |
| GoogLeNet | Multi-branch Inception | Multi-scale 특징과 계산 절감 | 복잡한 branch 설계 | 효율적 multi-scale CNN 이해 |
| ResNet | Residual block | 안정적인 깊은 학습, 강력한 baseline | 깊은 모델의 latency | 범용 분류·탐지·분할 backbone |
| MobileNet | Depthwise separable Conv | 모바일용 경량화 | Hardware별 실제 속도 차이 | Edge·모바일 추론 |
| EfficientNet | MBConv + Compound Scaling | 정확도와 효율의 균형 | 큰 해상도와 depthwise 비용 | 제한된 자원의 고성능 분류 |
| ViT | Global Self-Attention | 전역 관계와 대규모 전이학습 | Token 수의 제곱 attention 비용 | 이미지 분류, 멀티모달 backbone |
| Swin Transformer | Shifted Window + Hierarchy | 고해상도와 dense task에 유리 | 구현 복잡성 | 탐지·분할·고해상도 vision |

### 6. 실무 선택 가이드

- **빠른 기준선이 필요하면:** ResNet-18 또는 ResNet-50부터 시작해 데이터 pipeline과 metric을 검증한다.
- **모바일·임베디드 배포라면:** MobileNet 계열을 우선 검토하고 실제 장치의 latency, 메모리, 전력 사용량을 측정한다.
- **정확도와 모델 크기의 균형이 중요하면:** EfficientNet-B0/B1 같은 작은 모델부터 비교한다.
- **대규모 사전 학습과 이미지 분류가 중심이면:** ViT를 검토하되 입력 해상도와 patch 수에 따른 attention 비용을 확인한다.
- **탐지·분할 또는 고해상도 이미지라면:** 계층적 feature를 제공하는 ResNet, Swin Transformer 등의 backbone을 비교한다.
- **최종 선택 시:** 논문의 정확도나 FLOPs만 보지 말고 현재 데이터의 validation metric, peak memory, throughput, 단일 이미지 latency를 같은 환경에서 측정한다.
