```python
import torch
import torch.nn as nn

x = torch.rand(1, 3, 5, 5) # 1개의 샘플(batch), 3개의 채널(channel), 5x5 크기의 값을  가지는 랜덤 텐서를 생성

flatten = nn.Flatten() #레이어는 입력 텐서를 평탄화하는데 사용된다. 
x_flattened = flatten(x) #x를 Flatten 레이어를 통해 변환합니다. 

print("Original shape:", x.shape)
print("Original data:", x)
print("Flattened shape:", x_flattened.shape)
print("Flattened data:", x_flattened)
```

## Fully Connected Layer

```python
fc_layer = nn.Linear(in_features=75, out_features=3)#평탄화 된 텐서의 차원수와 일치 해야함. 
activation = nn.ReLU()  

output = activation(fc_layer(x_flattened))  

print(output) 
```

```python
#완전 연결 레이어 층이 2개인 경우 
fc_layer1 = nn.Linear(in_features=75, out_features=20)  
fc_layer2 = nn.Linear(in_features=20, out_features=3)   
activation = nn.ReLU()  

output1 = activation(fc_layer1(x_flattened))  
output2 = fc_layer2(output1)  # 두 번째 완전 연결 레이어를 적용합니다.

print(output2)  
```

# Softmax function



```python
import torch.nn.functional as F

fc_layer1 = nn.Linear(in_features=75, out_features=40)  
fc_layer2 = nn.Linear(in_features=40, out_features=20)
fc_layer3 = nn.Linear(in_features=20, out_features=3)   
activation = nn.ReLU()  # 활성화 함수로 ReLU 사용

output1 = activation(fc_layer1(x_flattened)) 
output2 = activation(fc_layer2(output1))       
output3 = fc_layer3(output2)                   

probabilities = F.softmax(output3, dim=1)

print("Probabilities:", probabilities)
print("Sum of probabilities:", torch.sum(probabilities))
print("softmax value sum: ", torch.sum(probabilities).item())  # .item()을 사용하여 스칼라 값으로 변환
```

# 3D 이미지와 필터를 텐서로 구현 

```python
import numpy as np
import matplotlib.pyplot as plt

# 빨간색 채널
red_channel = np.array([
    [1, 2, 0, 2, 1],
    [0, 1, 1, 0, 0],
    [1, 0, 2, 0, 1],
    [0, 1, 1, 2, 0],
    [1, 0, 1, 0, 0]
])

# 녹색 채널
green_channel = np.array([
    [0, 1, 1, 0, 1],
    [0, 2, 1, 1, 0],
    [0, 0, 2, 0, 1],
    [0, 0, 1, 1, 0],
    [1, 0, 2, 2, 0]
])

# 파란색 채널
blue_channel = np.array([
    [1, 0, 2, 0, 1],
    [0, 0, 0, 1, 1],
    [1, 0, 2, 1, 2],
    [1, 0, 1, 0, 0],
    [0, 0, 1, 2, 0]
])

# 필터의 각 채널
filter_red_channel = np.array([[1, 0], [0, 1]])
filter_green_channel = np.array([[1, 2], [0, 1]])
filter_blue_channel = np.array([[2, 0], [0, 0]])

# 3차원 입력 배열
image_color = np.stack((red_channel, green_channel, blue_channel), axis=-1)
filter_color = np.stack((filter_red_channel, filter_green_channel, filter_blue_channel), axis=-1)

print(filter_color)
print(f"image color shape: {image_color.shape}")
print(f"filter color shape: {filter_color.shape}")
```

```python
input_tensor = torch.tensor(image_color, dtype=torch.float).unsqueeze(0).permute(0, 3, 1, 2)
# channel-first 형식으로 변환 (batch, channel, height, width)
filter_tensor = torch.tensor(filter_color, dtype=torch.float).unsqueeze(0)  
filter_tensor = filter_tensor.permute(0,3,1,2) 

conv_layer = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=2)
#Conv2d 레이어를 생성하고, 입력 채널 수와 출력 채널 수를 1로 설정하고, 커널 크기를 2로 설정합니다.
with torch.no_grad():
    conv_layer.weight = nn.Parameter(filter_tensor)
    conv_layer.bias = nn.Parameter(torch.zeros(1)) 
#conv_layer의 가중치와 편향을 설정합니다. 가중치는 filter_tensor로 설정하고, 편향은 0으로 초기화합니다.
output_tensor = conv_layer(input_tensor)
#conv_layer를 사용하여 input_tensor에 대해 합성곱 연산을 수행하고, 결과를 output_tensor에 저장합니다.
print('Output shape after convolution:', output_tensor.shape)
print('Output tensor after convolution:', output_tensor)

#max pooling layer를 생성하고, kernel_size와 stride를 2로 설정합니다. 2x2 영역에서 최대값을 추출하고, stride를 2로 설정하여 겹치지 않게 이동합니다.
# moving window 방식으로 입력 텐서를 슬라이딩하면서 최대값을 추출합니다.
max_pool_layer = nn.MaxPool2d(kernel_size=2, stride=2)
output_tensor_max_pool = max_pool_layer(output_tensor)

print('Output shape after max pooling:', output_tensor_max_pool.shape)
print('Output tensor after max pooling:', output_tensor_max_pool)
```

# 완전 연결층이 3개인 경우

```python
# Flatten 적용
flatten = nn.Flatten()
flattened = flatten(output_tensor_max_pool)

# 완전 연결 레이어 정의
fc1 = nn.Linear(in_features=4, out_features=50)  
fc2 = nn.Linear(in_features=50, out_features=20)  
fc3 = nn.Linear(in_features=20, out_features=10)  
fc4 = nn.Linear(in_features=10, out_features=3)   

# 활성화 함수
relu = nn.ReLU()
softmax = nn.Softmax()

# 은닉층을 통한 전파
hidden_output1 = relu(fc1(flattened))      # 첫 번째 완전 연결 레이어와 ReLU 활성화
hidden_output2 = relu(fc2(hidden_output1)) # 두 번째 완전 연결 레이어와 ReLU 활성화
hidden_output3 = relu(fc3(hidden_output2)) # 세 번째 완전 연결 레이어와 ReLU 활성화
output = softmax(fc4(hidden_output3))      # 최종 완전 연결 레이어와 소프트맥스 활성화

print("Flattened input:\n", flattened)
print("\nHidden layer output shape:\n", hidden_output3.shape)
print("\nFinal output shape:\n", output.shape)
print("\nFinal output (probabilities):\n", output)
```
