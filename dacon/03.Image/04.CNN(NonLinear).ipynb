# 1. 비선형성

- 데이터의 복잡한 패턴과 관계를 학습하기 위해서는 비선형성이 필수적 
- 신경망에서 다차원 공간에서 데이터를 비틀고 구부릴 수 있는 능력을 부여 
- CNN에서도 이 비선형성은 매우 중요, 합성곱 레이어의 결과에 비선형 함수를 적용하여 더욱 복잡한 특성을 추출할 수 있게 하는것 

- 합성 곱 후에 주로 사용되는 비선형 함수중 하나가 `ReLU(Recitified Linear Unit)`이다. 

## 2. 스트라이드(Strides)

- 필터가 이미지의 모든 픽셀을 천천히 하나씩 탐색한다면 매우 정밀할 수 있지만, 시간도 오래 걸리고 계산 비용도 크게 늘어남. 

- 스트라이드는 필터가 이미지 위에서 얼마나 넓은 보폭으로 이동할 것인지를 결정하는 중요한 매개변수 

- 스트라이드는 필터가 이미지를 얼마나 빨리 지나가는지, 그리고 그 결과로 생성되는 특징 맵의 크기를 얼마나 줄일 것인지를 조절 
- 예를 들어, 스트라이드 값이 1이면 필터는 한 픽셀씩 매우 세밀하게 이미지를 탐색
- 반면, 2일 경우 한번에 두 픽셀씩 건너뛰며, 더 적은 양의 특징 맵을 생성하고 계산을 더 빠르게 할 수 있다. 


- `스트라이드 값을 너무 크게 설정할 경우?` 필터가 중요한 정보를 놓치고 지나갈 위험이 생김.  
- `스트라이드 값이 너무 작으면?` 너무 자세히 분석하게 되어 처리 시간이 길어짐. 신경망의 효율성을 저하 시킬 수 있다.  

## Padding 

- 필터가 이미지를 지나갈 때마다 결과물인 `특징 맵(feature map)이 점점 작아지는 문제`가 발생 
- 이로 인해 중요한 정보가 손실 될 수 있고, 신경망의 깊이가 깊어질 수록 이 문제는 더욱 심화 됨. 
- `필터를 사용한 컨볼루션 과정에서, 입력 이미지의 크기를 유지하고 싶다면?` -> padding 


- 패딩은 입력 특징 맵의 가장자리에 추가적인 픽셀(대개 0으로채워진) 을 덧붙여 입력 특징 맵의 크기를 인위적으로 늘립니다.  
- 이렇게 함으로써 필터가 입력 특징 맵의 모든 영역을 골고루 스캔할 수 있게 되고, 원본 이미지의 중요한 가장자리 정보도 보존할 수 있다. 

- 패딩을 사용하면 이미지의 가장자리 정보가 손실되지 않게 되며, 컨볼루션 연산을 반복할 때마다 발생하는 특징 맵의 축소를 방지. 
- 이는 네트워크가 이미지의 중요한 시각적 정보를 보다 잘 보존하고, 학습에 더 효과적으로 사용할 수 잇게 해줌.
