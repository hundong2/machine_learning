### 1. 특징 추출기 (Feature Extractor)

- 신경망은 이미지를 살펴보고 강아지의 중요한 부분들을 살핌. 
- `필터`라는 도구를 사용 -> 이 필터는 사진 속 강아지의 눈이나, 코, 귀와 같은 특징을 걸러내는 일종의 패턴 인식기
- 이렇게 신경망은 사진을 구성하는 여러 요소들을 식별하여 강아지를 어떻게 구성하고 있는지를 이해 

- 특징 추출기는 
    - `합성곱층(Convolution Layers)`, `풀링층(Pooling Layers)`으로 구성 

#### 합성곱 레이어(Convolution Layer)

- 합성곱 신경망(CNN) 의 핵심은 합성곱 레이어로, 이미지의 중요한 특성을 추출해내는 기본 단위. 
- `합성곱`이란 말은 사실 이미지와 일종의 필터를 겹쳐서 `특징을 도출`해내는 수학적 연산 
- 이미지 위에서 필터를 움직여가며 특정 패턴을 찾아내는 작업

### 2. 분류기(Classifier) 단계

- 이전에 추출된 특징들을 종합하여 신경망이 `이것은 강아지다` 또는 `고양이다`와 같은 결정을 내림.  
- 이 단계는 신경망이 이미지 속 객체를 정확히 분류하는 과정 



- 필터는 이미지의 특정 부분 즉 수용 영역(Receptive Field)에 적용 되어, 해당 영역의 특징을 감시하고, 이를 기반으로 특징 맵(feature map)을 생성 
- 이 과정은 이미지에서 각 부분에서 중요한 정보를 추출하여 특징 맵에 반영하는 방식 

- 합성곱 신경망(CNN)에서의 `필터`는 이미지 분류와 특성 추출의 핵심적인 구성요소 
- 각 필터는 학습과정을 통해 이미지의 중요한 특성을 감지하도록 최적화된 `가중치의 집합`
- 이 필터들은 학습을 통해 업데이트되는 가중치로 구성, 학습 과정에서 반복적으로 조정 됨. 
- 네트워크의 출력이 목표에 더 가까워지도록 도와주는 학습 가능한 매개변수, 네트워크가 훈련 데이터로부터 중요한 특징을 추출하고이해하는 데 필수적 

- 필터의 역할은 분류작업에서 `중요한 특성을 강조`, 이는 이미지를 분석할 때 다양한 각도에서 접근 할 수 있게 해줍니다. 
- 어떤 필터는 이미지의 가장자리를 감지하는 반면, 어떤 필터는 질감이나 특정 형태를 감지하는데 사용
