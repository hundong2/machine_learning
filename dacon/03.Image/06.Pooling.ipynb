# Pooling Layer

- 풀링층은 신경망의 계산 부하를 줄이는 동시에 입력 데이터의 중요한 특징을 유지하도록 설계되어, 보다 깊고 복잡한 아키텍처에서 효과적으로 기능합니다. 

# feature extraction

입력 -> convolution + ReLU -> pooling -> Convolution + ReLU -> pooling -> Classification 

- CNN의 구조 내에 자리잡고 있는 풀링(Pooling)층은 신경망이 다루는 이미지의 크기를 줄이는 역할을 하면서도, 중요한 특징을 보존하는 기능을 합니다. 
- 풀링 레이어는 보통 여러 개의 합성곱+ReLU 레이어 뒤에 오며, 특징 추출기(Feature Extractor)의 한부분을 구성 
- `풀링 레이어는 학습 가능한 매개변수를 가지고 있지 않다는 점`
    - 즉, 풀링은 순전히 고정된 연산을 수행 학습 과정에서 업데이트 되지 않음.  
    - 이미지로부터 특징들을 감지해내면, 풀링 층은 이러한 특징들의 가장 강력한 신호만을 남기고 나머지는 버림으로써 데이터의 `차원을 축소`

## Pooling 의 종류

### 1. Max Pooling 

- Max Pooling 은 특징 맵의 특정 영역에서 최대값을 선택하여 다음 레이어로 전달.  
- 이미지 내 가장 두드러진 특징을 유지하고자 할 때 사용됨. 이미지의 작은 변화에 대해 강인한 특징 표현을 가능하게 한다.  

- 가장 높은 값을 추출하여 새로운 특징 맵을 형성, 이과정을 통해 가장 중요한 정보는 보존 된다. 예를 들어 가장자리와 같은 구조적 정보를 강조하는 데 매우 효과적이다. 
- 이로 인해 모델이 작은 이미지 변화에도 견딜 수 있으며, 실제 이미지 처리시 중요한 세부정보를 잃지 않게 도와준다. 

## 2. Min Pooling 

- 윈도우 내에서 최소값을 선택하여 특징 맵을 생성하는 방법 
- 주로 이미지의 어두운 부분을 강조하고 싶을 떄 사용 
- 이미지의 밝은 픽셀보다 어두운 픽셀이 더 중요한 정보를 담고 있을 경우, 최소 풀링을 사용하여 그 정보를 보존, 

### 3. Average pooling 

- 윈도우 내의 모든 값의 평균을 계산하여 특징맵을 생성 
- 텍스처와 같은 부드러운 이미지 특성을 다룰 때 유용, 전반적인 특성의 평균적인 정보를 효과적으로 전달 할 수 있다. 

## 풀링의 중요한 이점 중 하나는 `위치 불변성(Location Invariance)`

- 계산 부하를 줄여준다. 
    - 데이터의 크기가 작아지면 연산에 필요한 자원이 줄어들어 더 빠른 훈련과 효율적인 인퍼런스를 가능하게 한다. 이는 큰 이미지 데이터셋을 다룰 때 특히 중요한 이점이 된다. 
- 과적합을 방지
    - 각 특징의 정확한 위치보다는 존재 여부에 더 중점을 두기 때문에, 모델이 훈련 데이터의 노이즈에 덜 민감해 질 수 있다.
