```python
import numpy as np

# ==========================================
# 1. 1차원 배열(벡터)의 내적 예제
# ==========================================
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 요소끼리 곱하고 더합니다: (1*4) + (2*5) + (3*6) = 4 + 10 + 18 = 32
dot_1d = np.dot(a, b)

print("--- 1차원 배열 연산 ---")
print(f"a와 b의 내적 결과: {dot_1d} \n")

# [활용 꿀팁] 가중 평균 앙상블에서 활용하기
모델예측값들 = np.array([80, 90, 100]) # 3개 모델의 예측 점수
가중치 = np.array([0.2, 0.3, 0.5])      # 모델별 가중치 (합이 1)

# (80*0.2) + (90*0.3) + (100*0.5) 연산을 np.dot 한 번으로 깔끔하게 처리!
최종예측값 = np.dot(모델예측값들, 가중치) 
print(f"가중 평균 예측값: {최종예측값} \n")


# ==========================================
# 2. 2차원 배열(행렬)의 곱셈 예제
# ==========================================
A = np.array([[1, 2], 
              [3, 4]]) 
# A는 2x2 행렬

B = np.array([[5, 6], 
              [7, 8]]) 
# B는 2x2 행렬

# 행렬 곱셈 수행
# 결과의 (0,0) 자리: A의 첫 번째 '행' [1, 2] x B의 첫 번째 '열' [5, 7] = 1*5 + 2*7 = 19
# 결과의 (0,1) 자리: A의 첫 번째 '행' [1, 2] x B의 두 번째 '열' [6, 8] = 1*6 + 2*8 = 22
dot_2d = np.dot(A, B)

print("--- 2차원 배열 연산 (행렬 곱셈) ---")
print("A와 B의 행렬 곱셈 결과:")
print(dot_2d)
```
