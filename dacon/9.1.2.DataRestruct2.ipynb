# pivot_table

- 데이터를 요약하고 집계 
- 다차원 분석
    - index, column을 여러개 지정하여 다차원적으로 데이터를 분석할 수 있다. 
    - 결측값 처리 : 피벗테이블을 만들 때 결측치가 발생할 경우 결측값을 특정 값으로 대체 할 수 있다. 
        - `aggfunc` 인수 

```python
pivot_table = pd.pivot_table(
    data=train,
    index=['gender', 'income_type'],
    columns='edu_type',
    values='income_total',
    aggfunc='median', #aggfunc에는 min, max,mean, median, sum, count가 올 수 있다. 
)

pivot_table
```
