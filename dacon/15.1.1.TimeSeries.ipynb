# 정현파의 기본요소 이해 

- 주파수, 주기, 진폭, 샘플링 

$$ 주기(T) = \frac{1}{주파수(f)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플링 주파수 설정
sampling_frequency = 1000  # 1000Hz
sampling_interval = 1 / sampling_frequency  

# 시간 축 생성 
t = np.linspace(0, 5, sampling_frequency * 5)
# 0~5 초 까지 생성, 샘플링 주파수에 따라 적절한 간격으로 나누어짐. 
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
# start: 시작값, stop: 끝값, num: 생성할 샘플의 수(기본값은 50)
# endpoint = True일 경우 stop값이 포함( 기본값 True )
# retstep = True 로 설정하면 간격(step)도 반환( 기본값은 False )
# dtype = 반환할 배열의 데이터 타입 

# 신호 주파수와 사인파 생성
freq_of_signal = 0.5  
magnitude_of_signal = 2  
signal = magnitude_of_signal * np.sin(2 * np.pi * freq_of_signal * t ) 
# np.sin함수를 사용하여, 주파수 0.5 Hz, 진폭 2의 사인파 신호를 생성 
# 신호 주기 계산
cycle_of_signal = 1 / freq_of_signal  

# 샘플링 시간, 주기, 진폭 출력
print(f"Sampling Time (샘플링 간격)  : {sampling_interval} seconds")
print(f"Period of Signal (신호 주기) : {cycle_of_signal} seconds --- Frequency of Signal (신호 주파수): {freq_of_signal} Hz")
print(f"Magnitude of Signal (신호 진폭) : {magnitude_of_signal}")
```

```python
plt.figure(figsize=(12, 4))
plt.plot(t, signal, label='Sine Wave', color='blue')
plt.title('Sine Wave Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 4))
plt.plot(t, signal, label='Sine wave (사인파)', color='b')

# 주기를 표시하는 세로선
plt.axvline(x=0, color='red', linestyle='--', label='Period Start')  # 시작점
plt.axvline(x=cycle_of_signal, color='blue', linestyle='--', label='Period End')  # 1주기 끝점
plt.axvline(x=2 * cycle_of_signal, color='blue', linestyle='--')  # 2주기 끝점

# 주기를 나타내는 화살표 (시작점에서 끝점까지)
plt.annotate('', xy=(cycle_of_signal, 0), xytext=(0, 0),
             arrowprops=dict(arrowstyle='<->', lw=2, color='green'))
plt.annotate('', xy=(2 * cycle_of_signal, 0), xytext=(cycle_of_signal, 0),
             arrowprops=dict(arrowstyle='<->', lw=2, color='green'))

# 진폭에 대한 텍스트 및 화살표
plt.annotate('', xy=(0, magnitude_of_signal), xytext=(0, -magnitude_of_signal),
             arrowprops=dict(arrowstyle='<->', lw=2, color='orange'))
plt.text(0.02, 1.5, f"Magnitude = {magnitude_of_signal}", fontsize=12, color='orange', verticalalignment='center')

# 주기(T=1초) 텍스트로 표시
plt.text(cycle_of_signal / 2 + 0.25, 0.25 * magnitude_of_signal, f"주기(T) = {cycle_of_signal} second", fontsize=12, color='green', horizontalalignment='center')

# 그래프의 축 범위 제한 (0~2.5초만 표시하여 주기 강조)
plt.xlim(0, 2.5)

plt.ylim(-1.5 * magnitude_of_signal, 1.5 * magnitude_of_signal)

plt.title('Sine Wave with Period and Magnitude (사인파와 주기 및 진폭)')
plt.xlabel('Time (시간) [seconds]')
plt.ylabel('Amplitude (진폭)')
plt.legend()
plt.grid(True)
plt.show()
```

# 사인파 수식
사인파의 기본 수식은 다음과 같습니다.

## 예시 파일
15.1.1.TimeSeries.ipynb

## 답변
기본형:
$$
y = A \sin(2\pi f t + \phi)
$$

의미:
- $A$ : 진폭
- $f$ : 주파수
- $t$ : 시간
- $\phi$ : 위상

지금 노트북 코드 형태는 위상이 0인 경우라서:
$$
y = A \sin(2\pi f t)
$$

예:
- 진폭 $A=2$
- 주파수 $f=0.5$

이면
$$
y = 2\sin(2\pi \cdot 0.5 \cdot t)
$$

# 각속도(Angular Velocity)
각속도는 **단위 시간당 얼마나 빠르게 각도가 변하는지**를 나타내는 값입니다.

## 예시 파일
15.1.1.TimeSeries.ipynb

## 답변
수식으로 표현하면:

$$
\omega = \frac{\theta}{t} = 2\pi f
$$

- $\omega$ (omega) : 각속도 (단위: rad/s)
- $\theta$ : 각도 (라디안)
- $t$ : 시간 (초)
- $f$ : 주파수 (Hz)

**초등 수준 비유:**  
선풍기 날개가 1초에 3바퀴 돈다면  
→ 1초에 $3 \times 2\pi = 6\pi$ 라디안만큼 각도가 변했다  
→ 각속도 = $6\pi$ rad/s

**사인파와의 관계:**  
$2\pi ft$ 자체가 "시간 $t$동안 돌아간 총 각도"이고,  
$2\pi f$ 부분이 각속도 $\omega$입니다.

그래서 사인파 수식을 $\omega$로 다시 쓰면:

$$
y = A\sin(\omega t + \phi)
$$

즉, $\omega = 2\pi f$ 로 **각속도가 클수록 더 빠르게 진동하는 파형**이 됩니다.

# 정현파의 위상 (Phase)

- 위상 앞섬(Phase Lead)과 위상 지연(Phase Lag)



```python
import numpy as np
import matplotlib.pyplot as plt

# 주어진 조건
magnitude_of_wave = 2  # 진폭
freq_of_wave = 1  # 주파수 1Hz
sampling_frequency = 100  # 샘플링 주파수 100Hz
T = 1 / freq_of_wave  # 주기 T

# 샘플링 간격 계산
sampling_interval = 1 / sampling_frequency  # 샘플링 간격

# 시간 축 생성 (2초 동안의 신호)
t = np.arange(0, 2, sampling_interval)

# 원래 사인파 신호 생성 (Original Signal)
signal_original = magnitude_of_wave * np.sin(2 * np.pi * freq_of_wave * t)

# T/4 만큼 지연된 신호 생성 (Phase Lag, 오른쪽 이동)
signal_phase_lag = magnitude_of_wave * np.sin(2 * np.pi * freq_of_wave * (t - T / 4))

# T/2 만큼 당긴 신호 생성 (Phase Lead, 왼쪽 이동)
signal_phase_lead = magnitude_of_wave * np.sin(2 * np.pi * freq_of_wave * (t + T / 2 ))

plt.figure(figsize=(10, 4))
plt.plot(t, signal_original, label='Original Signal', color='b')
plt.plot(t, signal_phase_lag, 'r--', label='T/4 Phase Lag (Delayed)')
plt.plot(t, signal_phase_lead, 'g--', label='T/2 Phase Lead (Advanced)')
plt.title('Sine Wave with Phase Shifts: Phase Lag and Phase Lead (Peaks Highlighted)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 주어진 조건
magnitude_of_wave = 1  # 진폭
freq_of_wave = 1  # 주파수 1Hz (T = 1초)
sampling_frequency = 100  # 샘플링 주파수 100Hz
T = 1 / freq_of_wave  # 주기 T

# 샘플링 간격 계산
sampling_interval = 1 / sampling_frequency  # 샘플링 간격

# 시간 축 생성 (2초 동안의 신호)
t = np.arange(0, 2, sampling_interval)

# 원래 코사인파 신호 생성 (Original Cosine Signal)
signal_original = magnitude_of_wave * np.cos(2 * np.pi * freq_of_wave * t)

# 3/4 T만큼 지연된 신호 생성 (Phase Lag, 오른쪽 이동)
signal_phase_lag = magnitude_of_wave * np.cos(2 * np.pi * freq_of_wave * (t - 3 * T / 4))

plt.figure(figsize=(10, 4))

# 원래 신호 그리기
plt.plot(t, signal_original, label='Original Cosine Signal', color='blue')

# 3/4 T 만큼 지연된 신호 그리기
plt.plot(t, signal_phase_lag, label=r'Cosine Signal with $3/4T$ Phase Lag', linestyle='--', color='red')

# 중요한 점 강조 (극대/극소점 등)
plt.scatter([T/4, 5*T/4], [1, 1], color='green', zorder=5, label=r'$T/4$ Peaks (Phase Lag)')
plt.scatter([T, 2*T], [1, 1], color='black', zorder=5, label=r'$T$ Peaks (Original Signal)')

# 제목과 레이블 설정
plt.title(r'Cosine Wave with $3/4T$ Phase Lag Applied', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.axhline(0, color='black',linewidth=0.5)
plt.grid(True)

# 범례 설정
plt.legend(loc='upper right', fontsize=10)

# 그래프 표시
plt.tight_layout()
plt.show()
```

# 샘플링 주파수에 따른 신호 정보 손실과 나이퀴스트 주파수 

$ f_{Nyquist} = 2 \times f_{signal} $

```python
import numpy as np

# 주파수: 1Hz (1초에 한 번 진동)
freq_of_wave = 1

# 샘플링 주파수 리스트
sampling_frequencies = [50, 20, 2.6, 2, 1]
total_time = 5  # 총 시간 5초

# 신호와 시간 데이터를 저장할 리스트
signals = []
time_axes = []

# 각 샘플링 주파수에 대해 반복
for sampling_frequency in sampling_frequencies:
    # 샘플링 시간 계산
    sampling_time = 1 / sampling_frequency

    # 샘플링 주파수에 따른 시간 축 생성
    t = np.linspace(0, total_time, int(sampling_frequency * total_time))

    # 정현파 신호 생성
    signal = np.sin(2*np.pi*freq_of_wave*t)

    # 샘플링 시간 출력
    print(f'Sampling time: {sampling_time:.3f} sec ({sampling_frequency} samples)')

    # 시간 축과 신호를 리스트에 저장
    time_axes.append(t)
    signals.append(signal)
```

```python
import numpy as np
import pandas as pd

sampling_freq = 1000 
t = np.linspace(0, 1, sampling_freq)

# 각 주파수를 가지는 신호 생성
low_freq_signal = np.sin(2 * np.pi * 2 * t)  # 2Hz
mid_freq_signal = np.sin(2 * np.pi * 10 * t)  # 5Hz
high_freq_signal = np.sin(2 * np.pi * 20 * t)  # 20Hz

pure_signal = low_freq_signal + mid_freq_signal + high_freq_signal  

# 노이즈 추가
noise = 0.3 * np.random.randn(sampling_freq)
final_signal = pure_signal + noise

# 시간과 final signal을 데이터프레임으로 만들기
df_signal = pd.DataFrame({
    'time': t,
    'pure_signal' : pure_signal,
    'final_signal': final_signal,
})

# 데이터프레임의 처음 몇 행 출력
print(df_signal.head())
```

```python
import matplotlib.pyplot as plt

# 그래프 시각화
plt.figure(figsize=(10, 12))

for i, (t, signal) in enumerate(zip(time_axes, signals)):
    plt.subplot(len(sampling_frequencies), 1, i+1)
    plt.plot(t, signal, marker='o', linestyle='-', color='blue')
    plt.title(f'Sampling time: {1/sampling_frequencies[i]:.3f} sec ({len(t)} samples)')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

# 시계열 데이터의 기본 구성: 추세, 계절성, 노이즈 



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#데이터 포인트의 총 개수
num_samples = 100 

# 날짜 생성
date_range = pd.date_range(start='2020-01-01', periods=num_samples, freq='D')

# 시계열 데이터 : 추세 
trend = np.linspace(0, 10, num_samples)

# 주기 설정 (일단위) 
cyclic_period = 20   # 20일마다 한 주기가 완성됨  

# 주파수 계산 (주기의 역수) 
frequency = 1 / cyclic_period  # 1/20 cycles per day 

# 시계열 데이터 : 계절성
t = np.arange(num_samples)
seasonality = 5 * np.sin( 2*np.pi*frequency*trend) 

# 시계열 데이터 : 노이즈 
noise = np.random.normal(0, 1, num_samples)

data = trend + seasonality + noise

# 데이터프레임 생성
df = pd.DataFrame({'Date': date_range, 'Value': data})
df.head()
```

```python
# 시각화
fig, axs = plt.subplots(4, 1, figsize=(12, 8))

# 추세
axs[0].plot(df['Date'], trend)
axs[0].set_title('trend')
axs[0].set_xlabel('날짜')
axs[0].set_ylabel('값')

# 주기성
axs[1].plot(df['Date'], seasonality)
axs[1].set_title('seasonality')
axs[1].set_xlabel('날짜')
axs[1].set_ylabel('값')

# 노이즈
axs[2].plot(df['Date'], noise)
axs[2].set_title('noise')
axs[2].set_xlabel('날짜')
axs[2].set_ylabel('값')

# 전체
axs[3].plot(df['Date'], df['Value'])
axs[3].set_title('전체 시계열 데이터')
axs[3].set_xlabel('날짜')
axs[3].set_ylabel('값')

plt.tight_layout()
plt.show()
```
