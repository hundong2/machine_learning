# 정현파의 기본요소 이해 

- 주파수, 주기, 진폭, 샘플링 

$$ 주기(T) = \frac{1}{주파수(f)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플링 주파수 설정
sampling_frequency = 1000  # 1000Hz
sampling_interval = 1 / sampling_frequency  

# 시간 축 생성 
t = np.linspace(0, 5, sampling_frequency * 5)
# 0~5 초 까지 생성, 샘플링 주파수에 따라 적절한 간격으로 나누어짐. 
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
# start: 시작값, stop: 끝값, num: 생성할 샘플의 수(기본값은 50)
# endpoint = True일 경우 stop값이 포함( 기본값 True )
# retstep = True 로 설정하면 간격(step)도 반환( 기본값은 False )
# dtype = 반환할 배열의 데이터 타입 

# 신호 주파수와 사인파 생성
freq_of_signal = 0.5  
magnitude_of_signal = 2  
signal = magnitude_of_signal * np.sin(2 * np.pi * freq_of_signal * t ) 
# np.sin함수를 사용하여, 주파수 0.5 Hz, 진폭 2의 사인파 신호를 생성 
# 신호 주기 계산
cycle_of_signal = 1 / freq_of_signal  

# 샘플링 시간, 주기, 진폭 출력
print(f"Sampling Time (샘플링 간격)  : {sampling_interval} seconds")
print(f"Period of Signal (신호 주기) : {cycle_of_signal} seconds --- Frequency of Signal (신호 주파수): {freq_of_signal} Hz")
print(f"Magnitude of Signal (신호 진폭) : {magnitude_of_signal}")
```

```python
plt.figure(figsize=(12, 4))
plt.plot(t, signal, label='Sine Wave', color='blue')
plt.title('Sine Wave Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 4))
plt.plot(t, signal, label='Sine wave (사인파)', color='b')

# 주기를 표시하는 세로선
plt.axvline(x=0, color='red', linestyle='--', label='Period Start')  # 시작점
plt.axvline(x=cycle_of_signal, color='blue', linestyle='--', label='Period End')  # 1주기 끝점
plt.axvline(x=2 * cycle_of_signal, color='blue', linestyle='--')  # 2주기 끝점

# 주기를 나타내는 화살표 (시작점에서 끝점까지)
plt.annotate('', xy=(cycle_of_signal, 0), xytext=(0, 0),
             arrowprops=dict(arrowstyle='<->', lw=2, color='green'))
plt.annotate('', xy=(2 * cycle_of_signal, 0), xytext=(cycle_of_signal, 0),
             arrowprops=dict(arrowstyle='<->', lw=2, color='green'))

# 진폭에 대한 텍스트 및 화살표
plt.annotate('', xy=(0, magnitude_of_signal), xytext=(0, -magnitude_of_signal),
             arrowprops=dict(arrowstyle='<->', lw=2, color='orange'))
plt.text(0.02, 1.5, f"Magnitude = {magnitude_of_signal}", fontsize=12, color='orange', verticalalignment='center')

# 주기(T=1초) 텍스트로 표시
plt.text(cycle_of_signal / 2 + 0.25, 0.25 * magnitude_of_signal, f"주기(T) = {cycle_of_signal} second", fontsize=12, color='green', horizontalalignment='center')

# 그래프의 축 범위 제한 (0~2.5초만 표시하여 주기 강조)
plt.xlim(0, 2.5)

plt.ylim(-1.5 * magnitude_of_signal, 1.5 * magnitude_of_signal)

plt.title('Sine Wave with Period and Magnitude (사인파와 주기 및 진폭)')
plt.xlabel('Time (시간) [seconds]')
plt.ylabel('Amplitude (진폭)')
plt.legend()
plt.grid(True)
plt.show()
```

# 사인파 수식
사인파의 기본 수식은 다음과 같습니다.

## 예시 파일
15.1.1.TimeSeries.ipynb

## 답변
기본형:
$$
y = A \sin(2\pi f t + \phi)
$$

의미:
- $A$ : 진폭
- $f$ : 주파수
- $t$ : 시간
- $\phi$ : 위상

지금 노트북 코드 형태는 위상이 0인 경우라서:
$$
y = A \sin(2\pi f t)
$$

예:
- 진폭 $A=2$
- 주파수 $f=0.5$

이면
$$
y = 2\sin(2\pi \cdot 0.5 \cdot t)
$$

# 각속도(Angular Velocity)
각속도는 **단위 시간당 얼마나 빠르게 각도가 변하는지**를 나타내는 값입니다.

## 예시 파일
15.1.1.TimeSeries.ipynb

## 답변
수식으로 표현하면:

$$
\omega = \frac{\theta}{t} = 2\pi f
$$

- $\omega$ (omega) : 각속도 (단위: rad/s)
- $\theta$ : 각도 (라디안)
- $t$ : 시간 (초)
- $f$ : 주파수 (Hz)

**초등 수준 비유:**  
선풍기 날개가 1초에 3바퀴 돈다면  
→ 1초에 $3 \times 2\pi = 6\pi$ 라디안만큼 각도가 변했다  
→ 각속도 = $6\pi$ rad/s

**사인파와의 관계:**  
$2\pi ft$ 자체가 "시간 $t$동안 돌아간 총 각도"이고,  
$2\pi f$ 부분이 각속도 $\omega$입니다.

그래서 사인파 수식을 $\omega$로 다시 쓰면:

$$
y = A\sin(\omega t + \phi)
$$

즉, $\omega = 2\pi f$ 로 **각속도가 클수록 더 빠르게 진동하는 파형**이 됩니다.

```python
import numpy as np
import matplotlib.pyplot as plt

# given condition
magnitude_of_signal = 3  # 진폭
freq_of_signal = 2  # 주파수 2Hz
sampling_interval = 0.001 

####################################################

# 시간 축 생성 
t = np.arange(0, 5, sampling_interval)  # 0에서 5초까지 샘플링 간격으로 생성

# 사인파 신호 생성
signal = magnitude_of_signal * np.sin(2 * np.pi * freq_of_signal * t)


# 신호 시각화
plt.figure(figsize=(12, 4))
plt.plot(t, signal)
plt.title(f'Sine Wave: Amplitude = {magnitude_of_signal}, Frequency = {freq_of_signal}Hz')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# given condition 
magnitude_of_signal = 5  # 진폭
cycle_of_signal = 2 # sec

# 주파수 계산
freq_of_signal = 1 / cycle_of_signal  # 주기에서 주파수 계산
sampling_frequency = 500  # 샘플링 주파수

####################################################

# 샘플링 간격 계산
sampling_interval = 1 / sampling_frequency  # 샘플링 간격

# 0에서 5초까지 샘플링 간격으로 시간축 t 생성
t = np.arange(0, 5, sampling_interval)  

# 사인파 신호 생성
signal = magnitude_of_signal * np.cos(2 * np.pi * freq_of_signal * t)

# 신호 시각화
plt.figure(figsize=(12, 4))
plt.plot(t, signal)
plt.title(f'Cosine Wave: Amplitude = {magnitude_of_signal}, Period = {cycle_of_signal} seconds')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 시간 설정 (100일 간의 데이터를 생성)
days = np.arange(0,100,1)  # 0일부터 100일까지

#일별 데이터에서 7일 주기 (1주일마다 반복되는 패턴)
cycle_7d = 7
freq_7d = 1 / cycle_7d  # 주기에서 주파수 계산

# 농산물 가격의 주기적인 변화 + 약간의 노이즈 추가
amplitude = 10  # 가격 변동의 크기
price_variation = amplitude * np.sin(2 * np.pi * freq_7d * days ) + np.random.normal(0, 2, len(days))
# 평균 0, 표준 편차 2의 정규분포를 따르는 노이즈를 추가, 현실 세계의 불규칙성 반영 


# 데이터 시각화
plt.figure(figsize=(10, 4))
plt.plot(days, price_variation, label='Price Variation (주기적 농산물 가격)', color='b')
plt.title('Weekly Cyclic Agricultural Price Data (주간 주기 농산물 가격 데이터)')
plt.xlabel('Day (일)')
plt.ylabel('Price (가격)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


hours = np.arange(0, 23, 1)  # 0시부터 23시까지

# 시간별 데이터에서 24시간 주기 (하루 동안 반복되는 패턴)
cycle_24h = 24
freq_24h = 1 / cycle_24h  # 주기에서 주파수 계산

# 전력 소비 패턴 생성 + 노이즈 추가
amplitude = 50  # 전력 소비량의 크기
power_consumption = amplitude * (np.sin(2 * np.pi * freq_24h * hours) + 2) + np.random.normal(0, 5, len(hours))

# 데이터 시각화
plt.figure(figsize=(10, 4))
plt.plot(hours, power_consumption, label='Power Consumption (전력 소비량)', color='g')
plt.title('Hourly Power Consumption Data (시간별 전력 소비량)')
plt.xlabel('Hour (시간)')
plt.ylabel('Power Consumption (전력 소비량)')
plt.grid(True)
plt.legend()
plt.show()
```
