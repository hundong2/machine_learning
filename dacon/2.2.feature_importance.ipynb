# Feature Importance

- 피처 중요도(feature importance)란?
  - 머신러닝에서 각 피처(feature)가 모델의 예측에 얼마나 중요한지를 나타내는 척도 
- 피처 중요도 측정 방법 
  - 머신 러닝 모델 중에는 자체적으로 각 피처의 중요도를 계산 하는 것들이 있다. 
    - 트리 기반 모델
    - 선형 모델 

## 트리 기반 모델에서의 피처 중요도 

- 트리 기반 모델( 결정 트리(decision tree)와 랜덤 포레스트(random forest))는 각 피처가 얼마나 많은 정보를 제공하는지에 따라 중요도를 할당 
- 데이터를 분류하거나 예측할 때 각 피처가 얼마나 유용한지를 평가하여 중요도를 산출 

## 선형 모델에서의 피처 중요도 

- 선형회귀 ( Linear Regression), 로지스틱 회귀(Logistic Regression) 와 같은 모델들은 각 피처에 할당된 가중치 ( Coefficient )의 크기와 방향을 통해 피처의 중요도를 평가 
- 가중치의 절대값이 클수록 해당 피처는 모델의 예측에 더 큰 영향을 미침. 

## 통계 기반 피처 중요도 

- 피처와 타겟 변수 (예측하려는 값) 사이의 관계를 수치적으로 평가 
- `상관 계수(correlation coefficients` -> 가장 흔히 사용되는 방법 
  - 두 변수 사이의 관계의 강도화 방향을 -1에서 +1 사이의 값으로 나타냄. 

## 피처 중요도는 어떻게 활용?

- 모델 해석과 설명 
  - 모델의 결정 기준을 파악
  - 동작방식을 설명 

- 데이터 전처리와 피처 엔지니어링 
  - 중요도가 낮은 피처를 제거, 중요도가 높은 피처를 기반으로 새로운 피처 생성 
  - 모델의 복잡성을 줄이고, 성능을 향상

## 피처 중요도 출력 

```python
# 피처와 타겟 변수 분리
X = housing_df.drop('target', axis=1)
y = housing_df['target']

model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(X, y)

# 피처 중요도 추출
feature_importances_dt = model_dt.feature_importances_
display(feature_importances_dt)

# feature_importances와 X.columns을 이용하여 중요도가 높은 순서로 정렬하여 출력

df_feature_importances_dt = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances_dt})
sorted_df_feature_importances_dt = df_feature_importances_dt.sort_values(by='Importance', ascending=False).reset_index(drop=True)

display(sorted_df_feature_importances_dt)
```

## bar graph

```python
import matplotlib.pyplot as plt
# 피처 중요도를 수직 바 그래프로 시각화
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(X.columns, feature_importances_dt, color='skyblue')

ax.set_xlabel("Feature")
ax.set_ylabel("Feature Importance")
ax.set_title("Feature Importance in Decision Tree Model")
plt.tight_layout()  # 레이블이 잘리지 않도록 조정
plt.show()
```

# 선형모델 기반의 피처 중요도 측정

- [coef Algorithm](./2.2.1.LinearRegression_coef.md). 


## 선형 모델 기반의 피처 중요도 측정 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# 피처 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 스케일링된 데이터로 선형 회귀 모델 학습
model_lr_scaled = LinearRegression()
model_lr_scaled.fit(X_scaled, y)

# 피처의 가중치를 피처 중요도로 사용 (스케일링된 모델)
feature_importances_lr = abs(model_lr_scaled.coef_)

feature_importances_lr


```
