```python
import os 
import zipfile
import shutil

def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)

folders = ['./sign_train', './sign_test' ]
for folder in folders:
    create_folder(folder)
    print(f"folder '{folder}' created successfully.")
    print(os.listdir(folder))
```

```python
train_zip_file = './sign_train.zip'
test_zip_file = './sign_test.zip'

train_final_path = './sign_train'
test_final_path = './sign_test'

def check_zip_file(zip_file_path):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_content = zip_ref.namelist()
        return True

def extract_and_move_files(zip_file_path, final_path):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(final_path)
    print(f"{zip_file_path} extracted and moved to {final_path} successfully.")
```

```python
print(f"Checking files: {train_zip_file} and {test_zip_file}")
if os.path.exists(train_zip_file):
    if check_zip_file(train_zip_file):
        extract_and_move_files(train_zip_file, train_final_path)
if os.path.exists(test_zip_file):
    if check_zip_file(test_zip_file):
        extract_and_move_files(test_zip_file, test_final_path)

for foleder in folders:
    print(f"folder '{folder}' final contents")
    print(os.listdir(folder)[:20])
```

```python
from glob import glob
import os

def get_train_data(data_dir):
    img_path_list = []
    label_list = []

    img_path_list.extend(glob(os.path.join(data_dir, '*.jpg')))
    img_path_list.sort(key=lambda x: int(x.split('/')[-1].split('.')[0]))  

    label_list.extend(train_df['label'])
    return img_path_list, label_list

def get_test_data(data_dir):
    
    img_path_list = []

    img_path_list.extend(glob(os.path.join(data_dir, '*.jpg')))
    img_path_list.sort(key=lambda x: int(x.split('/')[-1].split('.')[0]))

    return img_path_list

all_img_path, all_label = get_train_data('sign_train')
test_img_path = get_test_data('sign_test')

print(all_img_path[:10])
print(all_label[:10])
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

if 'all_img_path' in locals() and 'all_label' in locals():
    print(f"Total number of images: {len(all_img_path)}")
    print(f"Total number of labels: {len(all_label)}")
    # 이미지를 랜덤으로 16개 선택
    sample_indices = np.random.choice(len(all_img_path), 16, replace=False)
    sampled_img_paths = [all_img_path[i] for i in sample_indices]
    sampled_labels = [all_label[i] for i in sample_indices]

    # 4행 4열의 그리드로 이미지 출력
    fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(5, 5))

    for i, (img_path, label) in enumerate(zip(sampled_img_paths, sampled_labels)):
        ax = axes[i // 4, i % 4]
        img = Image.open(img_path)
        ax.imshow(img)
        ax.set_title(label)
        ax.axis('off')

    plt.tight_layout()
    plt.show() 
```
