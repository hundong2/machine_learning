# 로그 변환 ( Log Transformation )

- 데이터의 범위를 조정하고 왜곡된 분포를 개선하기 위해 사용되는 기법 
- 데이터 값 사이에 큰 차이가 있거나, 한쪽으로 치우친 분포를 보일때 이 변환 방법을 사용 
- 데이터 크기의 차이를 줄여서, 모든 데이터 포인트가 분석에 거의 동등한 영향을 미치도록 도와줌 

## 장점

- 비선형 관계를 선형 관계로 변환하는 효과가 있음 
- 극단 값의 영향을 줄이는데 효과적 

## 주의 사항

- 로그 변환을 할 때 주의할 점 중 하나는 데이터 세트에 0 또는 음수 값이 포함되어 있을 경우 로그 변환을 직접 적용할 수 없다. 
- $ log(0) = \infty $
- $ log(-1) $ 는 실수 범위에서는 정의 되지 않는다. 
- 이러한 값들을 사용하기 위해서는 반드시 전처리가 필요로 한다. 

````python
import numpy as np

data = [1, 10, 100, 1000, 10000]

log_transformed = np.log(data)

print(log_transformed)
````

## 정규 분포 ( 가우스 분포 ) 

- 데이터의 분포가 종 모양의 대칭 곡선을 그리는 형태 
- 데이터의 대부분이 평균값(또는 중앙값) 근처에 집중 
- 평균으로 부터 멀어질 수록 데이터의 빈도가 감소 

## 왜도 

- 각 특성의 왜도를 계산하는 과정은 데이터의 비대칭도를 평가하는 중요한 단계 
- 왜도는 데이터가 얼마나 비대칭적으로 분포하는지를 나타냄 
- 이상적인 정규 분포의 경우 왜도 값이 0에 가까움    
- `select_dtypes() 메서드로 수치형 데이터를 필터링` -> `skew()`로 각열의 왜도를 계산 

--- 

```python
age_skewness = train.select_dtypes('number').skew()
print(age_skewness)
```
