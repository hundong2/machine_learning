| `torch.int8` | i8 | 8 | -128 ~ 127 | 부호있는 8비트 정수 |

| `torch.uint8` | u8 | 8 | 0 ~ 255 | 부호없는 8비트 정수 |

| 타입 | 문자표기 | 비트 | 정밀도 | 설명 |

| `torch.float16` | f16 | 16 | ~3.3자리 | 반정밀도 (GPU 연산 가속) |
| `torch.bfloat16` or `torch.half` | bf16 | 16 | ~3.3자리 | Brain 반정밀도 |
| `torch.float32` or `torch.float` | f32 | 32 | ~7자리 | 단정밀도 (기본값) |
| `torch.float64` or `torch.double` | f64 | 64 | ~15자리 | 배정밀도 (double) |

| 타입 | 문자표기 | 메모리 | 설명 |

| `torch.complex128` or `torch.cdouble` | c128 | 128 | 실부/허부 float64 |

```python
#tensor의 타입 변경 
# 초기 텐서 생성 (정수 타입)
tensor = torch.tensor([1, 2, 3, 4])

print("1. 텐서의 타입 변경 function\n")
# 텐서 속성 변환
float_tensor = tensor.float() # int -> float
int_tensor = float_tensor.int() # float -> int
double_tensor = float_tensor.double() # float -> double
bool_tensor = int_tensor.bool() # int -> bool

print("처음 생성한 텐서:", tensor.dtype)
print('float_tensor의 타입 확인: ', float_tensor.dtype)
print('int_tensor의 타입 확인: ', int_tensor.dtype)
print('double_tensor의 타입 확인: ', double_tensor.dtype)
print('bool_tensor의 타입 확인: ', bool_tensor.dtype)

print("\n2. 텐서의 타입 변경 to method\n")
#텐서의 타입 변경 to method 
# 텐서 속성 변환
float_tensor2 = tensor.to(dtype=torch.float32)
int_tensor2 = float_tensor.to(dtype=torch.int32)
double_tensor2 = float_tensor.to(dtype=torch.float64)
bool_tensor2 = int_tensor.to(dtype=torch.bool)

print("처음 생성한 텐서:", tensor.dtype)
print('float_tensor의 타입 확인: ', float_tensor2.dtype)
print('int_tensor의 타입 확인: ', int_tensor2.dtype)
print('double_tensor의 타입 확인: ', double_tensor2.dtype)
print('bool_tensor의 타입 확인: ', bool_tensor2.dtype)
```

# 3. 차원(Dimension)과 형태(Shape)

- 0차원 텐서: 스칼라(단일 숫자)
- 1차원 텐서: 벡터(숫자의 배열)
- 2차원 텐서: 행렬(숫자의 2차원 배열)
- 3차원 텐서: 다차원 배열을 의미

```python
data = 5

scalar = torch.tensor(data)

print("스칼라: ", scalar)
print("스칼라 차원: ", scalar.dim())
print("스칼라 형태: ", scalar.shape)

data = [1, 2, 3]

vector = torch.tensor(data)

print("벡터: ", vector)
print("벡터 차원:", vector.dim())
print("벡터 형태:", vector.shape)

data = [[1, 2, 3],
        [4, 5, 6]]

matrix = torch.tensor(data)

print("행렬: ", matrix)
print("행렬 차원: ", matrix.dim())
print("행렬 형태: ", matrix.shape)

data = [
    [[1, 2, 3],
     [3, 4, 5]],
    [[5, 6, 7],
     [7, 8, 8]]
]
tensor3D = torch.tensor(data)

print("3차원 텐서", tensor3D)
print("3차원 텐서 차원:", tensor3D.dim())
print("3차원 텐서 형태:", tensor3D.shape)
```

# tensor의 형태 변환 


```python
import torch

# 초기 텐서 생성
x = torch.arange(16)  # 0에서 15까지의 숫자를 포함하는 텐서
print("텐서:", x)
print("텐서의 형태:", x.shape)

# reshape 함수 사용 예제
reshaped_x = x.reshape(4, 4)  # 4x4 텐서로 변경
print("\n변형 후 텐서:", reshaped_x)
print("변형 후 텐서의 형태:", reshaped_x.shape)
```

```python
#원본 텐서의 데이터를 공유하는 새로운 형태의 텐서를 반환 
#데이터의 실제 메모리 레이아웃을 변경하지 않으며, 단지 텐서에 대한 새로운 인덱싱 뷰를 제공
#만약 원본 텐서가 메모리상에서 연속적이지 않는경우 view()를 사용할 수 없습니다. 이 경우에는 reshape()를 사용해야 합니다.
viewed_x = x.view(4, 2, 2)  # 4x2x2 텐서로 변경
print("변형 후 텐서:", viewed_x)
print("변형 후 텐서의 형태:", viewed_x.shape)
```

```python
#차원 축소 
# 크기가 1인 차원을 포함하는 텐서 생성
tensor = torch.tensor([[[1, 2, 3, 4]]])  # 크기: (1, 1, 4)
print("원본 텐서:", tensor)
print("원본 텐서 형태:", tensor.shape)

# 모든 크기가 1인 차원을 제거
squeezed_tensor = tensor.squeeze()
print("\n변형된 텐서:", squeezed_tensor)
print("변형된 텐서 형태:", squeezed_tensor.shape)

# 특정 차원을 지정하여 squeeze
squeezed_dim_tensor = tensor.squeeze(0)  # 첫 번째 차원을 지정
print("\n변형된 텐서:", squeezed_dim_tensor)
print("변형된 텐서 형태:", squeezed_dim_tensor.shape)
```

```python
# 차원 확장 ununsqueeze() 메서드를 사용하여 크기가 1인 차원을 추가할 수 있습니다.
# 예를 들어, 1차원 텐서에 대해 차원을 확장하려면 unsqueeze()를 사용할 수 있습니다.
# 원본 텐서 생성
tensor = torch.tensor([1, 2, 3, 4])  # 크기: (4,)
print("원본 텐서:", tensor)
print("원본 텐서 형태:", tensor.shape)

# 첫 번째 차원을 추가하여 차원 확장
unsqueeze_tensor_first_dim = tensor.unsqueeze(0)
print("\n변형된 텐서:", unsqueeze_tensor_first_dim)
print("변형된 텐서 형태:", unsqueeze_tensor_first_dim.shape)

# 마지막 차원을 추가하여 차원 확장
unsqueeze_tensor_last_dim = tensor.unsqueeze(-1)
print("\n변형된 텐서:", unsqueeze_tensor_last_dim)
print("변형된 텐서 형태:", unsqueeze_tensor_last_dim.shape)
```

# 텐서 결합 stack

- cat과는 다르게, 모든 텐서에 대해 새로운 차원이 생기므로, 텐서들은 그 크기가 동일해야 한다. 
- stack은 데이터를 구조적으로 정리하고 싶을 때 유용 

```python
# 2x2 행렬 생성
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[5, 6], [7, 8]])
tensor3 = torch.tensor([[9, 10], [11, 12]])

# stack 함수 사용 예제
stack_tensors = torch.stack((tensor1, tensor2, tensor3), dim=0)  # 새 차원을 0번 차원으로 삽입하여 쌓기

print("stack_tensors\n", stack_tensors)
print('stack_tensors 텐서 형태: ', stack_tensors.shape)
```

```python
#텐서 분할 
# 10개의 요소를 가진 1차원 텐서 생성
tensor = torch.arange(10)  # 0부터 9까지의 숫자
print("원본 텐서:", tensor)
print("텐서 형태:", tensor.shape)

# chunk 함수 사용 예제
chunks = torch.chunk(tensor, 3, dim=0)  # 3개의 조각으로 분할
print("\n Chunks")
for i, chunk in enumerate(chunks):
    print(f"Chunk {i}: {chunk}, {chunk.shape}")
```

```python
# [10, 20] 크기의 텐서 생성
tensor = torch.randn(10, 20) 

# 첫 번째 차원(길이가 10)을 기준으로, 4개와 6개 요소를 포함하는 두 조각으로 나눔
split_sizes = (4, 6)  # 조각의 크기를 지정
splits = tensor.split(split_sizes, dim=0)  # 0번째 차원(행)을 기준으로 나눔

# 나누어진 조각들 확인
print(f"첫 번째 조각의 크기: {splits[0].size()}")
print(f"두 번째 조각의 크기: {splits[1].size()}")
```

# 4. 고급 인덱싱 기법 



```python
#gather() 메서드를 사용하여 특정 인덱스의 값을 선택할 수 있습니다. 예를 들어, 2차원 텐서에서 특정 행과 열의 값을 선택하려면 gather()를 사용할 수 있습니다.
# 예제: 2차원 텐서에서 특정 인덱스의 값을 선택
tensor = torch.tensor([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
indices = torch.tensor([[0, 2], [1, 0], [2, 1]])  # 선택할 인덱스
gathered_elements = tensor.gather(1, indices)  #
print("선택된 요소:\n", gathered_elements)
```

```python
# gather 예제
y = torch.tensor([[1, 2], [3, 4]])
gather_index = torch.tensor([[0, 0], [1, 0]])
gathered = y.gather(1, gather_index)  # 지정된 인덱스에서 값을 수집
print("추출한 텐서:\n", gathered)
```

PyTorch의 gather는 인덱스 텐서가 가리키는 위치의 값을 특정 축(dim)에서 뽑아오는 함수입니다.

핵심 한 줄
- y = x.gather(dim, index)
- dim 축을 따라 index가 지정한 위치의 값을 가져오고, 결과 shape은 index와 동일합니다.

1) 꼭 알아야 할 규칙
- x와 index는 차원 수(ndim)가 같아야 합니다.
- index의 dtype은 정수형(Long, int64)이어야 합니다.
- index의 각 값은 dim 축 범위 안이어야 합니다.
- output.shape = index.shape
- dim 이외의 축에서는 index와 x가 브로드캐스트 가능한 형태여야 합니다.

2) 직관적으로 이해하기
- dim=1이면 각 행(row)에서 열(column) 인덱스를 골라 뽑습니다.
- dim=0이면 각 열(column)에서 행(row) 인덱스를 골라 뽑습니다.

3) 기본 예제 (dim=1, 행별 열 선택)
- x
  [[10, 20, 30],
   [40, 50, 60]]

- index
  [[2, 0],
   [1, 1]]

- y = x.gather(1, index)

- y
  [[30, 10],
   [50, 50]]

설명
- 첫 행: 열 2, 0 선택 → 30, 10
- 둘째 행: 열 1, 1 선택 → 50, 50

4) dim=0 예제 (열별 행 선택)
- x
  [[10, 20, 30],
   [40, 50, 60],
   [70, 80, 90]]

- index
  [[2, 1, 0],
   [0, 2, 1]]

- y = x.gather(0, index)

- y
  [[70, 50, 30],
   [10, 80, 60]]

설명
- dim=0이므로 각 위치에서 행 번호를 index가 지정함

5) 실무에서 자주 쓰는 패턴
- 분류 모델에서 정답 클래스 점수 뽑기
  - logits shape: [batch, num_classes]
  - targets shape: [batch]
  - targets를 [batch, 1]로 바꾼 뒤 gather(1, targets_unsqueezed)
  - 각 샘플의 정답 클래스 logit만 추출 가능

6) 자주 나는 에러와 해결
- 에러: index dtype이 float
  - 해결: index.long() 사용
- 에러: index out of bounds
  - 해결: index 값이 0 이상, 해당 dim 크기 미만인지 확인
- 에러: 차원 수 불일치
  - 해결: unsqueeze로 차원 맞추기
- 에러: 예상과 다른 shape
  - 해결: 결과 shape은 index shape이라는 점 확인

7) gather vs index_select 차이
- gather
  - 위치마다 다른 인덱스를 줄 수 있음(세밀함)
- index_select
  - 한 축에서 공통 인덱스만 선택(간단함)

## view: 
- 메모리를 복사하지 않고 기존 텐서를 다른 모양으로 “바라보는” 방식입니다. 
- 그래서 원본 텐서가 연속(contiguous) 메모리여야 합니다. 아니면 에러가 납니다.
## reshape
- 가능하면 view처럼 동작하지만, 
- 필요하면 내부적으로 복사해서라도 원하는 모양을 만들어줍니다. 그래서 더 유연합니다.

```python
# (1, 5, 1) 형태의 랜덤 텐서 B 생성
tensor_B = torch.rand((1, 5, 1))

#(5,)로 축소
squeezed_tensor = tensor_B.squeeze()
print("squeezed_tensor shape:", squeezed_tensor.shape)

# (1, 5) 형태로 확장
unsqueezed_tensor = squeezed_tensor.unsqueeze(0)
print("unsqueezed_tensor shape:", unsqueezed_tensor.shape)
```

```python
# (3, 4) 형태의 랜덤 텐서 D1과 D2 생성
tensor_D1 = torch.rand((3, 4))
tensor_D2 = torch.rand((3, 4))

# 차원 0을 기준으로 결합
cat_tensor = torch.cat((tensor_D1, tensor_D2), dim=0)
print("cat_tensor shape:", cat_tensor.shape)

# 새로운 차원을 추가하여 결합
stack_tensor = torch.stack((tensor_D1, tensor_D2), dim=0)
print("stack_tensor shape:", stack_tensor.shape)
```
