# Activate Function

$ x1w1 + x2w2 + b = a $ -> $ h(a) = y $

```python
import torch
import torch.nn as nn

# 선형 층 정의 (입력 특성 10개, 출력 특성 1개)
linear_layer = nn.Linear(in_features=10, out_features=1)
```

```python
# 임의의 입력 텐서 생성 (배치 크기 1, 입력 특성 10개)
input_tensor = torch.randn(1, 10)
print("Input Tensor:\n", input_tensor)

# 선형 층을 통과
linear_output = linear_layer(input_tensor)
print("\nLinear Layer Output:\n", linear_output)

# 시그모이드 활성화 함수 적용
sigmoid_output = torch.sigmoid(linear_output)
print("\nSigmoid Output:\n", sigmoid_output)
```

## 탄젠트 함수 적용

- 탄젠트 함수는 출력 값을 -1과 1사이로 압축. 이는 데이터를 중심으로 정규화할때 유용 

```python
# 선형 층 정의 (입력 특성 10개, 출력 특성 1개)
linear_layer = nn.Linear(in_features=10, out_features=1)

# 임의의 입력 텐서 생성 (배치 크기 1, 입력 특성 10개)
input_tensor = torch.randn(1, 10)
print("Input Tensor:\n", input_tensor)

# 선형 층을 통과
linear_output = linear_layer(input_tensor)
print("\nLinear Layer Output:\n", linear_output)

# 하이퍼볼릭 탄젠트 활성화 함수 적용
tanh_output = torch.tanh(linear_output)
print("\nTanh Output:\n", tanh_output)
```

## ReLU(Recitified Linear Unit )

- ReLU 함수는 음수 입력을 0으로 바꾼다. 모델의 학습 속도를 개선하고, 그래디언트 소실 문제를 줄이는데 도움을 줌 

```python
# 선형 층 정의 (입력 특성 10개, 출력 특성 1개)
linear_layer = nn.Linear(in_features=10, out_features=1)

# ReLU 활성화 레이어 정의
relu_layer = nn.ReLU()

# 임의의 입력 텐서 생성 (배치 크기 1, 입력 특성 10개)
input_tensor = torch.randn(1, 10)
print("Input Tensor:\n", input_tensor)

# 선형 층을 통과
linear_output = linear_layer(input_tensor)
print("\nLinear Layer Output:\n", linear_output)

# ReLU 레이어를 통과
relu_output = relu_layer(linear_output)
print("\nReLU Layer Output:\n", relu_output)
```

## LeakyReLU

- ReLU와 유사하지만, 음수 입력에 대해 매우 작은 양의 기울기를 허용하여 뉴런이 죽는 문제를 완화 

```python
# 선형 층 정의 (입력 특성 10개, 출력 특성 1개)
linear_layer = nn.Linear(in_features=10, out_features=1)

# LeakyReLU 활성화 레이어 정의 (기울기 0.01)
leaky_relu = nn.LeakyReLU(negative_slope=0.01) # 음수 부분의 기울기를 0.01로 설정 

# 임의의 입력 텐서 생성 (배치 크기 1, 입력 특성 10개)
input_tensor = torch.randn(1, 10)
print("input_tensor:\n", input_tensor)

# 선형 층을 통과
linear_output = linear_layer(input_tensor)
print("\nlinear_output:\n", linear_output)

# LeakyReLU 레이어를 통과
leaky_relu_output = leaky_relu(linear_output)
print("\nLeaky ReLU Output:\n", leaky_relu_output)
```

# softmax

- softmax함수는 다중 클래스 분류 문제에서 출력 값을 확률 분포로 변환. 주로 `출력레이어`에서 사용

```python
# 선형 층 정의 (입력 특성 10개, 출력 특성 1개)
linear_layer = nn.Linear(in_features=10, out_features=1)

# Softmax 활성화 레이어 정의 (dim 파라미터는 softmax를 적용할 차원)
softmax = nn.Softmax(dim=1)

# 임의의 입력 텐서 생성 (배치 크기 1, 입력 특성 10개)
input_tensor = torch.randn(1, 10)
print("input_tensor:\n", input_tensor)

# 선형 층을 통과
linear_output = linear_layer(input_tensor)
print("\nlinear_output:\n", linear_output)

# Softmax 레이어를 통과
softmax_output = softmax(linear_output)
print("\nSoftmax Output:\n", softmax_output)
```

# LogSoftmax

- 로그소프트맥스 함수는 다중 클래스 분류 문제에서 softmax 함수의 로그 값을 계산. 
- 이 함수는 주로 학습 과정에서 크로스 엔트로피 손실과 함께 사용되어 계산을 단순화 하고 수치적 안정성을 높인다 `출력레이어` 또는, `손실 계산 전에` 사용됨. 

```python
# 선형 층 정의 (입력 특성 10개, 출력 특성 1개)
linear_layer = nn.Linear(in_features=10, out_features=1)

# Softmax 활성화 레이어 정의 (dim 파라미터는 softmax를 적용할 차원)
softmax = nn.LogSoftmax(dim=1)

# 임의의 입력 텐서 생성 (배치 크기 1, 입력 특성 10개)
input_tensor = torch.randn(1, 10)
print("input_tensor:\n", input_tensor)

# 선형 층을 통과
linear_output = linear_layer(input_tensor)
print("\nlinear_output:\n", linear_output)

# Softmax 레이어를 통과
softmax_output = softmax(linear_output)
print("\nSoftmax Output:\n", softmax_output)
```

## Dropout Layers

- 드랍아웃이란 학습 과정에서 신경망의 뉴런(노드)를 무작위로 선택하여 활성화하지 않음으로써, 즉 해당 뉴런의 출력을 0으로 만듦으로써 작동 
- 마치 신경망 내에서 여러 가지 다른 구조의 신경망이 학습되는 효과를 낳아, 모델이 특정 뉴런이나 뉴런의 조합에 과도하게 의존하는 것을 방지 

- Overfitting을 방지하는 효과적인 방법중 하나. 
- `과적합`이란 모델이 학습데이터에 너무 잘 맞춰져서 새로운 데이터나 테스트 데이터에 대한 성능이 떨어지는 현상. 

```python
# 드랍아웃 객체 생성, 드랍아웃 비율을 50%로 설정
dropout = nn.Dropout(p=0.5)

# 입력 데이터 텐서 생성
input_data = torch.randn(1, 10)  # 1x10 크기의 랜덤 텐서

# 드랍아웃 적용
output = dropout(input_data)

print("Input data:\n", input_data)
print("\nAfter applying dropout:\n", output)
```

# Model 만들기 

- sequencial container 

- 여러 모듈을 순서대로 쌓아 올릴 수 있게 해주며, 데이터가 정의된 순서대로 각 모듈을 통과 

```python
# Sequential 모델 정의
model = nn.Sequential(
    ### 여기에 모델 상세 내용을 작성하세요.
    nn.Linear(784, 128), # 입력층 -> 첫번째 은닉층
    nn.ReLU(), #첫 번째 은닉층의 활성화 함수
    nn.Dropout(p=0.2),#20% 확률로 드랍아웃
    nn.Linear(128, 64), # 128개의 뉴런, 64개의 뉴런으로 전달
    nn.ReLU(),
    nn.Linear(64, 10), #0~9사이의 숫자를 분류, 분류 문제이므로, 활성화 함수로는 LogSoftmax를 사용
    nn.LogSoftmax(dim=1)
)

# 모델 구조 출력
print(model)
```

```python
class CustomModel(nn.Module):
    def __init__(self):
        super(CustomModel, self).__init__()
        # 첫 번째 은닉층: 784개의 입력 특성을 받아 128개의 뉴런으로 전달
        self.fc1 = nn.Linear(784, 128)
        self.relu1 = nn.ReLU()  # 첫 번째 은닉층의 활성화 함수로 ReLU 정의

        # 두 번째 은닉층: 128개의 뉴런을 받아 64개의 뉴런으로 전달
        self.fc2 = nn.Linear(128, 64)
        self.relu2 = nn.ReLU()  # 두 번째 은닉층의 활성화 함수로 ReLU 정의

        # 출력층: 64개의 뉴런을 받아 10개의 출력을 생성
        self.fc3 = nn.Linear(64, 10)

        # 드랍아웃 적용: 확률은 20%
        self.dropout = nn.Dropout(p=0.2)

    def forward(self, x):
        # 첫 번째 은닉층을 통과한 후 ReLU 활성화 함수 적용
        x = relu1(fc1(x))
        
        # 드랍아웃 적용
        x = dropout(x)
        
        # 두 번째 은닉층을 통과한 후 ReLU 활성화 함수 적용
        x = self.relu2(self.fc2(x))
        
        # 출력층을 통과
        x = self.fc3(x)
        
        # 최종 출력을 위해 Softmax 적용
        return nn.Softmax(dim=1)(x)

model = CustomModel()
model
```
