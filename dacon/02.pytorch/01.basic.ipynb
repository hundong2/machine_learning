# 동적 계산 그래프(Dynamic Computation Graph)



````python
# 자동 미분
x = torch.tensor(1., requires_grad=True)
y = x * 2
y.backward()   # 미분 계산
print(x.grad)  # x에 대한 y의 미분 값 확인
````

## 4. pytorch 와 tensorflow비교



````python
import torch

# 텐서 생성
a = torch.tensor([2., 3.])
b = torch.tensor([6., 4.])

# 텐서 덧셈 연산
c = torch.add(a, b)

print(c)
````

````python
import tensorflow as tf

# 텐서 생성
a = tf.constant([2., 3.])
b = tf.constant([6., 4.])

# 텐서 덧셈 연산
c = tf.add(a, b)

print(c)
````
