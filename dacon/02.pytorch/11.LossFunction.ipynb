# 손실 함수 

## 평균 제곱 오차 손실 ( MSE, Mean squared Error Loss )

- 회귀 문제에 주로 사용, 예측값과 실제값의 차이를 제곱하여 평균 

## 교차 엔트로피 손실 ( Cross-Entropy Loss )

- 분류 문제에 주로 사용, 모ㅔㄹ이 실제 클래스를 정확히 예측하는데 얼마나 잘하는지를 측정
- torch.nn.CrossEntropyLoss()

## 이진 교차 엔트로피 손실 ( Binary Cross-Entropy Loss )

- 이진 분류 문제에 사용, 각 클래스에 대한 예측 확률을 사용하여 손실을 계산
- torch.nn.BCELoss

| 구분 | 이름 | 설명 | PyTorch 코드 |
| --- | --- | --- | --- |
| 분류 | 교차 엔트로피 손실 | 모델이 실제 클래스를 정확히 예측하는 데 얼마나 잘 하는지를 측정 | `torch.nn.CrossEntropyLoss()` |
| 분류 | 코사인 유사도 손실 | 두 벡터 간의 코사인 유사도를 사용하여 손실을 계산 | `torch.nn.CosineEmbeddingLoss()` |
| 분류/회귀 | 허브(Huber) 손실 | MSE와 MAE의 조합으로, 오차가 작을 때는 MSE, 클 때는 MAE로 작동 | `torch.nn.HuberLoss()` |
| 분류/회귀 | 마진 순위 손실 | 두 입력 샘플 간의 상대적 거리를 학습하는 데 사용 | `torch.nn.MarginRankingLoss()` |
| 다중 레이블 분류 | 다중 레이블 소프트 마진 손실 | 다중 레이블 분류 문제의 손실을 계산 | `torch.nn.MultiLabelSoftMarginLoss()` |
| 이진 분류 | 이진 교차 엔트로피 손실 | 각 클래스에 대한 예측 확률을 사용하여 손실을 계산 | `torch.nn.BCELoss()` |
| 이진 분류 | 이진 교차 엔트로피 손실 with 로짓 | 로짓(모델 출력)을 사용하여 이진 분류의 손실을 계산 | `torch.nn.BCEWithLogitsLoss()` |
| 회귀 | 평균 제곱 오차 손실 (MSE) | 예측값과 실제값의 차이를 제곱하여 평균낸 값 | `torch.nn.MSELoss()` |
| 회귀 | 평균 절대 오차 손실 (MAE) | 예측값과 실제값의 차이의 절대값을 평균낸 값 | `torch.nn.L1Loss()` |

```python
import torch.nn as nn

criterion = nn.CrossEntropyLoss()
print(criterion)
```

```python
criterion = nn.L1Loss()
print(criterion)
```

```python
criterion = nn.MSELoss()
print(criterion)
```
