## 데이터 정규표현식

- `.`: 어떤 한 개의 문자와 일치. 단, 행을 바꾸는 개행 문자(`\n`)은 제외   
- `^` : 문자열의 시작  
- `$` : 문자열의 끝  
- `|`: 두 패턴 중 하나와 일치. OR 연산 ,shift 와 + 를 입력하는 것으로 사용할 수 있음
- `\`: 특수 문자를 일반 문자로 사용, 특수 시퀀스를 나타내는데 사용됨.

### 특수 시퀀스 

- `\d` : 모든 숫자와 일치 ( 0,1,2, ..., 9)
- `\D` : 숫자가 아닌 모든 문자와 일치합니다. 
- `\s` : 모든 공백 문자와 일치 
- `\S` : 공백 문자가 아닌 모든 문자와 일치 
- `\w` : 문자, 숫자, 밑줄 문자와 일치 
- `\W` : `\w`에 해당하지 않는 모든 문자와 일치  

### 반복자 

- `*`: 바로앞의 문자가 0회 이상 반복될 때 일치. 즉, 문자가 없거나 여러번 있을 수 있음
- `+`: 바로앞의 문자가 1회 이상 반복될 때 일치. 적어도 한번은 해당 문자가 있어야 한다는 것을 의미 
- `?`: 바로 앞의 문자가 0회 또는 1회 있을 때 일치. 이는 문자가 있거나 없을 수 있다는 것을 의미 
- `{n}`: 바로 앞의 문자가 정확히 n회 반복 
- `{n,}`: 바로 앞의 문자가 n회 이상 반복 될 때 일치 
- `{n,m}`: 바로 앞의 문자가 최소 n회, 최대 m회 반복 될때 일치 

### example 

- `\d{2,4}`: 숫자이면서 최소 2자리 최대 4자리의 연속된 숫자
- `\S+`: 문자열 hello world가있을 때 공백이 아닌 문자가 1회 이상 연속되는 부분 hello와 world를 각각 별도로 찾아내어일치 시킴
