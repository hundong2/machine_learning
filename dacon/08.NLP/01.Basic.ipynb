# Natural Language Processing, NLP

- 인간읜 언어를 컴퓨터가 이해하고 처리할 수 있도록 하는 놀라운 기술 분야 

## 자연어 처리의 활용 분야 

- 자연어 이해(Natural Language Understanding, NLU ): 텍스트의 의미를 파악하고 해석하는 과정
- 자연어 생성(Natural Language Generation, NLG): 데이터나 정보를 자연스러운 언어 텍스트로 변환하는 기술 
- 음성 인식(Speech Recognition): 음성 데이터를 텍스트로 변환하는 과정
- 기계 번역(Machine Translation): 하나의 언어로 작성된 텍스트를 다른 언어로 번역하는 기술
- 감정 분석(Sentiment Analysis): 텍스트에서 감정이나 태도를 파악하는 과정 

## 2.1 자연어 이해(Natural Language Understanding)

- 의도 인식(Intent Detection)
    - 일반적으로 컴퓨터는 미리 설정되어 있는 명확한 명령만을 인식할 수 있음. 
    - 사람의 언어는 같은 의미를 가진 수많은 표현이 존재 -> 특정 목표나 목적을 정확하게 식별 
- 개체명 인식(Named Entity Recognition, NER)
    - 텍스트에서 중요한 정보(인물 이름, 장소, 날짜 등)을 식별하고 분류 
- 감정 분석(Sentiment Analysis)
    - 텍스트의 감정적 색채를 분석하여 긍정적, 부정적 또는 중립적 감증을 파악 
- 의미역 결정(Semantic Role Labeling) 
    - 문장 내에서 주어, 동사, 목적어 등의 역할을 분석하여 문장의 의미 구조를 해석
- 문맥 이해(Contextual Understanding)
    - 대화나 문서 내에서 이전과 이후의 맥락을 통해 현재의 텍스트 의미를 정확히 이해 

- NLU는 챗봇, 음성 인식 SW, 고객 서비스 자동화 도구 등에서 중요하게 활용, 사용자의 요구를 정확하게 파악하고 적절히 대응하기 위해 필수적인 기술 

## 2.2 자연어 생성 (Natural Language Generation, NLG )

- 구조화 된 데이터를 바탕으로 자연스러운 인간 언어의 텍스트를 생성하는 과정 
- 데이터에서 의미를 추출하고, 이를 사용자가 이해할 수 있는 언어로 변환하는 것을 목표 

- `데이터 기반 설명 생성`
    - 대량의 데이터를 분석하고 이를 설명하는 텍스트를 생성 
- `자동 보고서 작성`
    - 기업의 재무 데이터나 시장 조사 결과를 분석하고 요약 보고서를 자동으로 생성 
- `제품 설명 생성`
    - 온라인 쇼핑 사이트에서 제품의 특징을 자동으로 설명하는 텍스트를 생성
- `대화형 시스템 응답 생성`
    - 사용자의 질문이나 요청에 대해 자연스럽고 의미 있는 응답 생성 

## 3. 자연어 처리 프로세스 


1. 데이터 수집 
    - 다양한 소스에서 텍스트 데이터를 수집하고, 이 데이터를 NLP 시스템이 분석하고 학습할 수 있는 형태로 준비 
    - 데이터 수집 방법
        - 웹 크롤링 : python의 BeautifulSoup, Scrapy 같은 도구
        - API 사용
        - 데이터 베이스 엑세스 
        - 데이터 스크래핑 : PDF,Word docs, 전자 테이블 등에서 텍스트 추출, OCR(Optical Character Recognition) 
        - 데이터 구매 
2. 데이터 전처리 
    - 수집 된 데이터는 종종 불완전하거나 노이즈가 많을 수 있다. 
    - 데이터를 분석하기 전에 클리닝, 정규화, 토큰화, 불용어 제거, 어간 추출 및 표제어화 같은 전처리 과정을 거쳐야 한다. 

3. 특성 추출
    - 전처리 된 데이터에서 유용한 특성을 추출
    - 일반적으로 텍스트를 수치화하는 과정을 포함 
4. 모델링 및 훈련
    - 모델링 및 훈련은 자연어처리(NLP) 프로세스에서 핵심적 단계 중 하나, 특정 작업을 수행할 수 있는 모델을 설계, 모델을 설계한 이후 화보한 데이터에서 유용한 패턴을 학습
5. 평가 
    - 자연어 처리 모델의 평가는 모델이 실제 언어 데이터를 얼마나 잘 이해하고 처리하는 지 측정하는 과정
    - NLP 모델 평가에는 여러가지 방법과 지표가있으며, 모델의 유형과 수행하고자 하는 특정 작업에 따라 달라질 수 있다. 

## 4. 자연어 처리 도구 

### 4.1 transformers ( Hugging face 의 Transformers )

- 이 라이브러리는 BERT, GPT,RoBERTa 등과 같은 최신 모델을 포함 
- 텍스트 분류, 질문 답변, 텍스트 생성 등 다양한 NLP 작업을 위한 인터페이스를 제공 
- 사용하기 쉬운 API를 통해 모델의 파인 튜닝과 배포를 간단하게 할 수 있도록 돕는다. 

### 4.2 Dataset ( Hugging face )

- Hugging face 의 라이브러리는 NLP연구와 개발을 위한 다양한 데이터셋에 쉽게 접근할수 있도록 설계
- 많은 표준 데이터셋을 포함
- 메모리 효율적인 데이터 로딩, 캐싱, 데이터 전처리 기능을 제공 

### 4.3 Tokenizers (Hugging face)

- Hugging face의 tokenizers 라이브러리는 고성능 텍스트 토크나이징을 위해 설계
- 매우 빠르고, 다양한 토크나이징 기법(BPE,WordPiece)를 지원, 사전 훈련된 토크나이저를 쉽게 사용할 수 있다. 

### 4.4 Gensim

- python기반 오픈소스 라이브러리, 주로 텍스트 마이닝과 자연어 처리 작업을 위해 사용
- 주로 텍스트 마이닝과 워드 임베딩에 특화
- 리소스가 제한된 환경에서 간단히 사용 가능 

### 4.5 etc...

- NLTK(Natural Language toolkit )
    - 파이썬으로 작성된 강력한 자연어 처리 라이브러리 
    - 교육용으로설계, 텍스트 분석, 토큰화, 품사 태깅, 파싱, 의미 분석 등 다양한 NLP작업 지원
- KoNLPy
    - 한국어 자연어 처리를 위한 라이브러리 ( 파이썬 )
    - 한국어의 특성에 맞춘 다양한 형태소 분석기와 도구들을 제공
