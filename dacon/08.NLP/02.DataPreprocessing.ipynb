# 데이터 전처리와 텍스트 정제

- 데이터를 기계 학습 모델이 이해하고 처리할 수 있는 형태로 변환하는 단계 

1. 텍스트 정제(Text Cleaning) : 
    - 문자, 공백, 특수 기호 등을 제거하여 데이터를 깨끗하게 만드는 과정
2. 토큰화(Tokenization): 
    - 텍스트를 의미 있는 단위로 분리하는 과정, 단어, 구, 또는 문장일 수 있으며, 이 과정을 통해 생성된 토큰들을 처리와 분석을 위해 사용 
3. 정규화(Normalization): 
    - 정규화는 텍스트 데이터 내의 변형된 형태를 표준 형태로 통일하는 과정, "am", "are", "is" 의 동사는 "be"로 통일하거나, 오타나 다른 표기법을 정정하는 것을 포함 
4. 불용어 제거(stop word removal): 
    - 불용어는 자주 등장하지만 분석에는 별 도움이 되지 않는 단어들, "the", "is", "at"등과 같은 단어들, 이 단어들을 제거하면 데이터 분석의 효율성을 높일 수 있다. 
5. 임베딩(Embedding): 
    - 임베딩은 단어나 문장 같은 텍스트를 고차원에서 저차원의 연속적인 벡터로 변환하는 기술, 이 벡터는 텍스트의 의미적, 문맥적 특성을 포착할 수 있으며, 이러한 벡터 표현은 머신러닝 알고리즘에서 사용될 수 있다. 

## 텍스트 정제란?

- 자연어 처리(NLP) 프로젝트에서 초기 단계에서 진행 
- 텍스트 정제 과정에서 학습에 불필요한 텍스트를 제거하거나, 모델이 학습하기 좋은 단어로 텍스트를 대체 

### 노이즈의 예

- 오타, 분법적 오류, 불필요한 구두점, 공백처리, 특수문자, 대소문자, 숫자, 이모티콘, 문장부호, HTML 태그 

### 2. 기본 데이터 정제 

```python
text = "  Hello, World! This is an with #hash and @mentions \t\n"

# 텍스트를 소문자로 변환
lowercase_text = text.lower()
print(lowercase_text)

# 텍스트의 양 끝 공백을 제거
cleaned_text = lowercase_text.strip()
print(cleaned_text)

# 특정 문자 제거: '#'과 '@'
cleaned_text = text.replace('#', '').replace('@', '')
print(cleaned_text)
```

### 3. 정규 표현식을 이용한 데이터 정제

- `re.sub()`함수는 정규 표현식을 사용하여 텍스트에서 패턴에 매칭되는 부분을 다른 문자열로 대체 


```python
import re

# 예시 텍스트
text = "Hello,    World! This\tis an\nexample text."

# 다중 공백을 하나의 공백으로 축소
pattern = r'\s+'
cleaned_text = re.sub(pattern, ' ', text)
print(cleaned_text)

text = "This is an example text with numbers 12345 and 67890."

# 모든 숫자 제거
pattern = r'\d+'
cleaned_text = re.sub(pattern, '', text)
print(cleaned_text)
```

```python
import emoji

# 예시 텍스트
text = "Here is a smiley 😊 and a wink 😉, let's remove them!"

print(emoji.is_emoji("😊"))  # True
print(emoji.emoji_count(text))  # 2
# emoji 라이브러리를 사용하여 이모티콘 제거
cleaned_text = emoji.replace_emoji(text, replace='')

print(cleaned_text)
```

```python
from bs4 import BeautifulSoup

# 예시 HTML 문자열
html_doc = """
<html>
<head>
    <title>The Dormouse's story</title>
</head>
<body>
    <p class="title"><b>The Dormouse's story</b></p>

    <p class="story">Once upon a time there were three little sisters; and their names were
    <a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
    <a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
    <a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
    and they lived at the bottom of a well.</p>

    <p class="story">...</p>
</body>
</html>
"""

# BeautifulSoup 객체 생성, 'html.parser' 사용
soup = BeautifulSoup(html_doc, 'html.parser')

# BeautifulSoup의 get_text() 메소드를 사용하여 HTML에서 모든 텍스트 추출
cleaned_text = soup.get_text()
print(cleaned_text)
```
