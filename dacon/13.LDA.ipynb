# LDA

`LDA(선형 판별 분석)`는 고차원 데이터를 저차원으로 축소하는 데 사용되는 기법으로,   
주로 분류 문제에서 데이터의 클래스를 가장 잘 구분할 수 있는 축을 찾는 데 초점을 맞춥니다.   
이 방법은 클래스 간 분산은 최대화하고, 클래스 내 분산은 최소화하는 방식으로 작동합니다.   
LDA는 데이터 내에 존재하는 클래스 간의 차이를 명확하게 하여 데이터의 시각화 및 해석을 용이하게 하며, 데이터의 복잡성을 감소시킵니다.   

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 데이터 읽어오기
train = pd.read_csv('train.csv')

# 일부 칼럼 제거 
train = train.drop(['index', 'type'], axis = 1)
display(train.head())
    
# 독립 / 종속변수 설정
x_train = train.drop('quality', axis = 1)
y_train = train['quality']

# 데이터에 대한 표준화
scaler = StandardScaler()  
scaler.fit(x_train)
scaled_train = scaler.transform(x_train)
```

## 답변
차원 축소 알고리즘은 크게 **정답(Target/Label) 데이터를 활용하느냐 마느냐**(지도/비지도) 그리고 **데이터를 평면으로 보느냐 입체적으로 보느냐**(선형/비선형)에 따라 나뉩니다.

### 1. PCA (Principal Component Analysis, 주성분 분석) - [가장 많이 쓰임⭐]
- **유형:** 비지도 학습 / 선형 변환
- **원리:** 오로지 '데이터들이 가장 넓게 퍼져있는 방향(분산이 가장 큰 축)'을 찾아 그곳으로 데이터를 투영시킵니다. 정답지(Label)를 전혀 보지 않습니다.
- **장점:** 계산이 매우 빠르고 알고리즘이 단순하며 해석하기 쉽습니다. 
- **언제 쓰나요?** 이미지 압축, 노이즈 제거, 데이터를 머신러닝에 넣기 전 피처(Feature)를 줄이는 **가장 기본적인 전처리 과정**에서 거의 무조건 1순위로 쓰입니다.

### 2. LDA (Linear Discriminant Analysis, 선형 판별 분석)
- **유형:** 지도 학습 / 선형 변환
- **원리:** PCA와 달리 **정답지(Label)를 봅니다.** 같은 정답(클래스)을 가진 데이터끼리는 최대한 뭉치게 하고, 다른 정답을 가진 데이터들끼리는 최대한 멀리 떨어지도록(분산되도록) 하는 새로운 축을 찾습니다.
- **언제 쓰나요?** 분류(Classification) 모델을 만들기 전에, 모델이 정답을 더 쉽게 맞출 수 있도록 데이터를 예쁘게 정렬해주는 목적으로 주로 쓰입니다.

### 3. t-SNE (t-Distributed Stochastic Neighbor Embedding)
- **유형:** 비지도 학습 / 비선형 변환
- **원리:** 데이터 포인트들 사이의 거리를 비율(확률)로 변환하여, 고차원에서 가까웠던 데이터들은 저차원(2D, 3D)에서도 가깝게 묶이도록 복잡하게 곡면을 펴줍니다.
- **장점:** 시각화 결과물이 정말 아름답고 군집이 명확하게 보입니다.
- **단점:** 계산이 굉장히 느리고 매번 돌릴 때마다 결과가 바뀔 수 있어 머신러닝의 입력 피처로는 잘 쓰지 않습니다.
- **언제 쓰나요?** 데이터를 모델에 넣기 전에 사람들이 눈으로 보고 **데이터의 숨겨진 군집 구조를 파악(시각화)** 할 때 가장 많이 사랑받는 알고리즘입니다.

> **💡 결론:** 일반적인 머신러닝 파이프라인(전처리 -> 피처 축소 -> 모델 학습)에서는 압도적으로 **PCA**가 가장 많이 활용됩니다. 시각화를 통한 데이터 탐색(EDA) 단계에서는 **t-SNE**가 많은 사랑을 받습니다.
