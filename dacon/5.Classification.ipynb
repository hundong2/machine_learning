## 모델 성능 평가 지표의 정의와 활용 



- 분류 문제의 대표적인 성능 평가 지표 ( Accuracy, Precision, Recall, F1-score)를 confusion matrix를 통하여 정확한 개념을 이해 합니다. 

# 검증 데이터, 학습 데이터 분할 코드 

```python
from sklearn.model_selection import train_test_split

# 특성과 타겟 변수 분리
train_x = train_diabets.drop(['ID','Outcome'], axis=1)
train_y = train_diabets['Outcome']

# 학습 데이터와 검증 데이터로 분리
X_train, X_val, y_train, y_val = train_test_split(train_x, train_y, test_size=0.25, random_state=42)

# 분리된 데이터의 크기 확인

display("분할 후의 학습 데이터셋 크기:", X_train.shape )
display("분할 후의 검증 데이터셋 크기:", X_val.shape )
```

```
'분할 후의 학습 데이터셋 크기:'
(489, 8)
'분할 후의 검증 데이터셋 크기:'
(163, 8)
```

## DecisionTreeClassifier 모델 객체 생성 및 train_diabets 25% 데이터 검증으로 분할 

```python
from sklearn.tree import DecisionTreeClassifier

# Decision Tree 모델 생성 및 학습
clf_diabets = DecisionTreeClassifier(random_state=42)
clf_diabets.fit(X_train, y_train)

# 모델을 사용하여 검증 데이터에 대한 예측 수행
y_pred = clf_diabets.predict(X_val) 

print("첫 10개의 예측 데이터 :", y_pred[:10])
```

--- 

```
첫 10개의 예측 데이터 : [0 0 1 0 0 1 0 0 1 1]
```

## 분할 데이터 셋에대한 성능 검증 

- accuracy, precision recall, f1 score를 통하여 분류 성능 평가
- `sklearn.metrics`의 confusion_matrix, accuracy_score, precision_score, recall_score, f1_score 호출 

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

y_true = y_val
accuracy_diabets_DT = accuracy_score(y_true, y_pred) 
precision_diabets_DT =  precision_score(y_true, y_pred)
recall_diabets_DT =  recall_score(y_true, y_pred)
f1_score_diabets_DT =  f1_score(y_true, y_pred)

display(f"Accuracy:{accuracy_diabets_DT}")
display(f"precision:{precision_diabets_DT}")
display(f"recall: {recall_diabets_DT}")
display(f"f1_score: {f1_score_diabets_DT}")

# Confusion Matrix 얻기
cf_mat = confusion_matrix(y_val, y_pred)
display("Confusion Matrix\n", cf_mat) 
```


## average 매개변수의 의미 

- accuracy_score, precision_score, recall_score, f1_score
    - `average` 매개변수 : `이진분류`와 `다중 클래스 분류`에서 모델 성능을 집계하는 방식을 결정 

### 이진분류의 경우 

- 이진 분류에서는 average 매개변수를 지정할 필요가 없음. ( binary 가 기본값 ). 함수들은 양성 클래스에 대한 성능 지표를 자동으로 계산. 

### 다중 클래스 분류의 경우 

- average 매개변수를 통해 다양한 집계 방식을 선택
- macro : 클래스 간에 균등한 가중치를 부여하여 평균을 계산, 클래스 불균형을 고려하지 않습니다. 
- weighted : 각 클래스 샘플 수에 따라 가중치를 부여하여 평균을 계산, 클래스 불균형을 고려 
- micro : 전체 TP, FN, FP를 집계하여 계산. 클래스 불균형에 영향을 받지 않으며, 전체적인 성능을 나타냄

## multi 

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 학습 데이터와 검증 데이터로 분할
X_train, X_val, y_train, y_val = train_test_split(train_x, train_y, test_size=0.25, random_state=42)

# Decision Tree Classifier 모델 생성 및 학습
clf_wine_DT = DecisionTreeClassifier(random_state=42)
clf_wine_DT.fit(X_train, y_train)

# 검증 데이터에 대한 예측 수행
y_pred = clf_wine_DT.predict(X_val)

# 성능 지표 계산 및 출력
accuracy_wine_DT =  accuracy_score(y_val, y_pred)
precision_wine_DT =  precision_score(y_val,y_pred, average='macro')
recall_wine_DT =  recall_score(y_val, y_pred, average='macro')
f1_score_wine_DT =  f1_score(y_val, y_pred, average='macro')

display(f"Accuracy:{accuracy_wine_DT}")
display(f"precision:{precision_wine_DT}")
display(f"recall: {recall_wine_DT}")
display(f"f1_score: {f1_score_wine_DT}")
```
