## 데이터 분류 문제 유형 

```python
import pandas as pd
train_diabets = pd.read_csv('train_diabets.csv')
train_diabets.head()

# 타겟 변수 범주가 몇개로 구성이 되어 있는지?
unique_data = train_diabets['Outcome'].unique()
print(unique_data)
# [0 1]
# target 칼럼에서 범주별로 몇개의 데이터가 있는지
target_freq = train_diabets['Outcome'].value_counts()
print(target_freq)
#0    424
#1    228
#Name: Outcome, dtype: int64

```

## Scikit-learn

```python
from sklearn.datasets import load_iris  
import pandas as pd 

iris = load_iris()
 

# 데이터셋을 DataFrame으로 변환합니다.
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

# test 

```python
import pandas as pd
train = pd.read_csv('train_wine.csv')
test = pd.read_csv('test_wine.csv')

train_x = train.drop('quality', axis = 1 )
train_y = train['quality'] 

display(train.head())
display(train_x.head())
display(test.head())
```

```python
from sklearn.ensemble import RandomForestClassifier 

model_rf = RandomForestClassifier(random_state = 42) 

# 모델 학습
model_rf.fit(train_x, train_y)
# 테스트 데이터에 대한 예측 
predict = model_rf.predict(test)
print(predict[:10])
```
