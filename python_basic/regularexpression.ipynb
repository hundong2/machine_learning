## ğŸ“‹ ë¬¸ì œ ì„¤ëª…

## ğŸ¯ ëª©í‘œ

1. **ì²« ë²ˆì§¸ ê·¸ë£¹**: `Name: ` ë‹¤ìŒì— ë‚˜ì˜¤ëŠ” ì´ë¦„ì„ ìº¡ì²˜
2. **ë‘ ë²ˆì§¸ ê·¸ë£¹**: `Age: ` ë‹¤ìŒì— ë‚˜ì˜¤ëŠ” ë‚˜ì´(ìˆ«ì)ë¥¼ ìº¡ì²˜
3. ì¶”ì¶œëœ ì •ë³´ë¥¼ ì‚¬ìš©í•˜ì—¬ ì›í•˜ëŠ” í˜•íƒœë¡œ ì¶œë ¥

```python
import re

print("=== ì´ë¦„ê³¼ ë‚˜ì´ ì¶”ì¶œ ì‹¤ìŠµ ===")

# í…ŒìŠ¤íŠ¸ ë°ì´í„°
text_data = """
Personal Information:
Name: John Doe, Age: 45 years old
Contact: Name: Jane Smith, Age: 32 years old  
Record: Name: Bob Brown, Age: 30 years old
Additional: Name: Alice Johnson, Age: 28 years old
"""

print("ì›ë³¸ ë°ì´í„°:")
print(text_data)

print("\n=== 1ë‹¨ê³„: íŒ¨í„´ ë¶„ì„ ===")
print("ì°¾ì•„ì•¼ í•  íŒ¨í„´:")
print("- 'Name: ' ë‹¤ìŒì— ì´ë¦„ (ë¬¸ìì™€ ê³µë°±)")
print("- 'Age: ' ë‹¤ìŒì— ë‚˜ì´ (ìˆ«ì)")
print("- ê°ê°ì„ ë³„ë„ ê·¸ë£¹ìœ¼ë¡œ ìº¡ì²˜")

# íŒ¨í„´ êµ¬ì„± ìš”ì†Œë³„ ì„¤ëª…
print("\níŒ¨í„´ êµ¬ì„±:")
print("- Name: ë¶€ë¶„: 'Name: '")
print("- ì´ë¦„ ê·¸ë£¹: '([A-Za-z\\s]+)' - ì˜ë¬¸ìì™€ ê³µë°±ì„ 1ê°œ ì´ìƒ")
print("- ì¤‘ê°„ ë¶€ë¶„: ', Age: '")
print("- ë‚˜ì´ ê·¸ë£¹: '(\\d+)' - ìˆ«ìë¥¼ 1ê°œ ì´ìƒ")

# ì •ê·œí‘œí˜„ì‹ íŒ¨í„´
pattern = r"Name: ([A-Za-z\s]+), Age: (\d+)"

print(f"\nì™„ì„±ëœ íŒ¨í„´: {pattern}")

print("\n=== 2ë‹¨ê³„: íŒ¨í„´ í…ŒìŠ¤íŠ¸ ===")

# findallì„ ì‚¬ìš©í•˜ì—¬ ëª¨ë“  ë§¤ì¹­ ì°¾ê¸°
matches = re.findall(pattern, text_data)

print("findall() ê²°ê³¼ (íŠœí”Œ ë¦¬ìŠ¤íŠ¸):")
for i, (name, age) in enumerate(matches, 1):
    print(f"  {i}. ì´ë¦„: '{name}', ë‚˜ì´: '{age}'")

print("\n=== 3ë‹¨ê³„: ì›í•˜ëŠ” í˜•íƒœë¡œ ì¶œë ¥ ===")

print("ìš”êµ¬ì‚¬í•­ì— ë§ëŠ” ì¶œë ¥:")
for name, age in matches:
    # ì´ë¦„ ëì˜ ê³µë°± ì œê±°
    clean_name = name.strip()
    print(f"Name: {clean_name}, Age: {age} years old")
```

```python
print("\n=== 4ë‹¨ê³„: finditer()ë¥¼ ì‚¬ìš©í•œ ë” ìì„¸í•œ ì •ë³´ ===")

# finditerë¥¼ ì‚¬ìš©í•˜ì—¬ ë§¤ì¹­ ìœ„ì¹˜ì™€ ìƒì„¸ ì •ë³´ í™•ì¸
matches_detailed = re.finditer(pattern, text_data)

print("finditer() ê²°ê³¼ (ìƒì„¸ ì •ë³´ í¬í•¨):")
for i, match in enumerate(matches_detailed, 1):
    print(f"  {i}. ì „ì²´ ë§¤ì¹­: '{match.group()}'")
    print(f"     - ì²« ë²ˆì§¸ ê·¸ë£¹ (ì´ë¦„): '{match.group(1)}'")
    print(f"     - ë‘ ë²ˆì§¸ ê·¸ë£¹ (ë‚˜ì´): '{match.group(2)}'")
    print(f"     - ë§¤ì¹­ ìœ„ì¹˜: {match.start()}-{match.end()}")
    print()

print("=== 5ë‹¨ê³„: ë‹¤ì–‘í•œ í˜•íƒœì˜ ë°ì´í„° ì²˜ë¦¬ ===")

# ë” ë³µì¡í•œ í…ŒìŠ¤íŠ¸ ë°ì´í„°
complex_data = """
Employee Records:
Name: John Michael Doe, Age: 45 years old
Name: Mary Jane Smith, Age: 32 years old
Name: Bob, Age: 30 years old
Name: Alice Marie Johnson-Brown, Age: 28 years old
Invalid: Age: 25 years old, Name: Missing
Name: David Lee, Age: unknown years old
"""

print("ë³µì¡í•œ ë°ì´í„°:")
print(complex_data)

# ê°œì„ ëœ íŒ¨í„´ (í•˜ì´í”ˆë„ í¬í•¨)
improved_pattern = r"Name: ([A-Za-z\s\-]+), Age: (\d+)"

print(f"\nê°œì„ ëœ íŒ¨í„´: {improved_pattern}")
print("ê°œì„ ì‚¬í•­: í•˜ì´í”ˆ(-) ì¶”ê°€ë¡œ ë³µí•©ì„±ëª… ì²˜ë¦¬")

improved_matches = re.findall(improved_pattern, complex_data)

print("\nê°œì„ ëœ íŒ¨í„´ ê²°ê³¼:")
for name, age in improved_matches:
    clean_name = name.strip()
    print(f"Name: {clean_name}, Age: {age} years old")

print("\n=== 6ë‹¨ê³„: ëª…ëª…ëœ ê·¸ë£¹ ì‚¬ìš© ===")

# ëª…ëª…ëœ ê·¸ë£¹ì„ ì‚¬ìš©í•œ íŒ¨í„´
named_pattern = r"Name: (?P<name>[A-Za-z\s\-]+), Age: (?P<age>\d+)"

print(f"ëª…ëª…ëœ ê·¸ë£¹ íŒ¨í„´: {named_pattern}")

named_matches = re.finditer(named_pattern, complex_data)

print("ëª…ëª…ëœ ê·¸ë£¹ ê²°ê³¼:")
for match in named_matches:
    name = match.group('name').strip()
    age = match.group('age')
    print(f"Name: {name}, Age: {age} years old")
    print(f"  - ì´ë¦„ë§Œ: {match.group('name').strip()}")
    print(f"  - ë‚˜ì´ë§Œ: {match.group('age')}")
    print()
```

## ğŸ” íŒ¨í„´ ìƒì„¸ ë¶„ì„

### ê¸°ë³¸ íŒ¨í„´: `r"Name: ([A-Za-z\s]+), Age: (\d+)"`

| êµ¬ì„±ìš”ì†Œ | ì„¤ëª… | ì—­í•  |
|---------|------|------|
| `Name: ` | ë¦¬í„°ëŸ´ ë¬¸ìì—´ | 'Name: ' ë¬¸ìì—´ê³¼ ì •í™•íˆ ì¼ì¹˜ |
| `(` | ê·¸ë£¹ ì‹œì‘ | ì²« ë²ˆì§¸ ìº¡ì²˜ ê·¸ë£¹ ì‹œì‘ |
| `[A-Za-z\s]+` | ë¬¸ì í´ë˜ìŠ¤ | ì˜ë¬¸ìì™€ ê³µë°± 1ê°œ ì´ìƒ |
| `)` | ê·¸ë£¹ ì¢…ë£Œ | ì²« ë²ˆì§¸ ìº¡ì²˜ ê·¸ë£¹ ì¢…ë£Œ |
| `, Age: ` | ë¦¬í„°ëŸ´ ë¬¸ìì—´ | ', Age: ' ë¬¸ìì—´ê³¼ ì •í™•íˆ ì¼ì¹˜ |
| `(` | ê·¸ë£¹ ì‹œì‘ | ë‘ ë²ˆì§¸ ìº¡ì²˜ ê·¸ë£¹ ì‹œì‘ |
| `\d+` | ìˆ«ì í´ë˜ìŠ¤ | ìˆ«ì 1ê°œ ì´ìƒ |
| `)` | ê·¸ë£¹ ì¢…ë£Œ | ë‘ ë²ˆì§¸ ìº¡ì²˜ ê·¸ë£¹ ì¢…ë£Œ |

### ê°œì„ ëœ íŒ¨í„´: `r"Name: ([A-Za-z\s\-]+), Age: (\d+)"`

**ì¶”ê°€ëœ ê¸°ëŠ¥:**
- `\-`: í•˜ì´í”ˆ ë¬¸ì ì¶”ê°€ë¡œ ë³µí•©ì„±ëª… ì²˜ë¦¬ (Johnson-Brown)
- ë” ë‹¤ì–‘í•œ ì´ë¦„ í˜•íƒœ ì§€ì›

### ëª…ëª…ëœ ê·¸ë£¹ íŒ¨í„´: `r"Name: (?P<name>[A-Za-z\s\-]+), Age: (?P<age>\d+)"`

**ì¥ì :**
- ê·¸ë£¹ì— ì˜ë¯¸ìˆëŠ” ì´ë¦„ ë¶€ì—¬
- `match.group('name')`, `match.group('age')` ë°©ì‹ìœ¼ë¡œ ì ‘ê·¼
- ì½”ë“œ ê°€ë…ì„± í–¥ìƒ

```python
print("=== 7ë‹¨ê³„: ë°ì´í„° ë³€í™˜ ë° í¬ë§·íŒ… ===")

# ì¶”ì¶œí•œ ë°ì´í„°ë¥¼ ë‹¤ë¥¸ í˜•íƒœë¡œ ë³€í™˜
transformation_data = """
Profile: Name: Emma Watson, Age: 34 years old
Profile: Name: Daniel Radcliffe, Age: 35 years old
Profile: Name: Rupert Grint, Age: 36 years old
"""

pattern = r"Name: ([A-Za-z\s]+), Age: (\d+)"
matches = re.findall(pattern, transformation_data)

print("ì›ë³¸ ë°ì´í„°:")
print(transformation_data)

print("ë‹¤ì–‘í•œ í˜•íƒœë¡œ ë³€í™˜:")

# 1. ì‚¬ì „(Dictionary) í˜•íƒœë¡œ ë³€í™˜
print("\n1. ì‚¬ì „ í˜•íƒœ:")
people_dict = {}
for name, age in matches:
    clean_name = name.strip()
    people_dict[clean_name] = int(age)

for name, age in people_dict.items():
    print(f"  '{name}': {age}ì„¸")

# 2. JSON ìŠ¤íƒ€ì¼ë¡œ ì¶œë ¥
print("\n2. JSON ìŠ¤íƒ€ì¼:")
import json
people_list = []
for name, age in matches:
    people_list.append({
        "name": name.strip(),
        "age": int(age)
    })

print(json.dumps(people_list, indent=2, ensure_ascii=False))

# 3. CSV í˜•íƒœë¡œ ë³€í™˜
print("\n3. CSV í˜•íƒœ:")
print("ì´ë¦„,ë‚˜ì´")
for name, age in matches:
    clean_name = name.strip()
    print(f"{clean_name},{age}")

# 4. ë‚˜ì´ëŒ€ë³„ ë¶„ë¥˜
print("\n4. ë‚˜ì´ëŒ€ë³„ ë¶„ë¥˜:")
age_groups = {"20ëŒ€": [], "30ëŒ€": [], "40ëŒ€": []}

for name, age in matches:
    clean_name = name.strip()
    age_num = int(age)
    
    if 20 <= age_num < 30:
        age_groups["20ëŒ€"].append(clean_name)
    elif 30 <= age_num < 40:
        age_groups["30ëŒ€"].append(clean_name)
    elif 40 <= age_num < 50:
        age_groups["40ëŒ€"].append(clean_name)

for age_group, names in age_groups.items():
    if names:
        print(f"  {age_group}: {', '.join(names)}")

print("\n=== 8ë‹¨ê³„: ì‹¤ìŠµ ë¬¸ì œ ===")

challenge_data = """
Student Records:
Name: ê¹€ì² ìˆ˜, Age: 22 years old
Name: ì´ì˜í¬, Age: 21 years old  
Name: ë°•ë¯¼ìˆ˜, Age: 23 years old
Name: John Smith, Age: 25 years old
"""

print("ë„ì „ ê³¼ì œ: í•œê¸€ ì´ë¦„ë„ ì²˜ë¦¬í•  ìˆ˜ ìˆëŠ” íŒ¨í„´ ë§Œë“¤ê¸°")
print("í…ŒìŠ¤íŠ¸ ë°ì´í„°:")
print(challenge_data)

# í•œê¸€ì„ í¬í•¨í•œ íŒ¨í„´ (\wëŠ” í•œê¸€ë„ ë§¤ì¹­)
korean_pattern = r"Name: ([\w\s\-]+), Age: (\d+)"

print(f"\ní•œê¸€ ì§€ì› íŒ¨í„´: {korean_pattern}")
print("\\wëŠ” í•œê¸€, ì˜ë¬¸, ìˆ«ì, ì–¸ë”ìŠ¤ì½”ì–´ë¥¼ ëª¨ë‘ í¬í•¨")

korean_matches = re.findall(korean_pattern, challenge_data)

print("\nê²°ê³¼:")
for name, age in korean_matches:
    clean_name = name.strip()
    print(f"Name: {clean_name}, Age: {age} years old")
```

# metacharacters for regular expression

- `.` : ì„ì˜ì˜ ë‹¨ì¼ ë¬¸ìì™€ ì¼ì¹˜  
- `^` : ë¬¸ìì—´ì˜ ì‹œì‘ ë¶€ë¶„ì„ ì˜ë¯¸    
- `%` : ë¬¸ìì—´ì˜ ë ë¶€ë¶„ì„ ì˜ë¯¸
- `*` : ì•ì˜ ë¬¸ìê°€ 0ë²ˆ ì´ìƒ ë°˜ë³µë¨ì„ ì˜ë¯¸
- `+` : ì•ì˜ ë¬¸ìê°€ 1ë²ˆ ì´ìƒ ë°˜ë³µë¨ì„ ì˜ë¯¸
- `?` : ì•ì˜ ë¬¸ìê°€ 0ë²ˆ ë˜ëŠ” 1ë²ˆ ë‚˜íƒ€ë‚¨ì„ ì˜ë¯¸
- `|` : `OR` ì—°ì‚°ì„ ì˜ë¯¸í•˜ë©°, ì˜ˆ: a|b ëŠ” `a` or `b`ì™€ ì¼ì¹˜
- `()` : ê·¸ë£¹í™”ë¥¼ ì˜ë¯¸í•˜ë©°, ë¶€ë¶„ íŒ¨í„´ì„ ë¬¶ì–´ ì¤ë‹ˆë‹¤.
- `{n}` : në²ˆ ë°˜ë³µ ë˜ëŠ” íŒ¨í„´ì— ì¼ì¹˜, a{3} : aaa ì— ì¼ì¹˜
- `{n,}` : në²ˆ ì´ìƒ ë°˜ë³µë˜ëŠ” íŒ¨í„´ì— ì¼ì¹˜

```python
import re
# ì…ë ¥ í…ìŠ¤íŠ¸
text = """
This is a Test. The result should include This sentence. Another test? The code should ignore this. Today is a good day.
"""

pattern = r"T[^.]*\." # ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•˜ì„¸ìš”.
sentences = re.findall(pattern, text)
print(sentences)  # ['This is a Test.', 'The result should include This sentence.', 'Today is a good day.']
```

# ì •ê·œí‘œí˜„ì‹ ê·¸ë£¹í™”(Grouping)ì™€ ë°±ì°¸ì¡°(Backreferences)

ì •ê·œí‘œí˜„ì‹ì—ì„œ **ê·¸ë£¹í™”**ëŠ” íŒ¨í„´ì˜ ì¼ë¶€ë¥¼ ë¬¶ì–´ì„œ í•˜ë‚˜ì˜ ë‹¨ìœ„ë¡œ ì²˜ë¦¬í•˜ëŠ” ê¸°ëŠ¥ì…ë‹ˆë‹¤.
**ë°±ì°¸ì¡°**ëŠ” ì•ì—ì„œ ì •ì˜í•œ ê·¸ë£¹ì„ ë‹¤ì‹œ ì°¸ì¡°í•˜ì—¬ ê°™ì€ íŒ¨í„´ì„ ë°˜ë³µ ë§¤ì¹­í•˜ëŠ” ê¸°ëŠ¥ì…ë‹ˆë‹¤.

## ì£¼ìš” ê°œë…

### ê·¸ë£¹í™” `()`
- íŒ¨í„´ì„ ê·¸ë£¹ìœ¼ë¡œ ë¬¶ì–´ í•˜ë‚˜ì˜ ë‹¨ìœ„ë¡œ ì²˜ë¦¬
- ë§¤ì¹­ëœ ê·¸ë£¹ ë‚´ìš©ì„ ë³„ë„ë¡œ ì¶”ì¶œ ê°€ëŠ¥
- ë°±ì°¸ì¡°ì—ì„œ ì‚¬ìš©í•  ìˆ˜ ìˆë„ë¡ ë²ˆí˜¸ê°€ í• ë‹¹ë¨

### ë°±ì°¸ì¡° `\1, \2, \3...`
- ì•ì—ì„œ ì •ì˜í•œ ê·¸ë£¹ì˜ ë§¤ì¹­ ê²°ê³¼ë¥¼ ë‹¤ì‹œ ì‚¬ìš©
- `\1`ì€ ì²« ë²ˆì§¸ ê·¸ë£¹, `\2`ëŠ” ë‘ ë²ˆì§¸ ê·¸ë£¹ì„ ì˜ë¯¸
- ì¤‘ë³µëœ íŒ¨í„´ì´ë‚˜ ëŒ€ì¹­ì ì¸ êµ¬ì¡°ë¥¼ ì°¾ì„ ë•Œ ìœ ìš©

```python
import re

print("=== 1. ê¸°ë³¸ ê·¸ë£¹í™” ì˜ˆì œ ===")

# ì „í™”ë²ˆí˜¸ì—ì„œ ì§€ì—­ë²ˆí˜¸, êµ­ë²ˆ, ë²ˆí˜¸ë¥¼ ë¶„ë¦¬
phone_text = "ì—°ë½ì²˜: 010-1234-5678, 02-9876-5432, 031-555-7890"

# ê·¸ë£¹í™”ë¥¼ ì‚¬ìš©í•˜ì—¬ ì „í™”ë²ˆí˜¸ êµ¬ì„±ìš”ì†Œ ë¶„ë¦¬
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone_matches = re.findall(phone_pattern, phone_text)

print("ì „í™”ë²ˆí˜¸ ë¶„ë¦¬ ê²°ê³¼:")
for match in phone_matches:
    area_code, exchange, number = match
    print(f"  ì§€ì—­ë²ˆí˜¸: {area_code}, êµ­ë²ˆ: {exchange}, ë²ˆí˜¸: {number}")

print("\nê·¸ë£¹í™” ì—†ì´ ë§¤ì¹­:")
simple_pattern = r"\d{2,3}-\d{3,4}-\d{4}"
simple_matches = re.findall(simple_pattern, phone_text)
print(f"  ì „ì²´ ì „í™”ë²ˆí˜¸: {simple_matches}")

print("\n=== 2. ì´ë©”ì¼ ì£¼ì†Œ ê·¸ë£¹í™” ===")

email_text = "ì´ë©”ì¼: john.doe@example.com, admin@company.org, user123@domain.net"

# ì‚¬ìš©ìëª…ê³¼ ë„ë©”ì¸ì„ ë¶„ë¦¬
email_pattern = r"([A-Za-z0-9._%+-]+)@([A-Za-z0-9.-]+\.[A-Za-z]{2,})"
email_matches = re.findall(email_pattern, email_text)

print("ì´ë©”ì¼ ë¶„ë¦¬ ê²°ê³¼:")
for username, domain in email_matches:
    print(f"  ì‚¬ìš©ìëª…: {username}, ë„ë©”ì¸: {domain}")
```

```python
print("\n=== 3. ë°±ì°¸ì¡°(Backreferences) ê¸°ë³¸ ì˜ˆì œ ===")

# ì¤‘ë³µëœ ë‹¨ì–´ ì°¾ê¸°
text_with_duplicates = "This is is a test test. Hello hello world. Good good morning."

# ë°±ì°¸ì¡°ë¥¼ ì‚¬ìš©í•˜ì—¬ ì—°ì†ëœ ì¤‘ë³µ ë‹¨ì–´ ì°¾ê¸°
duplicate_pattern = r"\b(\w+)\s+\1\b"
duplicates = re.findall(duplicate_pattern, text_with_duplicates)
print("ì¤‘ë³µëœ ë‹¨ì–´:")
print(f"  ë°œê²¬ëœ ì¤‘ë³µ ë‹¨ì–´: {duplicates}")

# ì¤‘ë³µëœ ë‹¨ì–´ê°€ í¬í•¨ëœ ì „ì²´ ë§¤ì¹­ ë³´ê¸°
duplicate_full = re.finditer(duplicate_pattern, text_with_duplicates)
print("  ì „ì²´ ë§¤ì¹­:")
for match in duplicate_full:
    print(f"    '{match.group()}' - ë‹¨ì–´: '{match.group(1)}'")

print("\n=== 4. HTML íƒœê·¸ ë§¤ì¹­ ===")

html_text = "<div>ë‚´ìš©</div> <span>ê¸€ì</span> <p>ë¬¸ë‹¨</p> <div>ë‹¤ë¥¸ë‚´ìš©</div>"

# ì—¬ëŠ” íƒœê·¸ì™€ ë‹«ëŠ” íƒœê·¸ê°€ ê°™ì€ HTML ìš”ì†Œ ì°¾ê¸°
html_pattern = r"<(\w+)>.*?</\1>"
html_matches = re.findall(html_pattern, html_text)
print("ë§¤ì¹­ëœ HTML íƒœê·¸:")
print(f"  íƒœê·¸ë“¤: {html_matches}")

# ì „ì²´ HTML ìš”ì†Œ ì¶”ì¶œ
html_full = re.finditer(html_pattern, html_text)
print("  ì „ì²´ HTML ìš”ì†Œ:")
for match in html_full:
    print(f"    {match.group()} - íƒœê·¸: {match.group(1)}")
```
